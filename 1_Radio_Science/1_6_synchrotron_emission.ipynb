## 1.6 同步辐射

前一节讨论的热辐射把射电亮度与温度、电离气体和光深联系起来；同步辐射则代表射电天文学中最重要的一类非热连续谱机制。许多经典射电源之所以在厘米波和米波段异常明亮，并不是因为它们拥有不可思议的热温度，而是因为其中存在磁场和相对论电子。射电图像中的喷流、射电瓣、超新星遗迹壳层、星系盘非热连续谱和星系团弥散射电晕，都离不开同步辐射。

同步辐射这一节的核心不是罗列公式，而是建立一条判断链：磁场中的相对论电子如何辐射，电子能谱如何转化为射电谱指数，偏振为什么能追踪磁场几何，谱老化如何记录粒子冷却历史，以及这些性质为什么要求干涉观测具有多频率、宽带、偏振和空间尺度恢复能力。


### 1.6.1 单个相对论电子为什么会辐射

同步辐射的物理起点是带电粒子在磁场中的螺旋运动。洛伦兹力改变粒子速度方向，因而产生横向加速度；只要带电粒子被加速，就会辐射电磁波。在非相对论极限下，这种辐射称为回旋辐射，频率接近回旋频率。当天体中的电子达到相对论能量时，辐射会被强烈束射到电子瞬时速度方向附近，观测到的频率也会因相对论效应显著提高，这就是同步辐射。

若电子洛伦兹因子为 $\gamma$，磁场中垂直于速度方向的分量为 $B_\perp$，则同步辐射总功率的标度可以写成

$$
P_{\rm syn}\propto \gamma^2 B_\perp^2.
$$

更完整地说，辐射损失率可写为

$$
P_{\rm syn}={4\over 3}\sigma_T c\,\beta^2\gamma^2 U_B\sin^2\alpha_p,
$$

其中 $\sigma_T$ 是 Thomson 截面，$U_B=B^2/8\pi$ 是磁场能量密度，$\alpha_p$ 是电子速度与磁场方向之间的 pitch angle。这个式子说明，高能电子的能量损失非常快，而且磁场越强，损失越强。

单个电子的辐射并不是一条无限窄的谱线，而是在一个特征频率附近具有较宽的连续谱。常用特征频率为

$$
\nu_c\simeq {3e\over 4\pi m_e c}\,B_\perp\gamma^2.
$$

若把磁场写成微高斯量级，可以记成数量级关系

$$
\nu_c\simeq 4.2\,B_{\mu{\rm G}}\gamma^2\ {\rm Hz}.
$$

因此，在 $10\,\mu{\rm G}$ 的磁场中，$\gamma\sim 10^4$ 的电子主要在 GHz 量级辐射。同一片天体中的不同观测频率，实际上对应着不同能量电子的辐射贡献；多频射电观测因此具有“电子能量分层诊断”的意义。

### 1.6.2 从电子能谱到射电谱指数

真实天体中不会只有一个电子在辐射。喷流、射电瓣、超新星遗迹和星系盘中包含大量相对论电子，它们通常由激波、湍流或磁重联等过程加速。许多加速过程会产生近似幂律的电子能量分布：

$$
N(E)\,dE=N_0 E^{-p}\,dE.
$$

当这群电子在磁场中产生光学薄同步辐射时，观测到的通量密度也呈幂律：

$$
S_\nu\propto \nu^\alpha,
\qquad
\alpha=-{p-1\over 2}.
$$

这个关系非常重要，因为它把不可直接测量的电子能谱指数 $p$ 转化为可观测的射电谱指数 $\alpha$。典型光学薄同步辐射常有 $\alpha\sim -0.5$ 到 $-1.0$，比热自由-自由辐射的光学薄谱指数 $\alpha\simeq -0.1$ 陡得多。因而，宽带通量测量和谱指数图是区分热与非热连续谱的基本工具。

高能电子辐射得更强，同时也冷却得更快。随着时间推移，高频端对应的高能电子会先耗尽，谱在高频处发生变陡，这就是谱老化。若只考虑同步损失，电子在频率 $\nu$ 附近的典型寿命随磁场和频率近似满足 $t_{\rm syn}\propto B^{-3/2}\nu^{-1/2}$；若还考虑与宇宙微波背景等光子场的逆康普顿损失，则有效损失率取决于 $B^2+B_{\rm IC}^2$，其中 $B_{\rm IC}\simeq 3.25(1+z)^2\,\mu{\rm G}$。这解释了为什么射电瓣、星系团射电晕和老年电子群体常在高频变暗，而低频阵列能看到更古老的非热结构。

### 1.6.3 偏振、亮温与低频翻转

同步辐射的另一个关键特征是线偏振。单个电子辐射的电矢量与磁场投影有确定关系；若电子能谱为幂律，均匀磁场中的光学薄同步辐射理论最大线偏振度为

$$
\Pi_{\max}={p+1\over p+7/3}.
$$

对常见的 $p\sim 2$ 到 $3$，这个最大值可达约 $70\%$。真实天体中观测到的偏振度通常低得多，因为望远镜波束内可能混合多个磁场方向，湍流会打乱有序场，法拉第旋转和法拉第退偏也会改变不同频率下的偏振叠加。尽管理想值很少达到，偏振成像仍然是研究磁场几何、喷流结构和传播介质的核心方法。

亮温也能提供重要判断。普通热源很难长时间维持极高亮温；若由图像推得的 $T_b$ 远超热平衡温度，非热机制就变得不可避免。致密同步辐射源还可能受到同步自吸收影响：低频光子在源内部被同一群相对论电子吸收，使谱在低频端从下降幂律翻转为上升谱。理想均匀自吸收源的光学厚谱可接近 $S_\nu\propto\nu^{5/2}$，而外部自由-自由吸收也能造成低频翻转。因此，低频谱形的变化本身并不能自动说明是哪一种吸收机制，必须结合角尺度、亮温、环境和多频结构共同判断。


### 1.6.4 观测上如何识别同步辐射

同步辐射的识别通常依赖多条证据同时成立。最直接的证据是光学薄幂律连续谱；若谱指数明显为负，并且比自由-自由辐射更陡，就强烈指向非热电子。线偏振提供第二条证据，因为热自由-自由辐射本身几乎不产生显著线偏振，而同步辐射天然带有偏振。高亮温提供第三条证据，尤其是在紧致射电核、喷流和脉冲星风云中。空间上的谱指数变化则提供更深入的信息：核心、喷流、热点和射电瓣常具有不同谱指数，反映粒子注入、输运和冷却的历史。

这些诊断也直接影响干涉观测策略。研究同步辐射不能只追求单一频段的漂亮图像，而需要可靠的多频通量标定、匹配的 $uv$ 覆盖和共同分辨率下的谱指数成图；做偏振研究时，还需要处理泄漏项、绝对偏振角、法拉第旋转和带宽退偏；研究弥散射电瓣或星系晕时，还必须保证短基线对大尺度结构的敏感性。同步辐射因而是连接辐射机制和干涉观测设计的一个很好例子。


### 1.6.5 典型同步辐射源

同步辐射出现的前提是“磁场 + 相对论电子”，而这在宇宙中并不罕见。超新星遗迹的激波可以加速电子并形成壳状非热辐射；脉冲星风云由高速粒子风和磁场共同塑造；活动星系核喷流把能量输运到星系外，形成热点和射电瓣；星系盘中的宇宙线电子在微高斯磁场中产生弥散非热连续谱；星系团中的弥散射电晕和 relic 则记录了大尺度结构形成中的湍流与激波。

射电星系天鹅座 A 是最经典的例子之一。它展示了核心、喷流、热点和射电瓣这一整套结构，适合用来解释粒子加速、能量输运和谱老化之间的关系。

**图 1.6.3**：天鹅座 A，同步辐射的经典实例。核心、喷流、热点和射电瓣共同显示了活动星系核把能量输运到星系外介质的过程。


在这类系统中，中央活动星系核把能量注入喷流，喷流在远端与环境介质相互作用形成热点，随后带电粒子回流并填充射电瓣。热点附近更接近粒子新近加速的位置，通常预期其谱更平；离加速区更远、经历更长冷却时间的射电瓣，则更容易表现出高频变陡的老化特征。这种“形态 + 光谱”的联合判断，是射电天体物理中非常核心的分析方式。

### 1.6.6 与逆康普顿散射的联系

产生同步辐射的相对论电子也会与周围低能光子场相互作用，把后者散射到更高能量，这就是逆康普顿散射。若被散射的是电子自己产生的同步光子，称为同步自康普顿；若目标光子来自宇宙微波背景、恒星光场或尘埃红外辐射，则会形成不同的多波段辐射分量。

这一联系非常重要。射电数据直接测到的是同步辐射，但同步辐射强度同时依赖电子数密度和磁场强度，二者存在一定退化。若能把射电数据与 X 射线或毫米波中的逆康普顿信号结合起来，就能更严格地约束电子能谱和磁场。对喷流和射电瓣而言，逆康普顿是评估能量损失和磁场强度的重要工具；对星系团而言，热电子对宇宙微波背景造成的康普顿化还会产生 Sunyaev-Zel'dovich 效应，成为射电与毫米波宇宙学观测的重要桥梁。


### 1.6.7 本节小结

同步辐射来自磁场中的相对论电子，是典型的非热连续谱机制。电子能谱若近似幂律，光学薄同步辐射也表现为幂律谱；谱指数、偏振、亮温和低频翻转共同构成识别同步辐射的主要证据。由于高能电子冷却更快，多频成像和谱指数分布还能追踪粒子加速、输运和老化历史。

对干涉测量而言，同步辐射源的复杂性直接转化为观测需求：需要宽频覆盖来测谱，需要偏振校准来研究磁场，需要短基线恢复弥散结构，也需要长基线分辨喷流和热点。下一节将从连续谱转向谱线辐射。谱线观测更直接提供成分、速度和气体物理状态的信息，是射电天文学中另一条同等重要的主线。
