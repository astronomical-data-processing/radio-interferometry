## 7.x 延伸阅读与参考文献

第 7 章的跨度较大：Jones/RIME 属于测量方程的数学语言，接收机和相关器属于信号链工程，主波束、极化、传播效应和 RFI 又同时具有观测、仪器和数据处理含义。因此，本章的延伸阅读更适合按问题组织，而不是按书目顺序堆叠。

比较稳妥的阅读路径是先回到 `7.1` 与 `7.2`，把 Jones 语言、亮度矩阵、相关矩阵和 RIME 的符号体系固定下来；随后阅读 `7.3` 到 `7.5`，把模拟链路、数字相关器和主波束视为同一条系统响应链；最后进入 `7.6` 到 `7.8`，把极化、传播和 RFI 理解成真实观测偏离理想模型的主要来源。这样安排的好处，是每个仪器效应都会落到一个明确的数学位置，而不是变成孤立名词。


### 7.x.1 RIME、Jones 语言与方向依赖效应

本章理论骨架的核心文献是 Smirnov 关于 RIME 的系列论文。第一篇 [*Revisiting the radio interferometer measurement equation. I. A full-sky Jones formalism*](http://adsabs.harvard.edu/abs/2011A%26A...527A.106S) 给出全空域 Jones 形式，是把标量测量方程推广为矩阵测量方程的重要入口。第二篇 [*Calibration and direction-dependent effects*](http://adsabs.harvard.edu/abs/2011A%26A...527A.107S) 把校准和方向相关效应放进同一框架。第三篇 [*Addressing direction-dependent effects in 21 cm WSRT observations of 3C 147*](http://adsabs.harvard.edu/abs/2011A%26A...527A.108S) 展示这一框架怎样进入真实观测。第四篇 [*A generalized tensor formalism*](http://adsabs.harvard.edu/abs/2011A%26A...531A.159S) 则进一步把符号扩展为更一般的张量形式。

这些论文和本章的对应关系很直接。`7.1` 提供 Jones 向量和 Jones 矩阵，`7.2` 建立基本 RIME，`7.5` 到 `7.7` 则给出方向相关项、传播项和极化项的物理例子。阅读时宜始终追问同一个问题：某个物理效应到底应写在 Jones 链的哪一侧、是否依赖方向、是否依赖频率、是否随时间变化、是否会混合极化。


### 7.x.2 偏振、馈源与全相干标定

极化部分若只停留在 Stokes 参数定义，会很容易显得抽象。真正需要建立的是系统认识：馈源基底如何把极化信息编码进相关积，视差角怎样旋转馈源坐标系，D-term 又怎样把 Stokes I 泄漏成假偏振。Hamaker, Bregman & Sault (1996) 的 [*Understanding radio polarimetry. I. Mathematical foundations*](http://adsabs.harvard.edu/abs/1996A%26AS..117..137H) 是这一方向的经典文献，重点在于用线性代数语言描述极化和相干矩阵。Hamaker (2000) 的 [*Understanding radio polarimetry. IV. The full-coherency analogue of scalar self-calibration*](http://adsabs.harvard.edu/abs/2000A%26AS..143..515H) 则把全相干校准与极化保真度联系起来。

这些文献与 `7.6` 的关系尤其紧密。线馈源和圆馈源并不是两个互不相干的约定，而是同一电场在不同基底下的表达；D-Jones 也不是一个经验修正项，而是馈源响应不完美时自然出现的非对角耦合。


### 7.x.3 数字相关器、接收机与工程实现

`7.3` 和 `7.4` 涉及的许多问题属于信号链工程。Richard G. Lyons 的 *Understanding Digital Signal Processing* 适合补足采样、量化、窗函数、FFT、滤波和频谱泄漏等数字信号处理基础。Paul Horowitz 与 Winfield Hill 的 *The Art of Electronics* 虽然不是射电干涉测量教材，但对模拟前端、噪声、增益链、阻抗匹配和实际电路直觉非常有帮助。

这两类材料可以把“公式上的相关器”拉回硬件和软件中的相关器。ADC 位数会影响量化效率，PFB 窗函数会改变通道间泄漏，模拟前端的噪声温度会进入 SEFD，corner turn 和相关矩阵规模会决定数据搬运和算量瓶颈。现代干涉仪中的数字相关器不是抽象黑盒，而是一连串带有代价、近似和资源约束的设计决策。


### 7.x.4 与第 8 章校准的接口

进入第 8 章之前，第 7 章最值得反复回看的内容是若干接口知识。`7.1` 与 `7.2` 提供 Jones 矩阵、亮度矩阵、可见度矩阵和 RIME；`7.5` 说明主波束为什么属于方向相关增益；`7.6` 说明馈源旋转和 D-term 为什么会改变极化测量；`7.7` 说明传播效应为什么可被理解为传播 Jones 项；`7.8` 说明 flag 不是简单后处理，而会直接影响校准和成像权重。

本章最终要形成的判断是：观测系统不会忠实搬运天空信号，而会持续地重写、混合、衰减、旋转、污染和丢弃信息。校准的任务，就是尽可能把这些系统效应显式建模，并从数据中分离出可解释的天空信号。
