## 1.8 天文射电源<a id='science:sec:astronomical_radio_sources'></a>

射电天文学研究的对象极其广泛，从太阳和行星，到银河系内的电离气体、分子云、超新星遗迹和脉冲星，再到河外的恒星形成星系、活动星系核以及宇宙学背景辐射，几乎覆盖了现代天体物理的所有尺度。真正重要的不只是“这个源叫什么”，而是它通过什么物理机制辐射，在观测上呈现什么特征，以及这些特征如何决定望远镜、频段、分辨率和数据处理方法。

与光学图像相比，射电数据往往更直接地反映气体、磁场、相对论粒子和传播效应。很多在可见光下并不显眼甚至完全不可见的结构，在射电波段却十分突出。因此，本节不把射电源当作名录来背诵，而是把它们看成观测需求的来源：不同物理对象会自然要求不同的频率覆盖、角分辨率、时间分辨率、谱线能力、偏振能力和空间尺度响应。


### 1.8.1 射电源的观测诊断框架

射电源通常不能仅凭一张图像分类。连续谱和谱线首先提供两种互补信息：连续谱强调宽带辐射机制、谱指数和亮温，谱线则强调成分、速度和激发状态。对连续谱源，通量密度常写成 $S_\nu\propto\nu^\alpha$，不同谱指数能够区分自由-自由辐射、光学薄同步辐射和同步自吸收等情况。亮温则把面亮度转化为温度单位，若远高于普通热平衡温度，就会指向非热辐射、相干辐射或受激放大。

形态、角尺度、偏振和时间变化同样关键。致密核心、激射斑点和脉冲星定位问题要求长基线；超新星遗迹、星系盘、太阳和 CMB 等延展结构要求短间距与面亮度保真度；同步辐射常带有线偏振，而某些磁层或相干辐射可表现出强圆偏振；喷流演化、太阳暴发、脉冲星和 FRB 又把时间分辨率变成核心要求。所谓“射电源分类”，本质上就是把谱形、亮度、形态、偏振与时间变化这些观测量，同背后的物理环境对应起来。

### 1.8.2 太阳与空间天气

太阳是地球上最亮、同时也是最复杂的射电天体。它的射电辐射既包括较平稳的宁静太阳成分，也包括耀斑、日冕物质抛射、激波和磁重联等活动过程产生的剧烈暴发。对射电干涉观测而言，太阳尤其重要，因为它在同一个目标中叠加了热辐射、非热辐射、偏振、快速时变和传播效应。

宁静太阳在不同频段对应不同高度和光深条件下的色球层、过渡区与日冕等离子体；暴发期的太阳射电辐射则常由高能电子束、等离子体辐射、回旋脉泽辐射等机制主导。观测上，太阳射电爆发的通量密度可远高于典型河外源，频率漂移、偏振和形态演化又携带磁场拓扑与能量释放信息。由于源结构变化快、亮度高、动态范围要求高，太阳成像往往需要高时间分辨率、高频率分辨率以及专门的校准策略；它既服务于基础太阳物理，也与空间天气预报密切相关。


### 1.8.3 行星与磁层辐射

太阳系行星在射电波段同样重要，但辐射机制并不单一。类地行星和巨行星的大气、表面及地下结构可以通过热辐射来探测；具有强磁场的行星，尤其是木星，还会产生显著的非热磁层辐射。行星热辐射帮助约束温度结构、化学组成和表面介电性质，磁层辐射则揭示行星磁场、带电粒子分布以及与太阳风的相互作用。

某些行星低频辐射带有强圆偏振，是典型的磁场诊断量。对系外行星研究而言，寻找类似木星磁层辐射的低频信号，被视为探测行星磁场的一条潜在线路。这类目标说明，射电天文学不仅研究遥远星系，也能直接研究行星环境与空间等离子体物理；不同频段和角分辨率需求，也自然把单碟、低频阵和长基线干涉阵区分开来。


### 1.8.4 电离氢区、分子云与激射源

恒星形成区域是射电天文学最丰富的实验室之一。H II 区中的自由-自由辐射提供年轻大质量恒星周围电离气体的直接证据，而分子云中的转动跃迁谱线则让我们测量冷气体的温度、密度、化学组成和速度场。H II 区常表现为热连续谱源，谱指数和光学厚度随频率变化；分子云则更多依赖谱线诊断，例如 CO 及其同位素、致密气体示踪线以及复杂有机分子的跃迁。

激射源（maser）是这类环境中特别值得强调的目标。受激放大使某些分子跃迁呈现极高亮温、极窄线宽和极小角尺度。水、羟基和甲醇激射常用来研究恒星形成区、晚型恒星包层甚至活动星系核邻近环境。由于激射斑点极其致密，它们往往需要高角分辨率观测，甚至是 VLBI 级别的数据处理；同一批激射斑点的速度和自行还能提供三维运动学信息。


### 1.8.5 超新星遗迹

超新星遗迹是非热射电辐射的经典来源。超新星爆炸产生的激波会压缩并放大磁场，同时把粒子加速到相对论能量，进而产生同步辐射。许多超新星遗迹具有壳层结构、显著偏振和空间变化的谱指数，是理解同步辐射和宇宙线加速的直观例子。

从观测上看，超新星遗迹通常是延展源，可能呈壳层、填充中心或混合形态。射电辐射可以帮助测量激波与星际介质的相互作用、宇宙线加速、磁场结构以及能量反馈。与光学或 X 射线图像联合分析时，还能更完整地理解激波前后的热与非热成分。这类目标也自然引出干涉阵与单碟的互补关系：长基线能分辨精细壳层，但若缺乏短间距，则容易丢失大尺度面亮度结构；角分辨率提高并不自动意味着信息更完整。


### 1.8.6 脉冲星与中子星

脉冲星是射电天文学最具标志性的发现之一。它们本质上是高速自转、强磁场的中子星，射电发射通常具有相干辐射特征，因此亮温可高得惊人，远超一般热源和常见非热连续谱源。脉冲星观测的关键不只是平均通量密度，而是脉冲轮廓、周期、色散量、散射展宽、偏振和单脉冲统计。

这些量既反映脉冲星自身磁层物理，也记录信号穿过电离星际介质时受到的传播效应。色散量给出沿视线自由电子柱密度，散射尾反映小尺度电子密度起伏，偏振角摆动可用于研究磁层几何。从仪器角度看，脉冲星展示了与常规二维成像不同的一类射电实验：时间分辨率、相位稳定性、去色散处理和折叠算法常比图像动态范围更关键。很多时候，原始电压、时间序列或折叠轮廓本身就是科学产品。


### 1.8.7 宇宙微波背景

宇宙微波背景（CMB）是射电和毫米波天文学中最著名的热辐射背景之一。它的平均谱形极接近理想黑体，而真正丰富的宇宙学信息主要体现在微弱的温度起伏、极化结构以及二次各向异性中。与前面多数局域天体源不同，CMB 不是孤立目标，而是覆盖全天的大尺度背景场。

这类科学目标会把观测设计推向另一种方向：波束控制、系统噪声、扫描策略、前景分离、绝对校准和大角尺度保真度常常比极高角分辨率更重要。CMB 也提醒我们，在射电科学里“源”不一定是一个致密斑点；它可能是覆盖极大角尺度、需要统计方法和精密系统控制才能提取的背景场。


### 1.8.8 射电星系、类星体与活动星系核

河外射电源中最具代表性的类别，是由活动星系核（AGN）驱动的射电星系和类星体。它们通常由中心超大质量黑洞吸积供能，并通过喷流把能量输运到宿主星系乃至更大尺度环境中。射电辐射大多由同步辐射主导，因此常带有负谱指数和可观的线偏振。

活动星系核在射电上的形态非常多样。有的核心高度致密，需要 VLBI 才能分辨；有的喷流可以延伸到 kpc 乃至 Mpc 尺度；有的双叶结构极为显著，有的则在多频段联合观测中显示复杂的反馈过程。通过射电图像与谱指数分布，可以研究喷流能量学、电子老化、磁场结构以及 AGN 对宿主星系和星系团环境的反馈。这类目标是干涉测量应用最经典的舞台之一，因为核心、喷流、热点和射电叶之间跨越多个角尺度，需要不同阵列配置、频率和成像策略协同完成。

### 1.8.9 恒星形成星系

与由黑洞活动主导的射电星系不同，恒星形成星系的射电辐射更多反映恒星形成本身及其反馈。其总射电发射通常是两部分混合：一部分来自 H II 区的热自由-自由辐射，另一部分来自超新星遗迹和宇宙线电子产生的非热同步辐射。低频更容易由同步辐射主导，高频厘米波到毫米波则可能逐渐增加自由-自由和尘埃连续谱贡献。

这种混合成分使射电波段成为估计恒星形成率的重要工具，尤其在尘埃严重遮蔽的星系中更具优势。通过多频数据分离热与非热成分，还可以进一步推断大质量恒星形成、宇宙线传播和星系磁场结构。星暴星系则把这些过程推向极端，形成高表面亮度、高超新星率和复杂多尺度结构。这类源通常既有致密亮斑，也有较平滑的盘面结构，因此对角分辨率和面亮度灵敏度都有要求，是理解阵列配置和权重选择影响的典型案例。

### 1.8.10 动态射电天空与快速暴发现象

除了相对稳定的连续谱和谱线源，射电天空还包含大量时变目标。耀斑星、X 射线双星、磁星、脉冲星单脉冲、行星磁层事件以及快速射电暴（FRB），都属于这一类。它们的共同特点是时间结构本身就是核心科学信号，而不是成像后的附属信息。

快速暴发现象通常要求高时间分辨率、高频率分辨率和对传播效应的精细校正。以 FRB 为例，毫秒量级脉冲经过星际和星系际介质传播后，会出现频率相关的色散延迟和散射尾；定位宿主星系又需要干涉阵角分辨率和快速触发能力。动态射电天空因此把传统成像、波束形成、实时搜索、去色散和后续定位结合到同一条工作流中，展示了现代射电天文学越来越强的时间域特征。


### 1.8.11 从射电源到干涉测量需求

回顾这些射电源，可以看到科学对象本身决定了后续仪器问题的形态。致密源如 AGN 核、激射斑点和部分脉冲星定位问题，需要极高角分辨率，因此依赖长基线干涉阵。延展源如超新星遗迹、星系盘、太阳和 CMB 相关结构，则要求良好的短间距覆盖和面亮度保真度。强时变源如太阳暴发、脉冲星和 FRB，需要高时间分辨率与专门的校准、去色散或搜索流程。谱线源又要求高频率分辨率和严格的频谱标定，这与连续谱成像在数据处理上并不完全相同。

因此，理解射电源并不是为了记住几个例子，而是为了回答贯穿全书的问题：为什么需要干涉阵，为什么单一阵列配置不可能服务所有科学目标，为什么同样是“成像”，不同源会要求完全不同的频率设置、基线分布、校准策略和科学产品。后续章节中的可见度、采样、成像和定标，都可以从这些源的需求中找到物理动机。
