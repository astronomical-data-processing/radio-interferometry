## 4.2 基线及其空间表示<a id='visibility:sec:baseline_space'></a>


上一节把条纹和空间频率联系起来；本节转向干涉测量的几何基础。相关器给出的可见度样本由一条基线决定，而基线本身首先是一条真实的三维长度向量：它连接两面天线，方向和长度由阵列布局确定。

但是，干涉仪真正采样的不是这条三维物理向量本身，而是它相对于观测方向的投影。当地球自转时，地面上的物理基线几乎不变；可是相对于固定天源的投影基线会持续改变。这个持续改变的投影，就是 `uv` 轨迹的来源，也是地球自转合成能够增加傅里叶平面采样的原因。

因此，本节的任务是把同一条基线放进几个不同参考系中理解：地面本地坐标给出天线之间的物理分离，赤道参考系把这条分离与地球自转联系起来，而 `u,v,w` 坐标则把它写成相对于相位中心的测量坐标。


坐标变换的路径可以概括如下。首先，在本地笛卡尔坐标中用两面天线的位置差定义物理基线；然后，把坐标轴对齐到地理东、北、天顶方向，即 `ENU` 参考架；再利用台站纬度把 `ENU` 基线写到与地球自转轴相关的 $XYZ$ 参考架中；最后，以相位中心为 $w$ 轴，把基线旋转到 $(u,v,w)$ 参考架。每一步都没有改变同一条物理基线本身，只是在改变描述它的坐标轴。


<a id="vis:sec:421"></a> <!---\label{vis:sec:421}--->
### 4.2.1 物理基线

物理基线是两个天线单元在三维空间中的几何分离矢量。与之对应，投影基线是这条三维基线相对于某个观测方向的有效分量。在国际地球参考架或局部台站参考架中，物理基线可以近似视为常量；但对固定在天球上的源而言，投影基线会随着地球自转不断变化。

这里的“近似常量”并不意味着实际阵列没有细节。高精度处理中还要考虑地壳运动、潮汐、极移和参考架更新；但在本章的几何推导中，天线位置可视为固定，重点放在坐标变换和投影关系上。


#### 4.2.1.1 本地基线向量
<a id="vis:sec:4211"></a> <!---\label{vis:sec:4211}--->

三维空间中的位置可以用向量来描述。定义原点 $O$、一组正交基 $\mathcal{B}$，以及固定在地面的本地笛卡尔参考系 $\mathcal{R}$，其基向量为 $\mathbf{\hat{e}_x}, \mathbf{\hat{e}_y}, \mathbf{\hat{e}_z}$。设两面天线的位置向量分别为

$$
\mathbf{a}_1=\vec{OA}_1=x_1 \mathbf{\hat{e}_x} + y_1 \mathbf{\hat{e}_y} + z_1 \mathbf{\hat{e}_z},
$$

$$
\mathbf{a}_2=\vec{OA}_2=x_2\mathbf{\hat{e}_x} + y_2 \mathbf{\hat{e}_y} + z_2 \mathbf{\hat{e}_z}.
$$

若两天线之间的距离足够短，以至于可以忽略局部地球曲率，那么局部地面可近似看成平面。本地笛卡尔坐标下的物理基线定义为两位置向量之差：

$$
\mathbf{b} = \mathbf{a}_2 - \mathbf{a}_1 = \vec{A_1A_2}.
$$

这个定义说明，物理基线只依赖两面天线的相对位置，而与参考系原点的具体选取无关。交换天线顺序会得到相反基线 $-\mathbf{b}$，这也是可见度函数满足共轭对称关系的几何来源之一。

#### 4.2.1.2 $(E,N,U)$ 参考架：基线向量的全局定义

在前面的本地表达式中，$\mathbf{\hat{e}_x}, \mathbf{\hat{e}_y}, \mathbf{\hat{e}_z}$ 的方向仍然是任意指定的。为了让基线定义与地球表面的绝对方向联系起来，通常采用 `ENU` 参考架：$E$ 指向地理东，$N$ 指向地理北，$U$ 指向当地天顶。原点可以取阵列中心，也可以取某一面参考天线。

这样，基线向量虽然仍是在局部切平面附近描述，但坐标轴已经有了明确的地理含义。对一个紧凑阵列而言，`ENU` 坐标通常已经足以描述天线布局；对跨大陆 VLBI 阵列，则必须先在地心地固坐标中描述台站位置，再转换到每个台站或阵列中心附近的局部坐标。


#### 4.2.1.2 地平坐标系 $(\mathcal{A}, \mathcal{E})$ 中的基线表示


地面观测者通常用方位角 $\mathcal{A}$ 和高度角 $\mathcal{E}$ 描述天空方向。类似地，基线方向也可以用这两个角描述。这里把天线 1 放在地平参考系原点，$E,N,U$ 三个方向按照前面定义。

采用本书的方位角约定：$\mathcal{A}$ 从北向东增加，$\mathcal{E}$ 从地平线向天顶增加。若基线长度为 $|\mathbf{b}|$，则它在 `ENU` 坐标中的分量为

$$
\mathbf{b}_{\rm ENU}
=
|\mathbf{b}|
\begin{bmatrix}
\sin \mathcal{A} \cos \mathcal{E}\\
\cos \mathcal{A} \cos \mathcal{E}\\
\sin \mathcal{E}
\end{bmatrix}.
$$

第一行是东向分量，第二行是北向分量，第三行是天顶方向分量。若两面天线位于同一水平面，则 $U$ 分量为零，对应 $\mathcal{E}=0^\circ$。


<a id="vis:fig:422"></a> <!--\label{vis:fig:422}--></a>
<img src='figures/enu_alt_az.png' width=40%>


**图 4.2.2**：地平坐标系 $(\mathcal{A}, \mathcal{E})$ 与笛卡尔坐标系 $(E,N,U)$ 的关系

<b>约定提醒：</b> 方位角的零点和增大方向在不同软件或工程文档中可能不同。本章采用从北向东增加的天文约定；把公式写入程序时，必须与所用软件的约定保持一致。
</div>


#### 4.2.1.3 赤道坐标架 $(X,Y,Z)$ 中的基线表示

现在已经用 $\mathcal{A}$ 和 $\mathcal{E}$ 描述了 `ENU` 参考系中的基线。下一步是把它接到与地球自转轴相关的 $XYZ$ 参考架。对位于纬度 $L_a$ 的观测者，$Z$ 轴指向北天极，$X$ 轴位于天赤道平面并指向本地子午圈与天赤道的交点，$Y$ 轴补成右手系。这个参考架随地球一起转动，因此适合描述固定在地面上的基线。

几何上，`ENU` 平面和天赤道平面之间的夹角由台站纬度决定。把基线方向延伸到天球上，也可以用赤道坐标 $(H,\delta)$ 描述；这里的 $H$ 与 $\delta$ 描述的是“基线方向”在天球上的等效方向，而不是某个真实天源。


<a id="vis:fig:413"></a> <!--\label{vis:fig:413}--></a>
<img src='figures/xyz_XYZ.png' width=40%>


**图 4.2.3**：$(X,Y,Z)$ 参考架。$T$ 为当地天顶，$(X,Y)$ 为与北天极垂直的天赤道平面，$(N,S,E,W)$ 所在平面为天线局部平面

把 $(\mathcal{A},\mathcal{E},L_a)$ 转换到 $XYZ$ 后，基线可写成

$$
\begin{bmatrix}
X\\
Y\\
Z
\end{bmatrix}
=
|\mathbf{b}|
\begin{bmatrix}
\cos L_a \sin \mathcal{E} - \sin L_a \cos \mathcal{E} \cos \mathcal{A}\\
\cos \mathcal{E} \sin \mathcal{A} \\
\sin L_a \sin \mathcal{E} + \cos L_a \cos \mathcal{E} \cos \mathcal{A}
\end{bmatrix}.
$$

同一个向量也可以写成赤道形式

$$
\begin{bmatrix}
X\\
Y\\
Z
\end{bmatrix}
=
\begin{bmatrix}
|\mathbf{b}| \cos \delta \cos H\\
-|\mathbf{b}| \cos \delta \sin H\\
|\mathbf{b}| \sin \delta
\end{bmatrix}.
$$

这两种写法描述的是同一条基线，只是前者从地平角出发，后者从赤道角出发。它们之间的等价关系来自第 3 章讨论过的天极-天顶-方向球面三角。

<div class=advice>
<b>说明：</b> 这个转换本质上是把局部地平坐标绕东西方向倾斜到天赤道坐标。完整球面三角推导可参考附录 [A.1](../0_Introduction/2_Appendix.ipynb)。
</div>


### 4.2.2 $(u,v,w)$ 空间

现在可以把基线从 $XYZ$ 参考架转换到与相位中心直接相关的 `uvw` 参考架。设相位中心方向为 $\mathbf{s}_0$，其赤道坐标为 $(H_0,\delta_0)$。定义 $w$ 轴指向 $\mathbf{s}_0$，$u$ 轴位于局部切平面内并指向赤经增加方向，$v$ 轴补成右手系并大致指向赤纬增加方向。于是 $(u,v)$ 张成的平面垂直于相位中心方向。

这个定义非常重要：$u$ 和 $v$ 是成像中真正作为傅里叶变量使用的两个坐标，而 $w$ 记录基线沿视线方向的分量。窄视场且 $w(n-1)$ 可忽略时，成像方程近似退化为二维傅里叶变换；宽视场时，$w$ 项必须保留。


从 $XYZ$ 到 $(u,v,w)$ 的变换可以看成两次旋转。先绕 $Z$ 轴旋转，使新坐标系的一个轴对齐到相位中心所在时角；再绕新的 $X'$ 轴旋转，使 $w$ 轴指向相位中心。图 4.2.4 给出了这组旋转的几何关系。


<a id="vis:fig:414"></a> <!--\label{vis:fig:414}--></a>
<img src='figures/coordinates2b.png' width=50%>


**图 4.2.4**：$(X,Y,Z)$ 坐标系与 $(u,v,w)$ 坐标系之间的关系

由 $XYZ$ 到 `uvw` 的变换可写成

$$
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\mathcal{R}_{X'}^{\frac{\pi}{2}-\delta_0}
\mathcal{R}_{Z}^{\frac{\pi}{2}-H_0}
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}.
$$

按照线性代数中的通常约定，右侧旋转先作用。因此先执行绕 $Z$ 轴的旋转，再执行绕 $X'$ 轴的旋转。


第一步旋转矩阵为

$$
\mathcal{R}_{Z}^{\frac{\pi}{2}-H_0}
=
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0 \\
-\cos H_0 & \sin H_0 & 0 \\
0 & 0 & 1
\end{pmatrix}.
$$


第二步旋转矩阵为

$$
\mathcal{R}_{X'}^{\frac{\pi}{2}-\delta_0}
=
\begin{pmatrix}
1& 0& 0\\
0 & \sin \delta_0 & \cos \delta_0  \\
0 &-\cos \delta_0 & \sin \delta_0
\end{pmatrix}.
$$


两次旋转相乘后，得到常用的 $XYZ\rightarrow uvw$ 转换矩阵：

$$
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0
\end{pmatrix}
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}.
$$

矩阵的三行可以分别理解为：基线在相位中心切平面东西方向上的分量、南北方向上的分量，以及沿相位中心视线方向上的分量。


通常 `uvw` 坐标以观测波长为单位，而不是以米为单位。也就是说，先把物理基线除以波长 $\lambda$：

$$
\mathbf{b}_{\lambda}=\frac{\mathbf{b}}{\lambda}.
$$

于是

$$
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\frac{1}{\lambda}
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0
\end{pmatrix}
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}.
$$

这样写以后，$u$、$v$ 和 $w$ 本身是无量纲的“波长数”。不过由于 $l$ 和 $m$ 是角度的小量，成像中也常把 $u$、$v$ 称为以弧度倒数表示的空间频率。这个说法强调的是傅里叶共轭关系，而不是新的物理单位。


### 4.2.3 结论

本节把同一条基线依次放入本地坐标、地平坐标、赤道坐标和相位中心坐标中描述。物理基线由天线相对位置决定；投影基线由物理基线和观测方向共同决定；以波长为单位的 `uvw` 坐标则是干涉仪在可见度空间中的采样坐标。
