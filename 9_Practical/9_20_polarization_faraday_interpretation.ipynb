## 9.20 偏振与 Faraday 诊断：从 Stokes 图到磁化等离子体

第 9.9 节已经引入 Stokes 成像、偏振强度偏差、偏振角和 leakage；第 9.11 节进一步讨论交叉手相位校准和 RM synthesis。本节把这两部分连接起来，重点放在偏振图像如何转化为物理诊断。偏振观测不是在总强度图像旁边多做几张图，而是在复数 $Q+iU$ 的相位、幅度和频率变化中寻找磁场、有热电子的等离子体、几何投影和仪器误差共同留下的痕迹。

这一链条比连续谱总强度更脆弱。总强度 $I$ 的一个小标定误差通常表现为通量尺度偏差；偏振中同样大小的误差可能把未偏振强源泄漏成假 $Q/U$，也可能把偏振角整体旋转，进而改变 RM。Faraday 诊断还依赖 $\lambda^2$ 覆盖和通道化方式。若这些环节没有同时说明，偏振图中的一条 RM 结构既可能是天体物理信号，也可能是 calibration、beam、bandwidth 或分析约定的产物。


### 9.20.1 Stokes 参数和复偏振：$Q+iU$ 的相位才是偏振角

线偏振通常用

$$
P(\lambda^2)=Q(\lambda^2)+iU(\lambda^2)=L(\lambda^2)e^{2i\chi(\lambda^2)}
$$

表示，其中

$$
L=\sqrt{Q^2+U^2},\qquad p=\frac{L}{I},\qquad \chi=\frac{1}{2}\operatorname{atan2}(U,Q).
$$

偏振角前面的 $1/2$ 来自线偏振的自旋二性质：电场方向旋转 $180^\circ$ 后代表同一个线偏振状态，所以 $Q/U$ 平面中的相位是 $2\chi$。这也是偏振角容易出现 $n\pi$ 模糊的根源。若只把 $Q$ 和 $U$ 当作两张普通图像，而没有保持复数 $P=Q+iU$ 的相位关系，就很容易在频率合成、平滑或掩膜时破坏物理信息。

馈源基也会影响校准语言。在线偏振基中，平行手相关主要约束 $I$ 与 $Q$ 的组合，交叉手相关携带 $U$ 与 $V$ 的信息；在圆偏振基中，平行手相关主要约束 $I$ 与 $V$，交叉手相关携带 $Q$ 与 $U$。这并不改变天空本身的 Stokes 参数，却改变误差进入数据的路径。交叉手 delay 和 phase、D-term leakage、parallactic angle 覆盖、偏振角标定源，都是把相关器输出转化为天球 Stokes 图时必须经过的桥。


![polarization observable chain](figures/practical_polarization_observable_chain.png)

图中左侧把 Stokes cube 到偏振物理量的路径写成派生链。中上图显示 $Q/U$ 平面中的复偏振向量，向量相位对应 $2\chi$ 而不是 $\chi$。右上图显示外部 Faraday screen 的理想情形：$\chi$ 随 $\lambda^2$ 近似线性变化。下方面板强调 $I$、线偏振分数、偏振角和圆偏振 $V$ 描述不同物理量；把它们合成为单一“偏振强弱”会丢掉关键信息。


### 9.20.2 偏振强度偏差和角度误差

由于 $L=\sqrt{Q^2+U^2}$ 永远非负，即使真实线偏振为零，带噪声的 $Q$ 和 $U$ 也会给出正的 $L$。这就是 Ricean bias。高信噪比时常用近似去偏差公式

$$
L_{\rm deb}\simeq \sqrt{L^2-\sigma_{QU}^2},
$$

但这个公式只在 $L/\sigma_{QU}$ 足够大且 $Q$、$U$ 噪声近似相同、独立、高斯时才有明确意义。低信噪比偏振测量更适合用似然、上限或后验分布表达，而不是把负根号截断为零后继续当作普通通量使用。

偏振角的不确定度也由偏振信噪比控制。高信噪比近似下

$$
\sigma_\chi \simeq \frac{1}{2}\frac{\sigma_{QU}}{L}\quad {\rm rad}.
$$

这个关系说明，偏振角精度不由总强度 $I$ 的信噪比直接决定，而由 $Q/U$ 中的偏振信号决定。一个明亮但偏振分数很低的源，偏振角可以比一个较暗但强偏振的源更不稳定。RM 拟合进一步把不同频率上的角度误差组合起来，因此 band 内不同通道的 $Q/U$ 噪声、主波束校正和频率相关 leakage 都会进入 Faraday 诊断。


![Ricean bias and angle uncertainty](figures/practical_polarization_ricean_angle_uncertainty.png)

左图显示真实 $L=0$ 时测得的 $L$ 仍呈正分布，真实 $L=4\sigma$ 时测量分布也不以简单高斯方式围绕真值对称。右图给出高信噪比角度误差近似。偏振角进入 RM 拟合时，角度误差的单位通常是弧度；几度的系统角度偏差在长波长处会转化为可观的 RM 偏差。


### 9.20.3 校准误差：旋转、平移和假偏振

偏振校准误差可以粗略分成两类。第一类把 $Q/U$ 平面中的向量旋转，例如交叉手相位零点错误或偏振角标定错误。若这种旋转随频率变化，它会在 $\chi(\lambda^2)$ 中产生斜率，从而模拟或偏移 RM。第二类把总强度的一部分泄漏到偏振中，例如 D-term 未充分求解、off-axis leakage 或主波束偏振模型错误。由于许多射电源的真实偏振分数只有几个百分点，百分之一量级的 leakage 对弱偏振目标已经不是小修正。

交叉手 delay 的物理意义是两个正交偏振通道之间的相对群延迟。未校准的 delay 会使交叉手相位随频率线性变化；交叉手 phase 的常数项则对应 $Q/U$ 平面的整体旋转。D-term 描述馈源不完美正交造成的偏振泄漏，常需要有足够 parallactic angle 覆盖和合适的偏振校准源才能分离。绝对偏振角校准则把仪器 $Q/U$ 坐标固定到天球上的偏振角零点。缺少这一步时，偏振分数可能仍然可用，但 RM 和磁场方向解释会带有整体角度系统误差。


![polarization calibration error chain](figures/practical_polarization_calibration_error_chain.png)

图中把偏振校准误差分成三个效果：馈源基和 parallactic angle 决定天空偏振如何投影到相关器；交叉手 delay/phase 造成 $Q/U$ 旋转或频率相关旋转；D-term 和 off-axis leakage 造成类似加性向量的假偏振。右下图示意残余交叉手相位斜率怎样改变 $\chi(\lambda^2)$ 的斜率。实际分析中，亮源旁边的低偏振结构、主波束边缘的偏振增强、不同频段 RM 不一致，往往需要先回到这些校准环节排查。


### 9.20.4 Faraday depth、RM 与 $\lambda^2$ 规律

Faraday rotation 来自磁化等离子体中左右圆偏振传播相位速度的差异。在常用单位下，沿视线的 Faraday depth 可写为

$$
\phi(s)=0.812\int_s^0 n_e(l)\,B_\parallel(l)\,dl\quad {\rm rad\,m^{-2}},
$$

其中 $n_e$ 的单位为 $\mathrm{cm^{-3}}$，$B_\parallel$ 的单位为 $\mu\mathrm{G}$，路径长度单位为 pc。若偏振源位于一个外部、均匀、Faraday-thin 的旋转屏之后，则

$$
\chi(\lambda^2)=\chi_0+{\rm RM}\,\lambda^2,
$$

且 RM 可以近似等同于该屏的 Faraday depth。这个简单关系是许多 RM 图的基础，但它不是一般情形。若发射区和旋转介质混合，或视线中存在多个 Faraday 分量，$\chi$ 不再严格线性，偏振幅度也会随 $\lambda^2$ 退偏振。

更一般的表达是

$$
P(\lambda^2)=\int_{-\infty}^{+\infty} F(\phi)\,e^{2i\phi\lambda^2}\,d\phi,
$$

其中 $F(\phi)$ 是 Faraday dispersion function。RM synthesis 试图从有限、离散、有噪声的 $P(\lambda^2)$ 采样反演 $F(\phi)$。这与成像中的 Fourier 采样非常相似：有限 $\lambda^2$ 覆盖会产生一个点扩散函数，即 RMSF；观测到的 Faraday 谱是 $F(\phi)$ 与 RMSF 卷积后的 dirty Faraday spectrum。


### 9.20.5 RMSF：Faraday 空间中的点扩散函数

RM synthesis 的响应函数常写为

$$
R(\phi)=\frac{\sum_i w_i\,e^{-2i\phi(\lambda_i^2-\lambda_0^2)}}{\sum_i w_i},
$$

其中 $\lambda_0^2$ 是加权平均的 $\lambda^2$。RMSF 主瓣宽度决定 Faraday depth 分辨率，旁瓣决定不同 Faraday 分量之间的混叠风险。常用近似包括

$$
\delta\phi \simeq \frac{2\sqrt{3}}{\Delta\lambda^2},\qquad
\phi_{\rm max-scale}\simeq \frac{\pi}{\lambda_{\rm min}^2},\qquad
|\phi|_{\rm max}\simeq \frac{\sqrt{3}}{\delta\lambda^2}.
$$

这些公式给出量级判断：$\Delta\lambda^2$ 越大，Faraday 分辨率越高；最短波长限制可探测的 Faraday-thick 尺度；通道宽度限制可避免带宽退偏振的最大 Faraday depth。不同文献或软件对常数因子可能采用略有不同的定义，但物理依赖关系相同。


![Faraday RMSF](figures/practical_faraday_rmsf_lambda2_coverage.png)

图中比较了宽连续、窄带和带缺口的 $\lambda^2$ 采样。宽覆盖带来较窄 RMSF 主瓣；窄覆盖使 Faraday 分辨率变差；带缺口会增强旁瓣，使多个 Faraday 分量更容易互相污染。这个图说明 RM synthesis 不是简单地“把频率维 Fourier 变换一下”，而是带有明确采样函数和响应函数的反演问题。


### 9.20.6 Faraday 复杂性与退偏振

Faraday-thin 分量在 $F(\phi)$ 中近似为窄峰；Faraday-thick 结构在 $\phi$ 空间中具有有限宽度。多个 thin 分量可能在 RMSF 主瓣内混合成一个峰，也可能通过旁瓣制造假峰。Faraday-thick 结构在长波长处容易退偏振，因此即使总偏振强度下降，也不一定意味着磁场变弱；它可能表示视线内或 beam 内存在更强的 Faraday depth 分散。

退偏振可以来自多种机制。Bandwidth depolarization 是一个通道内部偏振角旋转过多，导致通道平均后 $Q/U$ 抵消。Beam depolarization 是一个合成波束内包含不同 RM 或不同本征偏振角的结构，空间平均后偏振减弱。Internal Faraday dispersion 则来自发射区和旋转介质混合，使不同深度的辐射在传播过程中相互抵消。区分这些机制通常需要同时比较偏振分数随频率的变化、RM 图的空间结构、总强度形态、beam 大小和频带设置。


![Faraday complexity](figures/practical_faraday_complexity_interpretation.png)

左图显示多个 Faraday 分量和一个 Faraday-thick 成分经过 RMSF 后会被混合成 dirty Faraday spectrum。右图显示单一 thin screen、多分量结构和 Faraday-thick 介质的 $|P(\lambda^2)|$ 行为差异。RM-CLEAN 可以减弱 RMSF 旁瓣，但不能自动把物理模型唯一化；当 $\lambda^2$ 覆盖有限、信噪比不高或存在 Faraday-thick 结构时，多个模型可能同样符合数据。


### 9.20.7 与真实软件工作流的对应

在 CASA 中，偏振成像通常从正确的 Stokes 输出、交叉手 delay/phase、leakage 和绝对偏振角标定开始。CARTA 适合检查 $I/Q/U/V$ 图、偏振角、偏振分数、频率切片和局部谱线；RM-Tools、pyrmsynth 或类似工具适合进行 RM synthesis、RMSF 检查和 Faraday 谱分析。若偏振产品来自 pipeline，也仍然需要检查校准日志、偏振校准源模型、parallactic angle 覆盖、D-term 解、频带 flag 和主波束偏振模型。

一个偏振结果至少应说明：采用的馈源基和 Stokes 符号约定；偏振校准步骤和校准源；是否完成绝对偏振角校准；$Q/U$ 噪声和偏振强度去偏差方法；偏振掩膜阈值；$\lambda^2$ 覆盖、通道宽度、RMSF 主瓣和旁瓣；RM synthesis 的权重、Faraday depth 网格、RM-CLEAN 参数；是否考虑 bandwidth、beam 和 internal depolarization；偏振角、RM 和偏振分数中的主要系统误差。只有这些信息齐全，偏振图才能从“有颜色的 RM 图”转化为可复现的磁场和等离子体诊断。


### 9.20.8 本节结论

偏振分析的核心对象是复偏振 $P=Q+iU$。偏振强度、偏振角、偏振分数和 RM 都是从这个复数随位置和频率变化中派生出来的量。由于 $L$ 的正定偏差、$\chi$ 的 $n\pi$ 模糊、校准误差对 $Q/U$ 的旋转和泄漏，以及有限 $\lambda^2$ 覆盖造成的 RMSF，偏振图像比总强度图像更依赖完整的误差叙述。

Faraday 诊断的深度在于它连接了磁场方向、热电子密度和几何结构；风险在于同一个 RM 峰可能对应简单外部屏，也可能对应多个 Faraday 分量或 Faraday-thick 介质。教材层面的训练目标不是只会运行 RM synthesis，而是能够把 Stokes 校准、偏振统计、$\lambda^2$ 采样和物理模型放在同一条解释链中。
