## 8.5 校准退化、模型不完备与解可信度

第 8 章前面把校准写成测量方程和最小二乘求解问题，并介绍了 1GC、2GC 和 3GC。还需要补上一层判断：校准解什么时候可信，什么时候只是把模型缺陷、噪声或方向相关误差吸收到增益中。真实处理里，求出一个增益解并不等于校准正确；残差降低也不等于科学量更准确。校准可信度来自多条证据的一致性，包括标尺、参考天线、模型完整性、解间隔、转移稳定性、残差结构和最终科学量。

本节集中讨论三个问题。第一，校准方程存在退化和规范自由度，必须由参考天线、通量标尺或相位约束固定。第二，天空模型不完备会使自校准把真实结构或模型缺陷吸收到增益中。第三，解可信度需要分级表达：可用于候选成像、可用于图像质量改善、可用于科学测量，所需证据并不相同。


![校准规范自由度](figures/calibration_degeneracy_gauge_freedom.png)

图 8.5.1 展示校准退化的基本思想。某些增益和天空模型的共同变换不会改变可见度拟合，必须由参考天线、通量标尺和相位中心固定。退化不是求解器故障，而是测量方程本身允许的自由度。


### 8.5.1 校准退化与规范选择

以方向无关增益为例，观测可见度可写成

$$
V_{pq}^{\rm obs}=g_p g_q^* V_{pq}^{\rm model}+n_{pq}.
$$

若所有天线增益同时乘上一个共同相位因子 $e^{i\phi}$，则 $g_p g_q^*$ 不变，观测方程无法确定绝对相位。因此需要参考天线固定相位规范。类似地，若通量标尺没有外部约束，增益幅度和天空模型通量之间也会存在尺度退化。极化校准中还会出现绝对偏振角、泄漏和交叉手相位的退化；方向相关校准中，天空模型与方向增益之间的自由度更复杂。

规范选择本身不是问题，但必须记录。参考天线变化、通量标尺模型变化、相位中心选择变化，都可能改变校准表的数值表达。若报告只比较增益曲线而不说明规范，很容易把等价变换误判为物理差异。


### 8.5.2 模型不完备与自校准偏差

自校准依赖当前天空模型。若模型缺少扩展结构、弱源、离轴强源或正确谱指数，求解器会试图用天线增益解释这些残差。这样做可能降低图像残差，却同时改变目标源通量、形态或光谱。自校准因此不是无条件的质量提升，而是一个模型和数据之间的闭环。


![模型不完备与校准偏差](figures/calibration_model_incompleteness_bias.png)

图 8.5.2 展示模型不完备如何影响校准。真实天空包含模型缺失结构时，求解增益可能吸收残差，最终导致目标图像的通量或形态偏差。自校准改善残差的同时，也可能改变科学量。


一个实用判断是比较自校准前后的守恒量和稳定量。闭合相位、闭合振幅、强源通量、目标区域积分通量、残差空间结构、解的时间平滑性、不同解间隔下的图像一致性，都可以作为证据。若自校准只在某个很短解间隔下改善图像，且目标通量明显变化，应怀疑过拟合或模型不完备。若残差降低但科学区域的结构随 mask 或模型版本大幅改变，结论应降级。


### 8.5.3 解间隔、自由度与信噪比

解间隔决定校准自由度。过短解间隔可以跟踪快速变化，但每个解的信噪比降低，并增加把噪声吸收到增益中的风险。过长解间隔信噪比较高，却可能无法跟踪真实大气、电离层或系统变化。合适解间隔应与物理变化时间尺度、校准源强度、带宽、目标动态范围和科学目标共同决定。


![解间隔折中](figures/calibration_solution_interval_tradeoff.png)

图 8.5.3 展示解间隔选择的折中。过短、合适和过长分别对应不同的解信噪比、跟踪能力和过拟合风险。解间隔不是经验旋钮，而是决定自由度和物理可解释性的参数。


### 8.5.4 校准解可信度分级

校准解可信度不应只有“通过/失败”两种状态。更合理的表达是分级。关键标尺缺失时，解不可用于科学测量；证据不完整但残差有所改善时，解只能作为候选；残差受控且转移稳定时，解可用于成像；只有当标尺、模型、残差、转移和误差预算共同支持时，解才可进入科学测量报告。


![校准解可信度阶梯](figures/calibration_solution_trust_ladder.png)

图 8.5.4 把校准解可信度分成不可用、候选可用、可用于成像和可用于科学测量。可信度等级来自标尺、模型、残差、转移和科学量的一致性。


### 8.5.5 教学案例：自校准后通量改变

设一幅连续谱图像在 phase-only 自校准后动态范围提升，但目标扩展源积分通量下降。若只看峰值信噪比，自校准似乎成功；若检查模型，发现初始模型主要由致密源组成，扩展结构没有进入 clean mask。此时增益解可能把扩展结构的一部分吸收到校准误差中。合理处理应回到模型、mask 和解间隔，比较自校准前后扩展源通量、残差结构和闭合量，并在报告中说明结论是否受模型不完备限制。

这个案例说明，校准可信度必须回到科学量。对点源动态范围有利的校准，不一定适合扩展通量测量；对成像残差有利的方向相关解，也可能因自由度过多而不适合精密通量或偏振测量。


### 8.5.6 本节结论

校准解是模型、数据、权重、约束和规范选择共同作用的结果。退化需要由参考天线和标尺固定，模型不完备会产生校准偏差，解间隔决定自由度和过拟合风险。判断校准是否可信，不能只看残差是否降低，而应检查标尺、模型、解平滑性、转移稳定性、闭合量、产品差异和最终科学量。
