## 9.x 延伸阅读与后续实践方向

第 9 章把数据检查、基础校准、成像、自校准、图像测量、连续谱源表、成像参数选择、QA 与失败模式识别、科学量测量、谱线物理量与运动学解释、偏振与 Faraday 诊断、短间距与联合成像、宽带宽场算法边界、观测设计与归档数据再分析、软件生态与可复现实践、VLBI 实践入口、低频与高频特殊观测体制、pipeline QA 与再处理决策、最小可复现处理项目、源表与检测产品验证、最小可运行处理工作流、平均展宽、谱线、宽带宽场、偏振、短间距、方向依赖处理和端到端连续谱案例串成一条实践链。进一步学习时，应把软件命令和测量方程对应起来：每一步究竟在改变数据、模型、权重、flag 还是图像解释。

推荐继续阅读 CASA Guides、WSClean 文档、DDFacet/KillMS 相关资料、PyBDSF 连续谱源搜索文档、SoFiA 谱线 source finding 文档、RM-Tools 与 RM synthesis 资料，以及对应望远镜的数据处理 cookbook。阅读这些材料时，重点不是复制命令，而是识别默认假设、适用频段、校准策略和质量诊断方式。


### 后续实践路线

较自然的训练顺序是：先对一个小型连续谱数据集完成 inspection、flag、1GC 和基础成像；随后做一轮保守 phase-only 自校准并评估图像质量；再用 9.14 的端到端案例把流程闭合到科学通量测量，并通过 9.15 进入连续谱源表生成、9.16 进入成像参数选择、9.17 进入 QA 与失败模式识别、9.18 进入图像科学测量，9.19 进入谱线速度约定、moment、柱密度、PV 图和 beam smearing 解释，9.20 进入偏振统计、校准误差、RMSF 和 Faraday 复杂性，9.21 进入最大可恢复尺度、mosaic、feather、joint deconvolution 和通量恢复验证，9.22 进入 MT-MFS、w-term、A/AW-projection 和方向相关校准的算法边界，9.23 进入观测设计与归档数据再分析，9.24 进入软件生态、provenance、参数文件和可复现验证，9.25 进入 VLBI 的延迟模型、fringe fitting、相位参考、SEFD 定标和天体测量，9.26 进入低频/高频观测中的频率标度、特殊校准节奏、QA 与误差预算，9.27 进入 pipeline weblog、自动产品 QA、归档数据再处理分支和 provenance，9.28 进入最小可复现处理项目、配置文件、依赖图、校验和与验证门，9.29 进入 PyBDSF/SoFiA 检测产品、选择函数、完整性、可靠性、CARTA region 和发布样本分支，9.30 进入最小可运行处理工作流、数据清单、环境记录、执行入口、失败重跑和运行报告；之后进入宽带主波束校正、谱线连续谱扣除和 moment 图；最后处理偏振、RM synthesis、短间距、方向依赖宽场成像、VLBI、特殊频段和 pipeline 案例深化。每一轮实践都应保留处理日志、关键诊断图和失败案例，因为真实数据处理能力往往来自对失败模式的识别。
