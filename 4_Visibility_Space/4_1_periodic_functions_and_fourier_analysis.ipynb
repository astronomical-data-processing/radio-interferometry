## 4.1 周期函数与傅里叶分析<a id='visibility:sec:periodic_fourier'></a>


第 2 章已经系统讨论了傅里叶级数和傅里叶变换；本节在进入可见度之前，再从物理图像的角度回看这些概念。干涉仪中的一条基线会在天空上形成一组正弦状条纹。当天空亮度与这组条纹相乘并在视场内积分时，得到的就是该空间频率上的一个复系数。这个复系数就是可见度。

因此，本节关注的不是新的数学定理，而是一个关键翻译：图像平面中的大尺度结构对应低空间频率，小尺度结构对应高空间频率；一条较短基线主要测低空间频率，一条较长基线主要测高空间频率。文献中常见的 Fourier plane、$k$-space、visibility space 和 `uv` plane 在本章语境中都指向同一件事：天空亮度被分解到不同空间频率后的表示。


### 4.1.1 周期函数

周期函数是理解干涉条纹的入口。本节只回顾最必要的性质：振幅、频率、相位以及复指数表示。更系统的数学讨论可参见 [第 2.2 节](../2_Mathematical_Groundwork/2_2_important_functions.ipynb)。


#### 4.1.1.1 周期函数的性质

周期函数是不断重复的函数，其中最简单的例子是正弦函数。对干涉测量而言，正弦函数不仅是数学例子，也可以看成一条基线在天空上形成的理想条纹模式。


```python
x=(np.arange(800)-400)/200.*np.pi
plt.title("Sine function")
plt.plot(x,np.sin(x))
```

**图 4.1.1**：简单的正弦函数

一个周期函数最基本的两个参数是幅值和频率。幅值决定波动强度，频率决定重复尺度；频率越高，结构越细。在图像语境中，低空间频率对应缓慢变化的大尺度亮度分布，高空间频率对应快速变化的细节。

此外，还可以引入幅值偏移和相位偏移。更一般的周期函数可写成

$$
f(x)=A\sin(kx+\phi)+C,
$$

其中 $A$ 是振幅，$k$ 是空间角频率，$\phi$ 是相位，$C$ 是常数偏移。常数偏移对应零空间频率成分；相位则决定这组条纹相对于坐标原点的位置。


```python
amplitude        = 2
frequency        = 1
phase_offset     = np.pi
amplitude_offset = 0
plt.title("General sine function")
plt.plot(x,np.sin(x))
plt.plot(x,amplitude*np.sin(frequency*x+phase_offset)+amplitude_offset)
```

**图 4.1.2**：基本正弦函数及其参数变化后的对比

图 4.1.2 展示了振幅、频率和相位改变后正弦函数如何变化。对后面的可见度讨论来说，最重要的是相位：同样的空间频率，只要相位不同，条纹峰谷在图像平面中的位置就会移动。复可见度的相位正是在记录这种位置信息。

<div class=advice>
<b>提示：</b>把正弦函数的相位偏移设为 $\frac{\pi}{2}$，得到的就是余弦函数。正弦和余弦可以合并成一个复指数，这正是复可见度同时包含实部和虚部的数学原因。
</div>


在很多物理教材中，频率 $f$ 也会写成角频率 $\omega$，两者满足关系

$$
\omega = 2\pi f
$$

#### 4.1.1.2 复杂周期函数

除正弦函数外，还有许多由周期行为构成的函数。对电磁波和干涉条纹而言，复指数形式尤其方便。根据欧拉公式，

$$
e^{ix} = \cos(x) + i \sin(x),
$$

一个复指数同时包含余弦和正弦两种相差 $90^\circ$ 的分量。相关器若能同时测量这两个正交分量，就能构造复可见度；只测其中一个分量，则会丢失一部分相位信息。


### 4.1.2 傅里叶分析：信号重建

下面回顾傅里叶分析中最重要的几个概念，目的在于建立“信号可以由不同尺度的周期成分组合而成”的直观认识。更详细的数学推导可参考 [第 2.3 节](../2_Mathematical_Groundwork/2_3_fourier_series.ipynb)。


#### 4.1.2.1 傅里叶级数

法国数学家约瑟夫·傅里叶的重要发现之一是：在一个有界区间内，任何可积函数，无论形状如何，都可以表示成正弦和余弦函数的线性组合。这里“有界”非常关键，因为我们实际上是在用一组周期函数去逼近原函数，而这个预先设定的区间就对应了这些周期成分的定义范围。

傅里叶级数的更多数学细节已在 [$\S$ 2.3 &#10142;](../2_Mathematical_Groundwork/2_3_fourier_series.ipynb) 中介绍。这里我们更关注它的物理意义。以下面这个任意函数 $f(x) = (x-0.5)^3+3$ 为例：

```python
# define range of x-axis
x=(np.arange(1200)-600.)/200
# calculate y as a function of x
y=(x-0.5)**3+2
# plot
plt.plot(x,y)
plt.title("Arbitrary function")
plt.ylabel("f(x)")
plt.xlabel("x")
```

**图 4.1.3**：示例中的任意函数

现在只考虑区间 $x=[-2,2]$。这个区间长度为 4，也就是希望用周期函数去复制原始函数的周期范围。由于 $f(x)$ 在这个区间上可积，可以用傅里叶级数来逼近它：

$$
f(t) = \frac{1}{2}a_0 + \sum_{n=1}^{\infty}[a_n\cos(w_n t) + b_n \sin(w_n t)],
$$

其中

$$
w_n = \frac{2\pi n}{x_1-x_0},
$$

$$
a_n = \frac{2}{x_1-x_0}\int_{x_0}^{x_1} f(t) \cos(w_n t)\,dt,
$$

$$
b_n = \frac{2}{x_1-x_0}\int_{x_0}^{x_1} f(t) \sin(w_n t)\,dt.
$$

注意 $a_0$ 前面有一个 $1/2$ 系数。实际计算时不可能令 $n\rightarrow\infty$，因此通常只取有限个系数，用不同的 $n_{\max}$ 来观察逼近效果。


```python
nmax=100
#x=(np.arange(1201)-800.)/200
x=(np.arange(1201)/300.-2)
y=(x-0.5)**3+2


def FourierSeriesApprox(xvals,yvals,nmax):
    approx=np.zeros_like(yvals)
    T=(xvals[-1]-xvals[0])
    w=2*np.pi/T
    dt=xvals[1]-xvals[0]
    approx=approx+1/T*(np.sum(yvals)*dt)
    
    an=np.zeros(nmax+1)
    bn=np.zeros(nmax+1)
    for n in (np.arange(nmax)+1):
        an[n]=2/T*np.sum(np.cos(w*n*xvals)*yvals)*dt
        bn[n]=2/T*np.sum(np.sin(w*n*xvals)*yvals)*dt
      
    
    for t in range(len(xvals)):
        for n in (np.arange(nmax)+1):
            #an=2/T*np.sum(np.cos(w*n*xvals)*yvals)*dt
            #bn=2/T*np.sum(np.sin(w*n*xvals)*yvals)*dt
            #approx[t]=approx[t]+an*np.cos(w*n*xvals[t])+bn*np.sin(w*n*xvals[t])
            approx[t]=approx[t]+an[n]*np.cos(w*n*xvals[t])+bn[n]*np.sin(w*n*xvals[t])
    return approx
        
yApprox=FourierSeriesApprox(x,y,nmax)
#plt.plot(x,y)
plt.figure(figsize=(18,12))
plt.plot(x,yApprox,label="Fourier series approx. with nmax=%i"%nmax)
plt.plot(x,y,label="Initial function")
plt.legend(loc="upper left")
```

**图 4.1.4**：任意函数及其傅里叶级数近似

可以看到，拟合在区间边缘通常比中间更差，而且当 $n_{\max}$ 较小时这种边缘偏差会更加明显。这是因为正弦和余弦函数必须周期性重复，而原函数未必与这种重复边界完全匹配。

**注意：** 每个 $n$ 都对应采样区间中的一个空间频率。傅里叶分析最核心的思想，就是找出哪些尺度承载了函数的大部分信息。随着 $n_{\max}$ 增大，我们能恢复越来越细的结构；当它趋向无穷时，对不连续特征的逼近也会更好。

物理学中常把这件事表述为“不同尺度承载不同能量”或“不同尺度承载不同功率”。例如，如果原函数是常数，那么它的全部信息都集中在常数项 $a_0$ 中。

傅里叶级数也可以写成复指数形式：

$$
\begin{aligned}
f(t) &= \sum_{n=-N}^N c_n e^{i w_n t}\\
w_n  &= \frac{2\pi n}{x_1-x_0}\\
c_n  &= \frac{1}{x_1-x_0} \int_{x_0}^{x_1} f(t) \cdot e^{-i w_n x}\,dx
\end{aligned}
$$

##### 4.1.2.2 傅里叶变换

上面的例子是在有限区间上讨论离散空间频率。若采样范围不断扩大、频率间隔不断变细，离散傅里叶级数自然走向连续形式，这就是傅里叶变换。当区间范围扩展到无限，即 $[x_0,x_1]\rightarrow[-\infty,\infty]$，并把离散级数推广为连续频率时，函数 $f$ 的傅里叶变换为

$$
\tilde{f}(s) = \int_{-\infty}^{\infty} f(t) e^{-i2\pi t s} \, dt.
$$

傅里叶变换给出函数在每一个频率上的权重分布。此前对傅里叶级数建立起来的直觉，在这里依然成立，只是从离散尺度推广到了连续尺度。这个思想既适用于时间频率，也适用于空间频率；对射电干涉而言，更关键的是空间频率。天空图像的二维傅里叶变换描述不同角尺度结构如何分布，而干涉仪恰恰直接采样这个傅里叶空间。


##### 4.1.2.2.1 边界问题

傅里叶变换相较于傅里叶级数的一大优势，是它不再依赖人为设定的有限边界。考虑下面的简单正弦波：

$$
f(x) = \sin(4x)
$$

如果我们试图在一个并非其周期整数倍的区间上计算它的傅里叶系数，就会遇到明显的边界问题。下面这个例子很能说明这一点：

```python
def FourierSeriesApprox(xvals,yvals,nmax):
    approx=np.zeros_like(yvals)
    T=(xvals[-1]-xvals[0])
    w=2*np.pi/T
    dt=xvals[1]-xvals[0]
    approx=approx+1/T*(np.sum(yvals)*dt)
    
    an=np.zeros(nmax+1)
    bn=np.zeros(nmax+1)
    for n in (np.arange(nmax)+1):
        an[n]=2/T*np.sum(np.cos(w*n*xvals)*yvals)*dt
        bn[n]=2/T*np.sum(np.sin(w*n*xvals)*yvals)*dt    
    
    for t in range(len(xvals)):
        for n in (np.arange(nmax)+1):
            #an=2/T*np.sum(np.cos(w*n*xvals)*yvals)*dt
            #bn=2/T*np.sum(np.sin(w*n*xvals)*yvals)*dt
            #approx[t]=approx[t]+an*np.cos(w*n*xvals[t])+bn*np.sin(w*n*xvals[t])
            approx[t]=approx[t]+an[n]*np.cos(w*n*xvals[t])+bn[n]*np.sin(w*n*xvals[t])
    return approx

nmax=1
x=(np.arange(1201))/800.
y=np.sin(4*x)
yApprox=FourierSeriesApprox(x,y,nmax)

x1=(np.arange(601-300)*np.pi)/600
y1=np.sin(4*x1)
y1Approx=FourierSeriesApprox(x1,y1,nmax)

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(x,yApprox,label="Fourier series approx. with nmax=%i"%nmax)
plt.plot(x,y,label="Initial function")
plt.legend(loc="upper left")
plt.title("Poorly-set boundary")
#plt.show()
plt.subplot(1,2,2)
plt.plot(x1,y1Approx,label="Fourier series approx. with nmax=%i"%nmax)
plt.plot(x1,y1,label="Initial function")
plt.legend(loc="upper left")
plt.title("Apppropriate boundary")
plt.show()
```

**图 4.1.5**：边界选择合适与不合适时的傅里叶逼近对比

上面两个例子中，函数本身没有变，唯一改变的是所选区间。边界选得合适时，只需要极少数傅里叶系数就能精确恢复原函数；边界选得不合适时，即使增加更多系数，也会出现明显偏差。有限视场、窗函数和采样边界在成像中会产生类似影响，它们最终会表现为旁瓣和泄漏。

傅里叶变换没有同样形式的有限边界问题，因为它是在整个空间上定义的；但真实观测仍然只能在有限时间、有限视场和有限 `uv` 覆盖内采样。因此，数学上的理想傅里叶变换和仪器中的有限采样之间始终存在差距。

##### 4.1.2.2.2 正弦函数的傅里叶变换

在边界合适的情况下，一个正弦函数只需要极少数傅里叶成分就能表示。这提示出一个重要事实：简单周期结构在傅里叶空间中对应非常集中的频率成分。

图像平面中的正弦或余弦函数，在 `uv` 平面中对应一个点。换句话说，正弦或余弦函数的傅里叶变换是位于对应频率位置上的 Dirac delta 函数。射电天文学中常说的“干涉条纹”，本质上就是这种图像平面周期结构与傅里叶平面单点之间的对应关系。

##### 4.1.2.2.3 Dirac delta 函数的傅里叶变换

现在来看另一个极端情形：图像中只有一个点源，其余位置全为零。这种情况通常用 Dirac delta 函数表示（见 [第 2.2 节](../2_Mathematical_Groundwork/2_2_important_functions.ipynb)）。它可以看成高斯函数在 $\sigma \rightarrow 0$、峰值趋于无穷而积分保持常数时的极限。

对于一维中心点源，有

$$
\begin{aligned}
f(t) &= \delta(t),\\
\tilde{f}(s) &= \int_{-\infty}^{\infty} f(t) e^{-i2\pi t s} \,dt\\
             &= \int_{-\infty}^{\infty} \delta(t) e^{-i2\pi t s} \, dt.
\end{aligned}
$$

由 delta 函数的定义可直接得到

$$
\tilde{f}(s) = e^{-i2\pi \cdot 0 \cdot s} = 1.
$$

这说明点源的傅里叶变换在整个傅里叶空间中是常数。所有空间频率都会“看到”这个点源，只是当点源偏离相位中心时，不同空间频率会带有不同相位。若点源位于 $(x_0,y_0)$，图像亮度写成

$$
f(x,y) = \delta(x-x_0)\delta(y-y_0),
$$

其二维傅里叶变换为

$$
\begin{aligned}
\tilde{f}(k_x,k_y) &= \int_{-\infty}^{\infty} f(x,y) e^{-i2\pi (k_x x + k_y y)} \, dx\,dy\\
                    &= e^{-i2\pi (k_x x_0 + k_y y_0)}.
\end{aligned}
$$

因此，中心点源对应常数可见度；偏心点源对应幅度近似恒定但相位随 `uv` 坐标线性变化的可见度。


##### 4.1.2.2.4 高斯函数的傅里叶变换

实际天空中的天体很少是理想点源，更常见的是近似二维高斯形状的源：

$$
f(x,y) = a \exp\left[-\frac{(x-\mu_x)^2}{2\sigma_x^2}\right]
          \exp\left[-\frac{(y-\mu_y)^2}{2\sigma_y^2}\right].
$$

其中 $\mu_x,\mu_y$ 描述源中心位置，$\sigma_x,\sigma_y$ 描述源在两个方向上的宽度。这个高斯源的傅里叶变换仍然是高斯包络乘以相位因子：

$$
\begin{aligned}
\tilde{f}(k_x,k_y)
&= 2\pi a \sigma_x \sigma_y
   \exp\left[-2\pi^2 (k_x^2 \sigma_x^2 + k_y^2 \sigma_y^2)\right]
   \exp\left[-i 2\pi (\mu_x k_x + \mu_y k_y)\right].
\end{aligned}
$$

若高斯源位于相位中心，即 $\mu_x=\mu_y=0$，相位因子退化为 1，只剩下实的高斯包络。这个结果揭示了一个非常重要的尺度互易关系：图像平面中越宽的源，在 `uv` 平面中越窄；图像平面中越紧致的源，在 `uv` 平面中越宽。短基线因此对大尺度结构更敏感，长基线则负责测量紧致结构和细节。


### 4.1.3 图像平面与傅里叶平面的关系

前面的讨论主要集中在单个源上。但真实天空并不只有一个源。那么，多源图像在傅里叶平面中会表现成什么样子呢？

设一幅图像中有 $n$ 个任意形状的源，每个源由函数 $f_n(x,y)$ 描述，则整幅图像可以写为

$$
B(x,y) = \sum_n f_n(x,y)
$$

对应的傅里叶平面则是

$$
\begin{aligned}
\tilde{B}(k_x,k_y) &= \text{FT}\big[B(x,y)\big]\\
                   &= \text{FT}\bigg[\sum_n f_n(x,y)\bigg]\\
                   &= \sum_n \text{FT}\bigg[f_n(x,y)\bigg]
\end{aligned}
$$

也就是说，一幅图像的傅里叶平面等于其中每个源傅里叶变换的线性叠加。这个结果非常重要，因为它告诉我们：傅里叶空间中的结构通常是多种源成分共同叠加的结果，不能简单“直接读图”。

干涉仪并不直接采样图像平面，而是直接采样傅里叶平面。本章后面的内容将讨论这种采样如何由基线、相关器和地球自转共同实现，以及不完整采样为何会在成像中留下脏波束。
