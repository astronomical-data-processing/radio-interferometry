### 术语表与符号表（GM）

检查首次出现的新定义是否使用了 *斜体*，并确认这些定义是否已经加入术语表；同时检查数学符号是否已经纳入符号表，并遵循基础数学排版规范：

1. 向量、标量和矩阵：
 * $a, A$ - 表示标量
 * $\mathbf{A}$, $\boldsymbol{\mathcal{A}}$ - 表示矩阵
 * $\mathbf{a}$ - 表示向量

2. $2\times2$ 极化矩阵与 $N\times N$ 非极化矩阵：
 * $\mathbf{A}$ - 表示 $2\times2$ 极化 Jones 矩阵
 * $\boldsymbol{\mathcal{A}}$ - 表示 $N\times N$ 非极化矩阵

3. Jones 与 Jacobian：
 * 使用 $\mathbf{J}$ 表示 Jones 矩阵，使用 $\mathbb{J}$ 表示 Jacobian 矩阵

4. 傅里叶变换：
 * 使用 $\mathscr{F}\{\cdot\}$ 表示傅里叶变换

5. 使用下标避免歧义：
 * 如果一个符号可表示多个不同物理量，应使用下标区分。例如 $\lambda$ 既可表示波长，也可表示 LM 阻尼因子，此时应写成 $\lambda_{\textrm{LM}}$ 来表示后者。新增此类符号时，也应一并加入术语表。

### 拼写风格

由于参与者的语言背景不同，文本拼写风格需要统一。本项目建议采用 [MNRAS 风格](http://www.oxfordjournals.org/our_journals/mnras/for_authors/#6.2 Spelling, grammar, punctuation and mathematics)，即总体遵循英式拼写，例如使用 `colour` 而不是 `color`，但多数 `-ize` 结尾仍保留 `z`。

可遵循的核心原则包括：

> ##### 标点
> 连字符（LaTeX 中一个短横）用于复合形容词，例如 `low-density gas`、`least-squares fit`、`two-component model`。这也适用于简单的单位修饰结构，例如 `1.5-m telescope`、`284.5-nm line`，但不适用于过于复杂的单位组合或范围表达。
>
> N-dash（LaTeX 中两个短横）比连字符更长，可用于：1）连接关键词；2）作为插入语的分隔；3）表示数值范围；4）连接专有名词组合，如 `Kolmogorov–Smirnov test`。
>
> M-dash（LaTeX 中三个短横）在 MNRAS 风格中通常不使用。
>
> ##### 拼写与语法
> 推荐使用英式英语，例如 `centre` 而不是 `center`，`labelled` 而不是 `labeled`。在 `-ise/-ize` 的选择上，可遵循期刊给出的具体规则。

如果全书后续需要统一为完全中文风格，可把这些规则视为历史编辑规范说明。
