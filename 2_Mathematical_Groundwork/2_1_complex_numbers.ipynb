## 2.1 复数、相位与相量表示<a id='math:sec:complex_numbers'></a><!--\label{math:sec:complex_numbers}-->

射电干涉测量面对的原始物理量是随时间快速振荡的电场。单个天线记录到的电压波形本身并不直接给出天空图像；真正有用的是不同天线接收信号之间的相位差、相关强度以及这些量随基线、频率和时间的变化。复数之所以成为干涉测量的基本语言，并不是因为它带来了形式上的简洁，而是因为一个复数正好可以同时记录两件事：振幅有多大，相位位于振荡周期的哪个位置。

这一节把复数看成相位的记账工具。代数规则仍然会给出，但重点放在它们为何适合描述波：复平面上的旋转对应相位推进，复共轭对应相位反向，两个相量相乘对应相位相加，而一个相量乘以另一个相量的共轭则直接给出相对相位。这最后一点正是复可见度的核心直觉。


### 2.1.1 复平面、模与共轭<a id='math:sec:the_field_of_complex_numbers'></a><!--\label{math:sec:the_field_of_complex_numbers}-->

复数通常写作

<a id='math:eq:complex_cartesian'></a><!--\label{math:eq:complex_cartesian}-->
$$
z=x+i y,
$$

其中 $x,y\in\mathbb{R}$，$i^2=-1$。从代数上看，$x$ 是实部，$y$ 是虚部；从几何上看，$z$ 是复平面上的一个点，也可以看成从原点指向该点的有向量。复数加法就是平面向量加法：

$$
(x_1+i y_1)+(x_2+i y_2)=(x_1+x_2)+i(y_1+y_2).
$$

乘法的定义稍微不那么直观，但它是复数真正有用的地方。利用 $i^2=-1$ 展开可得

$$
(x_1+i y_1)(x_2+i y_2)=(x_1x_2-y_1y_2)+i(x_1y_2+x_2y_1).
$$

这个规则的深层含义不是“多项式展开”，而是“伸缩加旋转”。为了看清这一点，需要把复数改写成极坐标形式。令

$$
|z|=\sqrt{x^2+y^2},\qquad \phi=\operatorname{atan2}(y,x),
$$

则

$$
z=|z|(\cos\phi+i\sin\phi).
$$

这里 $|z|$ 是复数的模，表示复平面中离原点的距离；$\phi$ 是幅角，表示相量相对实轴的转角。在波动问题中，这两个量分别对应振幅与相位。


复共轭定义为

$$
z^*=x-i y.
$$

几何上，复共轭把点 $(x,y)$ 反射到 $(x,-y)$，也就是把相位取反。模可以用共轭简洁地写成

$$
zz^*=(x+i y)(x-i y)=x^2+y^2=|z|^2.
$$

若 $z=|z|e^{i\phi}$，则共轭为

$$
z^*=|z|e^{-i\phi}.
$$

因此，对于两个相量

$$
z_1=A_1e^{i\phi_1},\qquad z_2=A_2e^{i\phi_2},
$$

有

$$
z_1z_2=A_1A_2e^{i(\phi_1+\phi_2)},\qquad
z_1z_2^*=A_1A_2e^{i(\phi_1-\phi_2)}.
$$

第一式说明复数乘法会把相位相加；第二式说明共轭乘积会把共同的相位参考消去，只留下相对相位。射电干涉测量中的相关器正是利用这一性质，把快速振荡的电场转换为随时间缓慢变化、可积累平均的复可见度。


### 2.1.2 欧拉公式：旋转的指数写法<a id='math:sec:eulers_formula'></a><!--\label{math:sec:eulers_formula}-->

欧拉公式指出，对于任意实数 $\phi$，

<a id='math:eq:euler_formula'></a><!--\label{math:eq:euler_formula}-->
$$
e^{i\phi}=\cos\phi+i\sin\phi.
$$

它并不是一个神秘恒等式，而是指数函数、正弦函数和余弦函数在泰勒展开中的自然汇合。把指数函数展开为幂级数，得到

$$
e^{i\phi}=1+i\phi+\frac{(i\phi)^2}{2!}+\frac{(i\phi)^3}{3!}+\frac{(i\phi)^4}{4!}+\cdots.
$$

利用 $i^2=-1$、$i^3=-i$、$i^4=1$，将实部和虚部分开：

$$
e^{i\phi}=\left(1-\frac{\phi^2}{2!}+\frac{\phi^4}{4!}-\cdots\right)
+i\left(\phi-\frac{\phi^3}{3!}+\frac{\phi^5}{5!}-\cdots\right).
$$

括号中的两串级数正是 $\cos\phi$ 与 $\sin\phi$ 的泰勒级数，因此得到欧拉公式。这个推导值得保留，因为它说明复指数不是人为记号，而是把圆周运动写成指数函数后的必然结果。

![欧拉公式的几何含义](figures/complex_plane_euler.png)

**图 2.1.1** 欧拉公式的几何含义。单位复数 $e^{i\phi}$ 位于单位圆上，其实部和虚部分别是 $\cos\phi$ 与 $\sin\phi$。


欧拉公式带来的好处，不只是把三角函数写短。复指数函数具有一个极其重要的性质：对它求导只会产生一个常数因子，

$$
\frac{d}{dt}e^{i\omega t}=i\omega e^{i\omega t}.
$$

也就是说，复指数是微分运算的本征函数。许多线性系统对复指数输入的响应只是改变其振幅和相位，而不会改变其频率形式。傅里叶分析正是建立在这个事实之上：复杂信号可以分解成许多复指数分量，线性系统对每个分量分别作用，最后再叠加回来。后面的可见度方程、卷积定理和 DFT 都会不断利用这一点。


### 2.1.3 实振荡、相量与双边频率表示<a id='math:sec:periodic_functions_and_complex_numbers'></a><!--\label{math:sec:periodic_functions_and_complex_numbers}-->

考虑一个实值余弦振荡

<a id='math:eq:real_cosine'></a><!--\label{math:eq:real_cosine}-->
$$
f(t)=a\cos(\omega t+\phi),\qquad a\ge 0.
$$

借助欧拉公式，可以把它写成复指数的实部：

<a id='math:eq:phasor_real_part'></a><!--\label{math:eq:phasor_real_part}-->
$$
f(t)=\operatorname{Re}\left\{a e^{i(\omega t+\phi)}\right\}.
$$

这个式子中，$a e^{i\phi}$ 是一个不随时间变化的复振幅，$e^{i\omega t}$ 则负责相量在复平面中的匀速转动。因此同一个实振荡可以理解为“一个旋转相量在实轴上的投影”。相位 $\phi$ 改变的不是振荡频率，而是相量起转时的方向；振幅 $a$ 改变的是圆周半径。

若需要把实信号完全写成复指数之和，则使用

$$
\cos\theta=\frac{e^{i\theta}+e^{-i\theta}}{2},
$$

得到

<a id='math:eq:two_sided_cosine'></a><!--\label{math:eq:two_sided_cosine}-->
$$
a\cos(\omega t+\phi)
=\frac{a}{2}e^{i\phi}e^{i\omega t}
+\frac{a}{2}e^{-i\phi}e^{-i\omega t}.
$$

这就是实信号的双边频率表示：正频率 $+\omega$ 的系数为 $(a/2)e^{i\phi}$，负频率 $-\omega$ 的系数是它的复共轭。后一条性质称为共轭对称性，它保证两项相加后虚部相互抵消，最终留下一个实函数。后面讨论傅里叶级数和连续傅里叶变换时，实信号频谱的共轭对称性会反复出现。

正弦函数不需要单独建立一套表示，因为

$$
\sin(\omega t+\phi)=\cos\left(\omega t+\phi-\frac{\pi}{2}\right).
$$

也就是说，正弦和余弦只是相位参考相差 $\pi/2$ 的同一种振荡。


### 2.1.4 相关测量中的相对相位<a id='math:sec:relative_phase_correlation'></a><!--\label{math:sec:relative_phase_correlation}-->

复数语言在干涉测量中最直接的用途，是描述两个天线接收信号的相对相位。设两个天线接收到同一窄带波的解析表示分别为

$$
E_p(t)=A_p e^{i(\omega t+\phi_p)},\qquad
E_q(t)=A_q e^{i(\omega t+\phi_q)}.
$$

相关器形成的基本量可以写成一个时间平均的共轭乘积：

$$
\langle E_p(t)E_q^*(t)\rangle
=\left\langle A_pA_q e^{i(\omega t+\phi_p)}e^{-i(\omega t+\phi_q)}\right\rangle
=A_pA_q e^{i(\phi_p-\phi_q)}.
$$

共同的快速因子 $e^{i\omega t}$ 与 $e^{-i\omega t}$ 相互抵消，留下的是两个接收信号之间的相位差。实际系统会有有限带宽、噪声、增益、延迟补偿和积分时间，表达式会更复杂；但“共轭乘积保留相对相位”这一点并不会改变。若几何延迟为 $\tau_g$，在常见相位约定下，相位差与频率近似满足

$$
\Delta\phi \simeq 2\pi\nu\tau_g,
$$

符号则取决于电场相位、傅里叶核和基线方向的约定。后续章节在写出完整可见度方程时，会明确采用的符号约定。

![相量共轭乘积与相对相位](figures/phasor_correlation_geometry.png)

**图 2.1.2** 两个相量的共轭乘积。左图显示两个同频相量的相位差；右图显示 $z_1z_2^*$ 的相位等于 $\phi_1-\phi_2$，共同的时间转动已经被消去。


下面的数值实验把三个同频率、不同振幅和不同初相位的振荡同时画在复平面和时间域中。第三幅图计算参考相量与另外两个相量的共轭乘积，可以看到输出相位不再跟随 $\omega t$ 快速旋转，而只由相对相位决定。


```python
frequency_hz = 2.0
omega = 2 * np.pi * frequency_hz
t = np.linspace(0, 1.0, 800, endpoint=False)
phases = np.deg2rad([0.0, 45.0, 120.0])
amplitudes = np.array([1.0, 1.0, 0.7])
labels = ['reference', 'phase lead 45 deg', 'phase lead 120 deg']
colors = ['tab:blue', 'tab:orange', 'tab:green']

phasors = [amp * np.exp(1j * (omega * t + phi)) for amp, phi in zip(amplitudes, phases)]
t0 = 0.18
idx0 = np.argmin(np.abs(t - t0))

fig, axes = plt.subplots(1, 3, figsize=(15, 4.6))

ax = axes[0]
for amp, phi, label, color, z in zip(amplitudes, phases, labels, colors, phasors):
    circle = amp * np.exp(1j * np.linspace(0, 2 * np.pi, 400))
    ax.plot(circle.real, circle.imag, color=color, alpha=0.25)
    ax.arrow(0, 0, z[idx0].real, z[idx0].imag,
             color=color, width=0.01, length_includes_head=True)
    ax.plot(z[idx0].real, z[idx0].imag, 'o', color=color, label=label)
ax.axhline(0, color='0.7', lw=1)
ax.axvline(0, color='0.7', lw=1)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('Real part')
ax.set_ylabel('Imag part')
ax.set_title(f'Phasors in the complex plane (t = {t[idx0]:.2f} s)')
ax.legend(loc='upper left', fontsize=9)

ax = axes[1]
for z, label, color in zip(phasors, labels, colors):
    ax.plot(t, z.real, color=color, lw=2, label=label)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Real part of the signal')
ax.set_title('Time-domain projection of the phasors')
ax.grid(alpha=0.3)
ax.legend(fontsize=9)

ax = axes[2]
reference = phasors[0]
for z, label, color in zip(phasors[1:], labels[1:], colors[1:]):
    visibility = reference * np.conj(z)
    ax.plot(visibility.real[0], visibility.imag[0], 'o', ms=9, color=color,
            label=f'{label} relative to reference')
    ax.arrow(0, 0, visibility.real[0], visibility.imag[0],
             color=color, width=0.01, length_includes_head=True)
    print(f'{label:>18s}: |z_1 z_2^*| = {np.abs(visibility[0]):.3f}, phase = {np.rad2deg(np.angle(visibility[0])):6.2f} deg')
ax.axhline(0, color='0.7', lw=1)
ax.axvline(0, color='0.7', lw=1)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('Real part')
ax.set_ylabel('Imag part')
ax.set_title(r'Relative phase: $z_1 z_2^*$')
ax.legend(fontsize=9, loc='upper left')

plt.tight_layout()
plt.show()
```

代码中的 `reference * np.conj(z)` 已经非常接近复可见度的代数骨架。真实相关器还要处理有限带宽、采样、量化、噪声和校准项，但基本结构仍是“一个信号乘以另一个信号的复共轭，再对时间或频率做平均”。这也是复共轭在干涉测量公式中出现得如此频繁的原因。


### 2.1.5 本节要点

复数可以看作平面上的点，也可以看作带有振幅和相位的相量。欧拉公式把圆周运动写成指数形式，使得振荡、相位平移和傅里叶分解都能用同一种语言表达。对射电干涉测量而言，最重要的结论是：$z_1z_2^*$ 的相位给出两个相量的相对相位，而这正是相关器从电场中提取复可见度的数学基础。
