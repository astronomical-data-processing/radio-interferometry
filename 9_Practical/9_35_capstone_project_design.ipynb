## 9.35 综合课程项目设计：从单次练习到研究训练

第 9 章前面的实践内容已经覆盖了从数据检查、校准、成像、自校准、源表、谱线、偏振、短间距、宽场宽带，到归档复盘、可复现工作流、样本包和处理报告的多个环节。若这些环节只以单节练习出现，训练会变成很多分散技能。综合课程项目的作用，是把技能重新组织成一条研究式路径：从科学问题出发，选择适合的数据，建立最小工作流，生成和验证产品，写出可复查报告，并形成带限制声明的发布包。

综合项目不等于把所有软件命令完整跑一遍。真正的目标是训练判断：科学目标能否被当前数据支持，处理参数为什么这样选，质量证据怎样影响结论，误差预算是否完整，结论边界是否清楚。一个项目可以很小，但必须让每个结论都能回到数据、参数和证据。这样实践训练才不只是工具操作，而是接近真实研究的推理过程。

本节给出一套项目设计框架。它适合课程、暑期学校、研究组训练或自学项目使用，也可以与 9.33 的轻量样本包和 9.34 的复查量规直接配套。重点不是规定唯一作业形式，而是说明一个项目怎样分阶段推进、怎样评价能力、怎样按不同深度设计任务，以及怎样逐步建立覆盖连续谱、谱线、偏振、短间距、宽场宽带和归档复盘的项目库。

### 9.35.1 综合项目应从科学问题开始

一个常见误区，是把综合项目设计成“完成某软件教程”。这样的项目能训练命令熟悉度，却很难训练科学判断。更好的起点是一个可以被数据回答或限制的科学问题。例如：一个连续谱场中某个扩展源的积分通量是否稳定；一个归档数据能否给出弥散辐射上限；一个谱线 cube 的 moment 图是否受 beam smearing 影响；一个偏振源的 Faraday 结构是否能由当前 $\lambda^2$ 覆盖分辨；一个短间距缺失的数据是否能支持总通量结论。

科学问题一旦明确，后续任务会自然展开。目标量决定频率、角尺度、速度分辨率、偏振覆盖或主波束区域；数据适配决定是否进入处理；最小工作流决定输入、环境和参数身份；产品验证决定图像、源表或 moment 图是否可信；误差预算决定结论置信度；限制声明决定哪些解释被支持，哪些仍然不能排除。

项目说明可以采用这样的句式：本项目关注某一目标量；给定数据或样本包保留了哪些证据；需要生成哪些产品；最终报告必须说明哪些限制。这样的设计让项目从一开始就带着科学方向，而不是在处理完成后才寻找解释。


### 9.35.2 里程碑：允许停止、回退和降级

阶段项目应按里程碑推进，而不是最后一次性提交。第一个里程碑检查科学问题与数据适配：目标量是否可测，频率、尺度、灵敏度和质量证据是否足够。第二个里程碑建立最小运行与初始产品：输入、环境、参数和运行编号是否清楚，初始图像或源表是否生成。第三个里程碑进入产品验证与误差预算：残差、局部噪声、通量标尺、主波束和数据限制是否被纳入。第四个里程碑完成报告复查与发布包：结论、限制声明、清单和产品身份是否一致。

里程碑的价值在于它允许项目在合理位置停止或回退。若第一个里程碑发现目标不可测，最好的结果可能是改写问题，而不是硬做处理。若第二个里程碑发现输入或参数错误，项目应回退到清单和配置。若第三个里程碑发现证据不足，结论应降级为候选、上限或限制声明。真实研究中，停止、回退和降级并非失败，而是保持结论可信的正常机制。


![综合项目里程碑](figures/practical_capstone_milestone_ladder.png)

图 9.35.2 展示四个里程碑：科学问题与数据适配，最小运行与初始产品，产品验证与误差预算，报告复查与发布包。每个里程碑都应能形成可复查产物，并允许停止、回退或降低结论等级。


每个里程碑都应有明确产物。数据适配阶段可以提交目标量说明、数据清单和适配矩阵；最小运行阶段可以提交运行报告、参数文件和初始产品；产品验证阶段可以提交差异图、局部噪声、源表检查或上限计算；最终阶段可以提交处理报告、限制声明和发布包 manifest。这样复查不会只发生在最后，而是在项目进行中持续校正方向。


### 9.35.3 能力评价：不仅看最终图像

综合项目的评价应覆盖概念、流程、诊断、测量和表达五个维度。概念理解关注能否把科学问题转化为可观测量和限制边界；流程组织关注能否建立可复查的运行和依赖关系；质量诊断关注能否识别异常、定位误差来源和判断模型不完备；科学测量关注能否给出通量、上限、源表、moment 或偏振量，并写出误差预算；证据表达关注能否提交可审查报告并根据复查意见修订。

评价深度可以分为基础达成、进阶达成和研究训练。基础达成强调看懂数据和产品；进阶达成强调能重跑、比较和解释取舍；研究训练强调能管理依赖、判断模型不完备、比较多种假设，并形成可审查结论。


![综合项目能力评价矩阵](figures/practical_capstone_assessment_matrix.png)

图 9.35.3 给出能力评价矩阵。评价不应只看最终图像，还应看能否解释目标量、组织流程、诊断误差、完成测量和表达证据。不同深度对应不同任务边界和证据要求。


这种评价方式能避免两个极端。一个极端是只看最终图像是否好看，忽视数据身份、误差和限制。另一个极端是只检查命令是否完整，忽视科学结论是否成立。综合项目应把二者连接起来：处理过程要可追溯，最终结论要可审查，图像和源表只是证据链中的一部分。


### 9.35.4 同一项目的三个深度

同一项目可以设计成三个深度，而不必为不同训练层次更换完全不同的案例。基础项目可以要求看懂数据与产品：检查元数据、质量报告、图像、源表或残差，并写出初步判断。进阶项目可以要求重跑与误差预算：改变成像权重、阈值、mask 或测量区域，比较产品差异并给出误差预算。研究项目可以要求完整复盘与发布包：说明数据适配、建立最小工作流、管理清单和依赖、形成限制声明，并整理可复查发布包。

三个深度可以共用同一输入，例如样本包、数据清单和图像产品；差异在于产物和证据要求。基础项目产物可以是质量判读和简短报告；进阶项目产物可以是重跑产品、差异审计和误差预算；研究项目产物可以是复盘报告、发布包和复查记录。这样的设计让课程材料具有连续性，也使不同阶段的训练可以相互衔接。


![同一项目的三个深度](figures/practical_capstone_depth_design.png)

图 9.35.4 展示同一项目如何设计为基础项目、进阶项目和研究项目。深度差异来自任务边界和证据要求，而不是换一套互不相干的案例。


例如，一个连续谱归档复盘项目在基础深度中可以只要求判断原图是否适合低面亮度测量；在进阶深度中要求用 taper 重跑并比较共同 beam 下的上限；在研究深度中要求建立数据清单、质量证据板、产品 manifest、处理报告和限制声明。三者共享同一科学目标，却训练不同层次的能力。


### 9.35.5 项目库：覆盖关键科学量和失败模式

一个长期可用的项目库不需要每个案例都很大，但应覆盖关键科学量和常见失败模式。连续谱项目可以训练成像、源表、通量和动态范围；谱线项目可以训练速度约定、moment 图、mask、柱密度和上限；偏振项目可以训练泄漏、旋转量、退偏振和 Faraday 分辨率；短间距项目可以训练缺通量、mosaic、feather 和联合成像；宽场宽带项目可以训练主波束、方向依赖和频谱指数偏差；归档复盘项目可以训练证据链、限制声明和发布包。

项目库的组织方式应服务能力覆盖，而不是只按软件分类。同一个软件可以出现在多个项目中，同一个科学问题也可以用不同工具实现。关键是每个项目都说明自己训练哪些科学量、哪些处理判断、哪些误差来源和哪些报告能力。


![项目库能力覆盖图](figures/practical_capstone_portfolio_map.png)

图 9.35.5 展示项目库的能力覆盖。连续谱、谱线、偏振、短间距、宽场宽带和归档复盘分别覆盖不同科学量和失败模式。项目库不必每个案例都很大，但应避免所有案例都训练同一种能力。


一个实用项目库可以从三类案例开始。第一类是标准路径案例，用于建立基本流程和期望结果。第二类是失败模式案例，用于训练诊断，例如校准残差、主波束边缘伪检、短间距缺失、mask 错误或源表选择函数变化。第三类是开放复盘案例，用于训练数据适配、限制声明和报告修订。三类案例相互配合，才能把实践训练从“照着做”推进到“知道为什么这样做”。


### 9.35.6 教学案例：低面亮度连续谱综合项目

以低面亮度连续谱归档复盘为例，综合项目可以这样展开。科学问题是：某个目标区域内是否存在可由当前数据约束的弥散连续谱结构。第一个里程碑要求判断归档数据是否适合该问题，重点检查频率、最大可恢复尺度、主波束范围、pipeline 质量报告和强源残差。若最大可恢复尺度明显小于目标尺度，项目应在此处改写问题，只讨论较小尺度结构或给出新观测需求。

第二个里程碑建立最小运行。保留归档校准，使用适合低面亮度目标的权重和 taper 重新成像，记录参数、运行编号和产品身份。第三个里程碑比较原图和重跑图，卷积到共同 beam，检查局部噪声、残差、主波束裁剪和强源排除区域，并计算通量或上限。第四个里程碑形成报告和发布包，写明上限适用的角尺度、频率、主波束范围和数据限制。

基础深度可以只完成第一个里程碑和简短判读；进阶深度完成重跑和误差预算；研究深度完成发布包与复查记录。这样同一项目既能训练基本概念，也能训练研究级证据链。


### 9.35.7 本节结论

综合课程项目把第 9 章的分散实践技能组织成研究式训练。项目应从科学问题开始，以里程碑推进，以阶段产物接受复查，以能力矩阵评价概念、流程、诊断、测量和表达。不同深度的项目可以共享同一输入，通过不同任务边界和证据要求形成基础、进阶和研究训练。

第 9.35 节的核心思想是：实践训练的最终目标不是完成软件步骤，而是形成可审查科学结论。数据选择、最小工作流、产品验证、误差预算、处理报告和发布包共同构成完整能力链。这样的项目设计能把教材内容落到真实训练场景，也能为后续建立项目库和样本包提供统一框架。
