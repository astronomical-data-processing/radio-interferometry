## 6.x 延伸阅读与参考文献 <a id='deconv:sec:refs'></a>


本章不宜只记成“会跑一次 CLEAN 的命令”。真正重要的是把去卷积理解为一个有先验、有停止准则、有质量评估、并最终通向源目录的反问题流程。因此，本节不再只是罗列书目，而是按“概念、算法、软件、工作流”四个层次组织延伸阅读，帮助你把第 6 章接到更真实的数据处理语境中。

### 6.x.1 本章主线回顾

* 如果只保留一个工作框架，可以把第 6 章压缩为：`dirty image / PSF -> 选择天空先验 -> minor / major cycle 去卷积 -> restored image 与 residual 检查 -> 源搜索与目录构建`。
* `6.1` 和 `6.2` 的核心，不是记住 `CLEAN` 的历史名称，而是理解为什么点源先验能够把一个病态反问题变成可操作的迭代算法。
* `6.3` 的重点，是区分“算法思想”和“工程实现”：Högbom、Clark、Cotton-Schwab、多尺度、多频率方法并不是互斥菜单，而是在不同数据规模、视场和谱结构下的不同折中。
* `6.4` 和 `6.5` 决定你是否真的能把图像用于科学分析。对残图、局部噪声、clean bias、伪影和源拟合误差缺少判断，往往比“不会跑命令”更危险。

### 6.x.2 建议对照的教材与讲义

* [<cite data-cite='taylor1999synthesis'>Synthesis Imaging in Radio Astronomy II</cite> &#10548;](http://adsabs.harvard.edu/abs/1999sira.conf.....T)：讲义式体例非常适合与本章并读，用来把课堂示例接回真实阵列数据处理。
* *Interferometry and Synthesis in Radio Astronomy*（Thompson, Moran, Swenson）：如果你希望把本章提升到可独立查阅公式、理解近似条件与工程约束的层次，这是最重要的系统参考书。
* [Essential Radio Astronomy &#10548;](https://science.nrao.edu/opportunities/courses/era)：NRAO 的开放教材。它对去卷积细节讲得不如专门的干涉测量教材深入，但很擅长把“图像、噪声、角分辨率、动态范围”放回完整的射电观测语境。
* *Tools of Radio Astronomy*：更适合补充接收机、观测系统和误差来源的背景，让去卷积不至于被理解成脱离仪器的纯数值问题。

### 6.x.3 CLEAN 家族与现代去卷积

* [<cite data-cite='1974A&AS...15..417H'>Högbom (1974)</cite> &#10548;](http://adsabs.harvard.edu/abs/1974A%26AS...15..417H)：经典图像域 `CLEAN` 的起点，也是理解“点源字典 + 逐步减去 PSF”最直接的论文。
* [<cite data-cite='1980A&A....89..377C'>Clark (1980)</cite> &#10548;](http://adsabs.harvard.edu/abs/1980A%26A....89..377C)：把 `CLEAN` 从小规模教学算法推进到大数据量可执行的工程实现。
* [<cite data-cite='1984AJ.....89.1076S'>Schwab (1984)</cite> &#10548;](http://adsabs.harvard.edu/abs/1984AJ.....89.1076S)：把 major / minor cycle 的思想更清楚地放进真实成像流程。
* [<cite data-cite='2008ISTSP...2..793C'>Cornwell (2008)</cite> &#10548;](http://arxiv.org/abs/0806.2228)：多尺度 `CLEAN`，适合把“天空先验不再是纯点源”这件事真正想明白。
* [<cite data-cite='2011A&A...532A..71R'>Rau & Cornwell (2011)</cite> &#10548;](http://arxiv.org/abs/1106.2745)：多尺度多频率联合去卷积，是宽带连续谱成像的关键入口。
* [<cite data-cite='2014MNRAS.444..606O'>Offringa et al. (2014)</cite> &#10548;](http://arxiv.org/abs/1407.1943)：`WSClean` 的代表性论文，适合观察现代成像器怎样把宽视场、并行化和多种去卷积策略组合到统一软件里。
* [<cite data-cite='2015A&A...576A...7D'>Dabbech et al. (2015)</cite> &#10548;](http://arxiv.org/abs/1412.5387)：如果你想越过传统 `CLEAN`，这是进入稀疏重构与更一般反问题方法的一个自然入口。

### 6.x.4 宽视场、模型先验与联合推断

* [<cite data-cite='1992A&A...261..353C'>Cornwell & Perley (1992)</cite> &#10548;](http://adsabs.harvard.edu/abs/1992A%26A...261..353C)：宽视场成像与非共面基线问题，是理解“为什么去卷积不能脱离成像几何”最经典的文献。
* [<cite data-cite='2002ApJ...570..447C'>Chang & Refregier (2002)</cite> &#10548;](http://arxiv.org/abs/astro-ph/0107085)：用 shapelet 基做源建模，帮助你把“模型”从 delta 函数或高斯进一步推广到更灵活的函数基。
* [<cite data-cite='2015MNRAS.450.1308L'>Lochner et al. (2015)</cite> &#10548;](http://arxiv.org/abs/1501.05304)：贝叶斯建模与不确定性传播视角，提醒我们去卷积不只是得到一张“好看的图”，还涉及模型比较与误差刻画。
* 本章当前只建立了这些方向的入门接口，还没有系统展开方向相关去卷积、联合校准-成像、偏振宽带联合重建、mosaic / single-dish 联合去卷积等更高级主题。继续深入时，建议把它们视为第 6 章的自然延伸，而不是彼此无关的新话题。

### 6.x.5 源搜索、局部噪声与目录构建

* [PyBDSF 文档 &#10548;](https://pybdsf.readthedocs.io/en/latest/)：可直接对应本章 `6.5` 的许多概念，包括均值与 RMS 图、Gaussian 拟合、Gaussian 分组、shapelet / wavelet 描述以及偏振和谱指数相关量。
* [AegeanTools 文档 &#10548;](https://aegeantools.readthedocs.io/en/stable/)：除了 `aegean` 本体，还包含 `BANE` 等和局部背景、局部噪声估计相关的工具，特别适合把“局部阈值”从概念推进到可执行流程。
* [Hancock et al. (2012): *Compact continuum source-finding for next generation radio surveys* &#10548;](https://arxiv.org/abs/1202.4500)：讨论下一代巡天条件下源搜索器的完整率、可靠率与参数恢复问题。
* [Hancock et al. (2018): *Aegean 2.0* &#10548;](https://arxiv.org/abs/1801.05548)：从工程与工作流角度说明源搜索器如何适应更大规模图像与更多科学场景。
* [Hydra II: source finder comparison &#10548;](https://arxiv.org/abs/2304.14357)：如果你准备真正做巡天目录，这类基准测试很重要，因为它会告诉你 `PyBDSF`、`Aegean`、`Selavy` 等软件在不同源形态和噪声条件下各自的偏差与盲点。
* 学到这里，最重要的观念不是“阈值设成几 sigma”，而是：背景是否平稳、局部 RMS 是否可信、相邻源是否需要 deblending、拟合模型是否与合成波束一致、目录参数是否已经包含主波束和系统误差的影响。

### 6.x.6 软件与实践建议

* [CASA `tclean` 文档 &#10548;](https://casadocs.readthedocs.io/en/latest/api/tt/casatasks.imaging.tclean.html)：官方文档把 `hogbom`、`multiscale`、`mtmfs`、`auto-multithresh`、`wproject` 等选项放在同一任务框架中，适合训练你从“参数组合”角度理解去卷积流程。
* [WSClean deconvolution 手册 &#10548;](https://wsclean.readthedocs.io/en/latest/deconvolution.html)：当前文档专门整理了 `auto-masking`、`local RMS thresholding`、`wideband deconvolution`、`multi-scale cleaning`、`MORESANE` 和压缩感知相关模式，非常适合与本章的 `6.3` 到 `6.5` 对照阅读。
* 对大多数初学者，一个稳妥的训练顺序是：
  1. 先把 `6.2` 到 `6.4` 的示例重新跑一遍，并且能解释每一次迭代为何停止、残图里还剩下什么。
  2. 再修改 `6.3` 到 `6.5` 里的源形态、噪声水平、阈值和 PSF 宽度，观察算法何时开始失效。
  3. 最后拿一幅真实连续谱图像，用独立的源搜索器生成目录，并和残图、RMS 图、波束参数一起做交叉检查。
* 如果你在真实数据里只追求“把图清干净”，通常很容易过度去卷积；如果你只追求“源表尽量多”，又很容易把伪影写进目录。成熟的数据处理习惯，是把去卷积、图像质量评估和源搜索视为一个闭环，而不是三个彼此孤立的步骤。
