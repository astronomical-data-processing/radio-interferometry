## 4.7 闭合量、缺短间距与 mosaicking 的可见度图像

第 4.6 节说明了可见度和天空亮度之间的 Fourier 关系。本节继续讨论三个在真实数据中反复出现、却容易被基础公式掩盖的问题：哪些信息不依赖天线增益而仍可用于判断源结构，缺少短基线为什么会破坏扩展源通量，mosaicking 为什么既是像面操作也是可见度空间操作。

这三个问题都说明同一件事：可见度不是一组可随意插值的图像像素，而是带有几何、仪器和采样结构的复测量量。闭合相位利用三条基线的相位和抵消天线相位误差；缺短间距说明 $uv$ 平面中心区域携带总通量和大尺度结构；mosaic 则通过多个指向改变主波束加权，使像面和 Fourier 空间中的信息被重新组合。

### 4.7.1 闭合相位与闭合振幅

若只考虑方向无关天线增益，基线 $ij$ 上的观测可见度可写为

$$
V_{ij}^{m obs}=g_i g_j^st V_{ij}^{m true},
$$

其中 $g_i=|g_i|e^{i\phi_i}$ 是天线 $i$ 的复增益。单条基线相位包含源结构相位和两个天线相位之差，因此并不能直接等同于天空结构。三台天线形成闭合三角形时，三条基线相位相加得到

$$
\Psi_{ijk}=rg V_{ij}^{m obs}+rg V_{jk}^{m obs}+rg V_{ki}^{m obs}.
$$

把每条基线中的天线相位展开，$\phi_i-\phi_j+\phi_j-\phi_k+\phi_k-\phi_i$ 正好抵消，剩下的只与真实可见度相位有关。这就是闭合相位。类似地，四台天线可以构造闭合振幅，例如

$$
C_{ijkl}= {|V_{ij}^{m obs}|\,|V_{kl}^{m obs}|\over |V_{ik}^{m obs}|\,|V_{jl}^{m obs}|},
$$

在理想天线增益模型下，天线振幅因子也会抵消。闭合量不需要知道每个天线的绝对相位或振幅，因此在自校准、VLBI 和质量诊断中非常重要。

![闭合相位三角形](figures/closure_phase_triangle.png)

**图 4.7.1** 三台天线构成的闭合三角形。每条基线相位都含有两个天线相位，但三条基线相位求和后，天线相位项互相抵消。闭合相位保留了源结构信息，同时对天线型相位误差不敏感。

闭合量并不是万能的。它只能抵消可写成天线因子的误差，不能抵消基线相关误差、方向相关误差、偏振泄漏、时间/频率平均误差和真实天空模型不完备带来的结构偏差。热噪声也会直接进入闭合量，低信噪比三角形的闭合相位会变得不稳定。因此，闭合量最适合作为“模型是否与数据中不依赖天线增益的部分一致”的诊断工具，而不是取代校准和成像的捷径。

在实践中，若自校准后普通残差明显改善，但闭合相位残差并未改善，问题往往不在天线相位表本身，而在天空模型、方向依赖项或数据选择。反过来，若闭合量稳定而天线解剧烈震荡，可能说明求解间隔过短、参考天线选择不佳或增益模型自由度过多。

![闭合量诊断时间序列](figures/closure_diagnostic_timeseries.png)

**图 4.7.2** 天线相位漂移会使单条基线相位随时间明显变化，但闭合相位和可以保持稳定。真实数据中，闭合量常用于区分天线型校准问题和源结构/模型问题。

### 4.7.2 缺短间距与总通量

Fourier 关系给出一个直接结论：$u=v=0$ 处的可见度对应表观天空的总通量。更一般地，靠近 $uv$ 原点的短基线约束大角尺度结构，远离原点的长基线约束小角尺度结构。若阵列没有足够短的基线，大尺度亮度分布就没有被测量到；即使长基线成像看起来有结构，扩展源的总通量和背景水平仍然会不可靠。

常用的最大可恢复尺度近似写作

$$
	heta_{m MRS}\simeq \eta {\lambda\over B_{\min}},
$$

其中 $B_{\min}$ 是最短有效基线，$\eta$ 是依赖权重、采样和成像细节的数量级因子。这个公式不是严格截止边界，而是提醒大于该尺度的结构会迅速变得难以恢复。缺短间距在图像中常表现为扩展源周围的 negative bowl、总通量低估、背景被迫接近零，以及多尺度 CLEAN 难以稳定判断大尺度分量。

![缺短间距的 Fourier 视图](figures/missing_short_spacing_fourier_view.png)

**图 4.7.3** 缺短间距的三个等价视角。左图显示 `uv` 平面中心区域未被采样；中图显示扩展源可见度的主要功率集中在低空间频率；右图展示短间距滤波后，宽广发射会被削弱并伴随负碗状结构。

缺短间距不是去卷积算法独自能解决的问题。算法可以在先验约束下选择某个看似合理的大尺度模型，但未被测量的 Fourier 分量无法凭空从数据中产生。单碟总功率、紧凑阵列配置、多配置合并和 joint deconvolution 的价值，正是补充不同空间频率范围的约束。若科学目标是扩展星系的总 H I 通量、星形成区的弥散分子气体或星系团低面亮度晕，短基线和单碟信息常常比最高角分辨率更重要。

### 4.7.3 Mosaicking 的双重图像

单个指向观测到的是主波束加权后的天空。若主波束为 $A(l,m)$，单个指向的表观亮度近似为 $A(l,m)I(l,m)$。在 Fourier 空间中，像面相乘对应可见度空间卷积；换言之，主波束的孔径函数会把邻近空间频率的信息混合到每个测量中。多个指向的 mosaic 通过改变主波束中心，在像面上覆盖更大区域，同时在 Fourier 空间中提供不同相位加权的约束。

这就是 mosaic 与普通拼图的区别。它不是简单把几张已经成好的图剪切相加，而是在测量方程层面引入多个指向的主波束。joint mosaic imaging 会同时考虑不同指向的主波束响应、噪声权重和重叠区域；若还加入单碟数据或紧凑阵列数据，便可以更好地约束大尺度结构。

![mosaicking 的像面与 Fourier 视图](figures/mosaic_image_uv_dual_view.png)

**图 4.7.4** Mosaic 的双重视角。像面中，不同指向给出不同主波束加权的表观天空；Fourier 空间中，主波束乘法对应孔径函数卷积，不同指向引入不同相位加权。Mosaic 能改善视场覆盖和部分低空间频率约束，但不能替代真正的零间距总功率测量。

### 4.7.4 教学案例：扩展星系的通量为什么少了一半

一个常见案例是附近星系的 H I 或低频连续谱晕。长基线数据给出漂亮的细节结构，峰值信噪比也很高，但与单碟或低分辨率巡天相比，总通量明显偏低。此时若只提高 CLEAN 迭代次数，通常不会解决问题，反而可能把残余旁瓣和噪声解释成大尺度成分。

合理的诊断顺序应从 Fourier 空间开始。首先检查最短基线对应的最大可恢复尺度是否小于目标角尺度；随后查看 `uv` 权重是否过度偏向长基线，是否使用了会进一步压低短基线贡献的均匀权重；再比较不同 taper 和不同阵列配置下的积分通量。若短基线覆盖本身不足，结论应降级为“高空间频率结构可靠，总通量和弥散成分不完整”。若科学问题必须依赖总通量，就需要加入紧凑配置、单碟、mosaic 或联合成像约束。

闭合量在这个案例中也有作用。若紧致源结构的闭合相位与模型一致，而扩展通量仍然缺失，问题可能主要是采样范围而不是天线相位校准；若闭合相位本身与模型不一致，天空模型或方向相关误差也必须同时检查。

### 4.7.5 本节结论

闭合量、缺短间距和 mosaic 都是可见度空间思维的直接延伸。闭合相位和闭合振幅揭示了哪些结构信息不依赖天线型增益；短基线和零间距决定扩展源总通量和大尺度结构是否被测量；mosaicking 则把主波束加权、多指向覆盖和 Fourier 空间卷积联系起来。进入第 5 章成像时，权重、网格化、视场大小和去卷积策略都应回到这些可见度约束中理解。
