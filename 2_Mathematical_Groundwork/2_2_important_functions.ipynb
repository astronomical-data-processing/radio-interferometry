## 2.2 干涉测量中常见的函数原型<a id='math:sec:important_functions'></a>

上一节用复数建立了振幅与相位的语言；本节转向另一类基础对象：函数形状。射电干涉测量中的许多公式，最终都可以归结为少数几种函数原型的组合。高斯函数描述平滑而局域的响应，矩形窗描述有限孔径、有限时间和有限带宽，`sinc` 函数描述矩形窗在频域中的旁瓣，delta 函数描述点源或理想采样点，Shah 函数则描述规则采样网格。

这些函数不应被看作孤立的定义清单。它们更像一套“形状词汇”：看到高斯，要联想到波束宽度和归一化面积；看到矩形窗，要联想到有限观测范围；看到 `sinc`，要联想到旁瓣、零点和分辨率；看到 delta 与 Shah 函数，要联想到点源、抽样和频谱复制。下面的安排从最容易画出的连续函数开始，逐步过渡到分布和采样函数。


### 2.2.1 高斯函数：平滑、局域与波束面积<a id='math:sec:gaussian_function'></a>

一维高斯函数可写为

<a id='math:eq:gaussian_1d'></a><!--\label{math:eq:gaussian_1d}-->
$$
g(x)=a\exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right].
$$

这里 $a$ 是峰值，$\mu$ 是中心位置，$\sigma$ 控制宽度。高斯函数在射电天文学中很常见：热噪声常用高斯统计描述，单碟主波束和合成波束的主瓣也常被近似为高斯。这个近似不是因为真实仪器响应必然严格高斯，而是因为高斯函数足够平滑、局域，并且由少数几个参数就能描述中心位置、宽度和总面积。

实际观测里常用半高全宽（full width at half maximum, FWHM）而不是 $\sigma$ 描述波束大小。令 $\mu=0$ 并求 $g(x)=a/2$ 的位置，得到

$$
\exp\left(-\frac{x^2}{2\sigma^2}\right)=\frac{1}{2},
$$

因此

<a id='math:eq:gaussian_fwhm'></a><!--\label{math:eq:gaussian_fwhm}-->
$$
x=\pm\sqrt{2\ln 2}\,\sigma,
\qquad
\mathrm{FWHM}=2\sqrt{2\ln 2}\,\sigma\simeq 2.355\sigma.
$$

![高斯函数的宽度约定与二维波束模型](figures/important_functions_gaussian.png)

**图 2.2.1** 高斯函数的宽度约定与二维椭圆高斯模型。左图标出 $\sigma$ 与 FWHM 的关系；右图显示二维波束常用的主轴、次轴和位置角描述。


高斯函数的另一个重要性质是面积可以解析计算。先考虑峰值为 1、中心在原点的一维情形：

$$
I=\int_{-\infty}^{+\infty}\exp\left(-\frac{x^2}{2\sigma^2}\right)dx.
$$

将积分平方，并把二维积分改写为极坐标，得到

$$
\begin{aligned}
I^2
&=\int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty}
\exp\left[-\frac{x^2+y^2}{2\sigma^2}\right]dxdy\\
&=2\pi\int_0^{\infty}r\exp\left(-\frac{r^2}{2\sigma^2}\right)dr
=2\pi\sigma^2.
\end{aligned}
$$

所以

<a id='math:eq:gaussian_integral'></a><!--\label{math:eq:gaussian_integral}-->
$$
\int_{-\infty}^{+\infty}a\exp\left[-\frac{(x-\mu)^2}{2\sigma^2}\right]dx
=a\sqrt{2\pi}\sigma.
$$

当 $a=1/(\sqrt{2\pi}\sigma)$ 时，高斯函数面积为 1，便成为均值为 $\mu$、方差为 $\sigma^2$ 的正态分布概率密度。这个归一化思想在后面处理噪声模型和点扩散函数时都会出现。


二维高斯函数在波束近似中尤其重要。若先不考虑旋转，峰值归一化的椭圆高斯可写为

<a id='math:eq:elliptical_gaussian'></a><!--\label{math:eq:elliptical_gaussian}-->
$$
G(l,m)=\exp\left[-\frac{1}{2}\left(\frac{l^2}{\sigma_l^2}+\frac{m^2}{\sigma_m^2}\right)\right].
$$

由于 $l$ 与 $m$ 两个方向可以分离，面积为

$$
\int\!\!\int G(l,m)\,dl\,dm=2\pi\sigma_l\sigma_m.
$$

若椭圆高斯被旋转一个位置角，面积仍然不变。原因是二维旋转矩阵的行列式为 1；它只改变坐标轴方向，不改变面积元素。把 $\sigma_l$、$\sigma_m$ 换成两个方向的半高全宽 $\theta_l$、$\theta_m$，得到射电天文学中常见的高斯波束面积近似

<a id='math:eq:gaussian_beam_area'></a><!--\label{math:eq:gaussian_beam_area}-->
$$
\Omega_{\rm G}
=\frac{\pi}{4\ln 2}\,\theta_l\theta_m
\simeq 1.1331\,\theta_l\theta_m.
$$

这里的角宽必须使用弧度，计算结果才是立体角。后面从 Jy/beam 转换到 Jy/sr、估计亮温或比较不同分辨率图像时，这个因子会频繁出现。


### 2.2.2 阶跃函数与矩形窗：有限范围的数学表达<a id='math:sec:boxcar_and_rectangle_function'></a>

许多观测过程都有明确的“开”与“关”。接收机只在某个时间段积分，频谱通道只覆盖有限带宽，孔径只在天线口径范围内非零。阶跃函数和矩形窗函数就是描述这种有限范围的最简单工具。

Heaviside 阶跃函数定义为

<a id='math:eq:heaviside'></a><!--\label{math:eq:heaviside}-->
$$
H(x)=
\begin{cases}
0, & x<0,\\
1, & x>0.
\end{cases}
$$

在 $x=0$ 处取 $0$、$1/2$ 或 $1$ 都是常见约定；对积分和傅里叶变换而言，单个点的取值通常不影响结果。由两个阶跃函数相减，可以构造区间 $[a,b]$ 上的方窗：

$$
\Pi_{a,b}(x)=H(x-a)-H(x-b).
$$

最常用的居中矩形函数写为

<a id='math:eq:rectangle'></a><!--\label{math:eq:rectangle}-->
$$
\Pi(x)=
\begin{cases}
1, & |x|\le \frac{1}{2},\\
0, & |x|>\frac{1}{2}.
\end{cases}
$$

矩形窗看起来简单，却会在频域产生长尾旁瓣；这正是有限观测范围带来谱泄漏和旁瓣结构的根源之一。


### 2.2.3 `sinc` 函数：矩形窗的频域阴影<a id='math:sec:sinc_function'></a>

归一化的 `sinc` 函数定义为

<a id='math:eq:sinc_normalized'></a><!--\label{math:eq:sinc_normalized}-->
$$
\operatorname{sinc}(x)=
\begin{cases}
1, & x=0,\\
\dfrac{\sin(\pi x)}{\pi x}, & x\ne 0.
\end{cases}
$$

未归一化形式常写作 $\operatorname{sinc}_{\rm u}(x)=\sin x/x$。文献中两种约定都很常见，因此必须检查定义；NumPy 的 `np.sinc(x)` 使用的是上面的归一化形式。

`sinc` 函数最重要的来源，是矩形窗的傅里叶积分。虽然后面才会系统建立傅里叶变换，这里可以先直接算出一个核心恒等式：

<a id='math:eq:rect_to_sinc'></a><!--\label{math:eq:rect_to_sinc}-->
$$
\begin{aligned}
\int_{-1/2}^{1/2}e^{-2\pi iux}\,dx
&=\left.\frac{e^{-2\pi iux}}{-2\pi iu}\right|_{-1/2}^{1/2}\\
&=\frac{e^{-\pi iu}-e^{\pi iu}}{-2\pi iu}
=\frac{\sin(\pi u)}{\pi u}
=\operatorname{sinc}(u).
\end{aligned}
$$

因此，一个在原始域中边界很硬的矩形窗，在傅里叶域中会变成带有振荡旁瓣的 `sinc` 响应。有限孔径、有限积分时间和有限带宽都会产生类似结构；后面讨论时间平均 smearing、带宽 smearing 和谱泄漏时，这个图像会直接派上用场。

![矩形窗与 sinc 响应](figures/important_functions_window_sinc.png)

**图 2.2.2** 矩形窗及其对应的 `sinc` 型响应。矩形窗越宽，`sinc` 主瓣越窄；矩形窗越窄，频域响应越宽。这是分辨率与观测范围互相制约的早期体现。


### 2.2.4 狄拉克 delta 函数：点源、脉冲与抽样点<a id='math:sec:dirac_delta_function'></a>

狄拉克 delta 并不是普通函数，而是一个分布。它最重要的性质不是“在原点无穷高、别处为零”这样的图像化描述，而是抽样性质：对足够良好的函数 $f$，有

<a id='math:eq:delta_sampling'></a><!--\label{math:eq:delta_sampling}-->
$$
\int_{-\infty}^{+\infty} f(x)\delta(x-a)\,dx=f(a).
$$

这个式子说明，$\delta(x-a)$ 可以从连续函数中抽取 $x=a$ 处的值。射电天文学中，理想点源可以用天球上的 delta 分布表示；理想化的采样点也可以用 delta 分布表示。点源之所以在傅里叶域中产生纯相位斜坡，根源也在这个抽样性质。

一种较稳妥的直觉，是把 delta 看成一族面积恒为 1、宽度逐渐缩窄的函数极限。例如

<a id='math:eq:delta_gaussian_limit'></a><!--\label{math:eq:delta_gaussian_limit}-->
$$
\delta_\epsilon(x)=\frac{1}{\sqrt{2\pi}\epsilon}
\exp\left(-\frac{x^2}{2\epsilon^2}\right),
\qquad
\delta(x)=\lim_{\epsilon\to 0}\delta_\epsilon(x)
$$

这个极限应当在积分意义下理解，而不是逐点理解。对任意固定的 $x\ne 0$，$\delta_\epsilon(x)$ 的确趋于 0；但总面积始终为 1，并在 $\epsilon\to 0$ 时集中到原点。

![delta 函数的窄化极限](figures/important_functions_delta.png)

**图 2.2.3** delta 函数的窄化极限。几条高斯曲线面积相同，宽度越窄，峰值越高；理想 delta 是这种“单位面积集中到一点”的分布极限。


delta 函数还满足两个常用关系。第一，缩放自变量会改变 delta 的归一化：

<a id='math:eq:delta_scaling'></a><!--\label{math:eq:delta_scaling}-->
$$
\delta(ax)=\frac{1}{|a|}\delta(x).
$$

这个因子来自变量替换。因为 $\delta(ax)$ 在 $x=0$ 附近抽样，但 $u=ax$ 使得 $du=|a|dx$；若没有 $1/|a|$，积分面积就不再等于 1。第二，阶跃函数的广义导数是 delta 函数：

$$
\frac{dH}{dx}=\delta(x).
$$

这表明“突然打开的边界”在导数意义下对应一个集中在边界处的脉冲。这个关系在处理窗函数边界、格林函数和响应函数时很有用。


### 2.2.5 Shah 函数：规则采样的理想模型<a id='math:sec:shah_function'></a>

单个 delta 函数表示一个理想采样点；规则排列的无穷多个 delta 函数就构成 Shah 函数，也称 Dirac comb。令采样间隔为 $T$，定义

<a id='math:eq:shah_function'></a><!--\label{math:eq:shah_function}-->
$$
\operatorname{III}_T(x)=\sum_{n=-\infty}^{+\infty}\delta(x-nT).
$$

若连续信号为 $f(x)$，用间隔 $T$ 进行理想采样可写为

<a id='math:eq:ideal_sampling_comb'></a><!--\label{math:eq:ideal_sampling_comb}-->
$$
f_s(x)=f(x)\operatorname{III}_T(x)
=\sum_{n=-\infty}^{+\infty}f(nT)\delta(x-nT).
$$

这个公式把“每隔 $T$ 取一个样本”的操作写成了普通乘法与 delta 抽样性质的组合。它是第 2.9 节采样理论的入口。

![Shah 函数与规则采样](figures/important_functions_shah_sampling.png)

**图 2.2.4** Shah 函数与规则采样。连续信号乘以一个 delta comb 后，只在采样位置留下加权脉冲。


Shah 函数最深刻的性质，是它在傅里叶变换下仍然是 Shah 函数，只是采样间隔取倒数并带有归一化因子。若采用后面将使用的约定

$$
\mathcal{F}\{f\}(u)=\int_{-\infty}^{+\infty}f(x)e^{-2\pi iux}\,dx,
$$

则

<a id='math:eq:shah_fourier_pair'></a><!--\label{math:eq:shah_fourier_pair}-->
$$
\mathcal{F}\{\operatorname{III}_T(x)\}
=\frac{1}{T}\operatorname{III}_{1/T}(u).
$$

因此，原始域中的规则采样会在傅里叶域中造成周期复制。采样间隔越大，频域复制越密；采样间隔越小，频域复制越疏。混叠现象正是这些复制谱相互重叠后的结果。第 2.9 节会把这个直觉发展成采样定理。


### 2.2.6 本节的概念线索

这一节的函数可以串成一条很短的逻辑链。高斯函数提供平滑局域响应的标准模型，并给出波束面积的常用近似；阶跃函数和矩形窗描述有限范围，而有限范围在傅里叶域中自然产生 `sinc` 型旁瓣；delta 函数把点源和理想采样点写成分布；Shah 函数则把规则采样写成一串 delta 函数。这样，从“一个波束是什么形状”到“有限采样如何改变频域结构”，后续傅里叶章节所需的主要函数原型就已经准备好了。
