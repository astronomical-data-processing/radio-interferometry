本节讨论展宽（smearing）现象，以及为了避免时间展宽和带宽展宽，需要满足哪些观测条件。我们将结合 JVLA 的一些实际例子，说明不同积分区间设置会如何影响天空中不同位置源的成像结果。

# 时间展宽与带宽展宽 <a id='time-and-bandwidth-smearing'></a>

时间展宽和带宽展宽也常被统称为平均效应（averaging）。它们出现在可见度被按有限的时间区间和频率区间进行平均时（参见 Bridle 和 Schwab 的讨论）。这种平均首先发生在相关器中，因为相关器输出的本来就是对一段有限时间和有限频率宽度上的平均测量值。除此之外，在后续数据处理中，人们还可能为了节省内存或降低计算成本，再次对可见度做进一步平均。

从观测效果上看，展宽通常表现为离轴源振幅下降。其直观理解如下：对于一个位于方向单位向量 $\boldsymbol{\sigma}$ 上、流量为 $S$ 的点源，其可见度相位会随着频率和时间发生变化。如果我们把一段相位快速旋转的信号直接做平均，那么平均后的复数向量长度就会变短，于是表观振幅减小。

因此，源越偏离相位中心、基线越长、时间平均区间越大、频率平均宽度越宽，展宽往往就越明显。

### 频率积分引起的相位变化 <a id='phase-changes-as-a-function-of-accumulation-frequency'></a>

由 $c=\lambda\nu$ 可知，当观测频率从 $\nu$ 变化到 $\nu+\Delta\nu$ 时，频率方向上的相位变化可写为

$\Delta \Phi_{\nu} = \frac{2\pi}{c}(\nu+\Delta \nu)~ \bf{u}\cdot(\bf{\sigma}-\bf{\sigma}_0)$，

其中 $\Delta\nu$ 表示频率步长，也就是单次平均所对应的带宽。随着 $\Delta\nu$ 增大，上式中的相位变化也会更快，因此更容易在频率平均过程中产生带宽展宽。

### 时间积分引起的相位变化 <a id='phase-changes-as-a-function-of-accumulation-times'></a>

基线向量 $\bf{u}=\bf{u}(t)$ 会随时间变化。设在一组积分时刻 $t_1, \cdots, t_n$ 上的可见度被平均成一个单独的可见度点 $V$。

此时时间方向上的相位变化可写为

$\Delta \Phi_{t} = \frac{2\pi}{\lambda} \bf{u}^{a}\cdot(\bf{\sigma}-\bf{\sigma}_0)$，

其中 $\bf{u}^{a}$ 是向量组 $\bf{u}(t_1), \cdots, \bf{u}(t_n)$ 的平均基线向量。随着积分时间区间 $\Delta t=(t_n-t_1)$ 增大，平均过程带来的净振幅损失也会更明显。

同时，展宽效应还会随着基线长度增加以及源偏离相位中心的距离增加而增强。

### 基线长度带来的影响 <a id='phase-changes-as-a-function-of-baseline-length'></a>

当 $\bf{u}_{pq}$ 的模更大，也就是基线更长时，相位变化会更快，因此源振幅 $S$ 会受到指数项
$\exp \{ i\Phi \}$
更强的衰减影响。换句话说，长基线通常比短基线更容易受到展宽的影响。

### 距相位中心越远的源 <a id='phase-changes-as-a-function-of-distance-from-the-phase-centre'></a>

对于远离望远镜指向中心的源，可见度域中的条纹旋转得更快，因此平均之后更容易表现为振幅下降。

这可以直观理解为：源到相位中心的角距离 $||\bf{\sigma}-\bf{\sigma}_0||$ 越大，相位变化就越明显，于是平均后的振幅损失也越大。

除了降低源的表观流量 $S$ 之外，展宽还会扭曲 PSF，因为不同基线所受到的衰减程度并不相同。
