## 9.18 从图像到科学量：通量、亮温、上限与误差预算

射电图像完成后，科学工作并没有结束。图像只是中间产品，真正进入论文、目录或物理模型的往往是通量密度、积分通量、亮温、上限、源大小、谱指数或线宽等科学量。若这些量的测量方法和误差来源没有说清，图像再漂亮也难以支撑可靠结论。

本节把第 9.5 节的基本图像测量扩展成一条更完整的实践链。它的核心问题是：一张单位为 Jy/beam 的 restored image，如何转化为带有方法说明和误差预算的科学量。这个问题贯穿连续谱、谱线和偏振分析，也与第 9.15 节的源表生成、第 9.17 节的 QA 失败模式直接相连。


### 9.18.1 科学测量需要先通过图像 QA

任何测量都应先确认图像产品本身可解释。残差图中若仍有亮源旁瓣、方向相关伪影或主波束边缘噪声，通量测量就会带有难以量化的系统误差。一个基本测量流程应先检查 residual、RMS map 和 synthesized beam，再选择 peak flux、aperture flux、Gaussian fit 或 catalogue measurement 等方法，随后完成单位转换、误差估计和结果报告。


![从图像产品到科学测量的流程](figures/practical_measurement_workflow.png)

**图 9.18.1** 从图像到科学量的测量流程。一个通量或上限若没有说明图像 QA、测量方法、单位转换、误差来源和区域定义，就不是完整的科学测量。


### 9.18.2 Jy/beam 图像中的积分通量

射电干涉恢复图像常用单位 Jy/beam。这个单位表示图像值已经按合成波束归一化，而不是每个像素直接给出 Jy。因此，对扩展区域求积分通量时，应把孔径内像素值求和，再除以以像素为单位的波束面积：

$$
S_{\rm int}=\frac{\sum_{i\in A}(I_i-I_{\rm bg})}{A_{\rm beam,pix}}.
$$

对于 FWHM 主轴和次轴分别为 $b_{\rm maj}$、$b_{\rm min}$ 的高斯恢复波束，若二者用像素为单位，则

$$
A_{\rm beam,pix}=\frac{\pi b_{\rm maj}b_{\rm min}}{4\ln2}.
$$

孔径本身也需要验证。孔径过小会漏掉源翼部，孔径过大会引入更多背景噪声和相邻源污染。一个实用做法是画孔径增长曲线，检查积分通量是否在某个尺度后趋于稳定。


![Jy/beam 图像中的孔径通量与增长曲线](figures/practical_measurement_jybeam_aperture_growth.png)

**图 9.18.2** Jy/beam 图像中的孔径测量。左图显示孔径和合成波束，中图显示孔径与背景环，右图显示积分通量随孔径尺度的增长曲线。通量稳定区比单个任意孔径更有诊断价值。


### 9.18.3 相关噪声：误差不能按像素数简单缩放

干涉图像中的相邻像素并不独立。恢复波束把噪声相关到大约一个 beam 的尺度，gridding、deconvolution 和权重也会引入额外相关结构。因此，孔径通量的不确定度不能简单写成 $\sigma_{\rm pix}\sqrt{N_{\rm pix}}$。更合理的一阶估计，是按孔径内独立 beam 数计算：

$$
N_{\rm beam}\simeq \frac{A_{\rm ap,pix}}{A_{\rm beam,pix}},
\qquad
\sigma_S \simeq \sigma_{\rm rms}\sqrt{N_{\rm beam}}.
$$

这里的 $\sigma_{\rm rms}$ 是 Jy/beam 单位下的局部图像 RMS。这个表达式仍是近似，因为真实图像中残差不一定是平稳高斯噪声；但它比独立像素假设更接近射电图像的统计结构。


![相关噪声与 beam-aware 误差缩放](figures/practical_measurement_correlated_noise.png)

**图 9.18.3** 独立像素噪声和相关图像噪声的对比。孔径误差应按独立 beam 数缩放，而不是按像素数简单缩放。过小的误差估计会让弱源检测和上限显得过于乐观。


### 9.18.4 亮温和上限：同一个 Jy/beam 在不同 beam 下不是同一个物理量

在射电天文学中，亮温常用于比较表面亮度。对高斯 beam，常用近似为

$$
T_B[\mathrm{K}]\simeq
1.222\times10^6\frac{S_\nu[\mathrm{Jy/beam}]}{\nu^2[\mathrm{GHz}]\,\theta_{\rm maj}[\mathrm{arcsec}]\,\theta_{\rm min}[\mathrm{arcsec}]}.
$$

这个公式说明，同样的 Jy/beam 峰值，在不同频率和不同 beam 下对应不同亮温。高分辨率图像的 Jy/beam 噪声可能较高，但对紧致高亮温结构更敏感；低分辨率图像可能更适合测量扩展低面亮度发射。

非探测上限也必须说明 beam、RMS 和区域假设。点源上限通常可用 $n\sigma_{\rm rms}$ Jy/beam 表示；扩展源上限需要乘上独立 beam 数的平方根，并说明假设的源面积。若源面积未知，只给一个单 beam 上限往往不足以约束扩展发射。


![亮温转换与上限示例](figures/practical_measurement_brightness_temperature_upper_limits.png)

**图 9.18.4** 亮温转换和上限估计示意。同一个 Jy/beam 峰值会因频率和 beam 大小对应不同亮温；上限也取决于图像 RMS、beam 和假设源面积。


### 9.18.5 误差预算：热噪声只是其中一项

射电图像测量的误差通常包含随机项和系统项。随机项来自热噪声和局部背景波动；系统项可能来自通量标尺、主波束模型、deconvolution 残差、短间距缺失、校准误差和背景估计。若这些误差近似独立，总误差可用平方和近似：

$$
\sigma_{\rm total}^2=\sigma_{\rm thermal}^2+\sigma_{\rm scale}^2+\sigma_{\rm PB}^2+
\sigma_{\rm bg}^2+\sigma_{\rm deconv}^2+\cdots.
$$

这个公式的意义不是声称所有项都能精确知道，而是要求测量报告显式说明主要误差来源。对于高信噪比亮源，通量标尺或主波束模型可能主导误差；对于弱扩展源，背景和 deconvolution 残差可能比热噪声更重要；对于短间距不足的数据，系统性 missing flux 甚至不能简单写成高斯误差项。


![图像测量误差预算](figures/practical_measurement_uncertainty_budget.png)

**图 9.18.5** 图像测量中的误差预算。热噪声、通量标尺、主波束模型、背景估计和去卷积残差共同决定最终误差。高信噪比并不自动意味着高精度。


### 9.18.6 结果报告应包含方法，而不只是数值

一个完整的射电图像测量结果至少应说明以下信息：使用哪一幅图像，图像单位和 beam 参数是什么；测量采用 peak、aperture、Gaussian fit 还是 catalogue value；孔径或拟合区域如何定义；局部 RMS 如何估计；是否做了 primary beam correction；误差预算包含哪些项；上限对应几倍 RMS 和什么源面积假设。

例如，一个扩展源积分通量不应只写成 $S=1.3\,\mathrm{mJy}$，而应说明它来自某一 restored image 的某个孔径，已扣除局部背景，按 beam 面积完成 Jy/beam 到 Jy 的转换，并给出热噪声、通量标尺和背景项的误差估计。这样，测量结果才可以被复查、比较和重新使用。


### 9.18.7 与真实软件工作流的对应

在 CASA 或 CARTA 中，peak、RMS、孔径统计和区域文件可以用于基础图像测量；`imfit` 可用于高斯拟合；PyBDSF 可用于连续谱源表；spectral cube 场景还需要 moment 图、mask 和通道相关噪声处理。不同工具给出的数值可能略有差别，关键是保持区域定义、beam、背景、RMS 和误差项一致。

本节与第 9.15 节的区别在于：第 9.15 关注自动源表生成，本节关注单个或少数目标的可解释科学测量。实际研究常同时需要二者：源表提供批量候选和统计结果，手工或半自动测量提供关键目标的更细致误差判断。


### 9.18.8 本节结论

图像到科学量的转换，是射电数据处理链中最容易被低估的一步。Jy/beam、beam area、局部 RMS、相关噪声、孔径选择、亮温转换和系统误差都可能改变最终结论。可靠的测量不只需要一个数值，还需要说明这个数值是如何从图像中得到的，以及它的误差边界来自哪里。

完成本节后，第 9 章已经形成一条更完整的实践链：从数据检查和校准，到成像参数、QA、源表和科学测量。后续扩展可以继续把这条链带入谱线物理解释、偏振、短间距和宽场方向相关案例。
