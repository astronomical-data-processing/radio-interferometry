## 2.3 傅里叶级数：从周期展开到谱表示<a id='math:sec:fourier_series'></a>

傅里叶级数是从“波形”走向“频谱”的第一步。上一节已经看到，矩形窗会对应 `sinc` 型响应，delta comb 与规则采样密切相关；这些联系背后共同的语言，就是把函数分解为一组正弦、余弦或复指数分量。

本节只讨论周期函数。周期性带来一个很重要的简化：允许的频率不是连续的，而是基频的整数倍。换句话说，一个周期函数的谱不是一条连续曲线，而是一串离散谐波。连续傅里叶变换会在下一节出现；它可以看成周期趋于无穷大时，谐波间隔逐渐变密后的极限。

![周期波形由谐波叠加而成](figures/fourier_series_harmonic_basis.png)

**图 2.3.1** 周期波形可以由不同频率、振幅和相位的谐波叠加而成。傅里叶级数的任务，是反过来从给定函数中求出这些谐波的系数。


### 2.3.1 周期函数与谐波基底<a id='math:sec:fourier_series_basis'></a>

设函数 $f(t)$ 的周期为 $T$，即

$$
f(t+T)=f(t).
$$

基频角频率定义为

$$
\Omega_0=\frac{2\pi}{T}.
$$

由于 $e^{i m\Omega_0(t+T)}=e^{i m\Omega_0t}e^{i2\pi m}=e^{i m\Omega_0t}$，所有

$$
\phi_m(t)=e^{i m\Omega_0t},\qquad m\in\mathbb{Z},
$$

都具有同样的周期 $T$。这些函数就是傅里叶级数的复指数基底。可以把它们理解为一组频率为 $m/T$ 的纯谐波：$m=0$ 是常数项，$m=1$ 是基频，$m=2$ 是二次谐波，依此类推；负的 $m$ 则对应反向旋转的复相量。


为了说明系数从哪里来，需要给周期函数空间定义一个内积：

<a id='math:eq:periodic_inner_product'></a><!--\label{math:eq:periodic_inner_product}-->
$$
\langle f,g\rangle=\frac{1}{T}\int_{-T/2}^{T/2}f(t)g^*(t)\,dt.
$$

在这个内积下，复指数基底彼此正交：

<a id='math:eq:complex_basis_orthogonality'></a><!--\label{math:eq:complex_basis_orthogonality}-->
$$
\langle \phi_m,\phi_n\rangle
=\frac{1}{T}\int_{-T/2}^{T/2}e^{i(m-n)\Omega_0t}\,dt
=\delta_{mn}.
$$

当 $m=n$ 时，被积函数恒为 1，积分结果为 1；当 $m\ne n$ 时，指数函数在一个周期内完成整数个转圈，正负相位贡献相互抵消，积分为 0。这个正交性是傅里叶系数公式的真正来源。系数不是猜出来的，而是把函数投影到每一个谐波基底上得到的。


### 2.3.2 复系数傅里叶级数<a id='math:sec:complex_fourier_series'></a>

若 $f(t)$ 足够规则，例如分段光滑并且每个周期内只有有限个跳变，则可以写成

<a id='math:eq:complex_fourier_series'></a><!--\label{math:eq:complex_fourier_series}-->
$$
f(t)=\sum_{m=-\infty}^{+\infty}a_m e^{i m\Omega_0t}.
$$

将等式两边与 $e^{i n\Omega_0t}$ 做内积，并利用正交性，得到

$$
\left\langle f,e^{i n\Omega_0t}\right\rangle
=\sum_{m=-\infty}^{+\infty}a_m\left\langle e^{i m\Omega_0t},e^{i n\Omega_0t}\right\rangle
=a_n.
$$

因此复系数为

<a id='math:eq:complex_fourier_coefficients'></a><!--\label{math:eq:complex_fourier_coefficients}-->
$$
a_m=\frac{1}{T}\int_{-T/2}^{T/2}f(t)e^{-i m\Omega_0t}\,dt,
\qquad m\in\mathbb{Z}.
$$

这里采用“分析式中负号、合成式中正号”的约定，后面的连续傅里叶变换也会沿用同一符号习惯。若 $f(t)$ 是实函数，则必有

$$
a_{-m}=a_m^*.
$$

这就是第 2.1 节中实振荡双边频率表示的推广：实信号的正、负频率系数互为复共轭。


在间断点处，傅里叶级数通常不会收敛到函数某一侧的值，而是收敛到左右极限的平均值：

$$
\frac{f(t_+)+f(t_-)}{2}.
$$

这并不是级数“失败”，而是周期谐波基底对跳变边界的自然处理方式。有限项截断时，跳变附近还会出现振铃和过冲；这个现象称为 Gibbs 现象。它在成像中有非常直接的类比：用有限空间频率样本表示带有尖锐结构的天空亮度时，也会出现旁瓣、振铃和局部误差。


### 2.3.3 实系数形式与复系数形式的关系<a id='math:sec:real_fourier_series'></a>

对于实函数，傅里叶级数也常写成正弦和余弦形式：

<a id='math:eq:real_fourier_series'></a><!--\label{math:eq:real_fourier_series}-->
$$
f(t)=\frac{c_0}{2}
+\sum_{m=1}^{\infty}c_m\cos(m\Omega_0t)
+\sum_{m=1}^{\infty}s_m\sin(m\Omega_0t).
$$

利用三角函数的正交性，可得

<a id='math:eq:real_fourier_coefficients'></a><!--\label{math:eq:real_fourier_coefficients}-->
$$
c_m=\frac{2}{T}\int_{-T/2}^{T/2}f(t)\cos(m\Omega_0t)\,dt,
\qquad
s_m=\frac{2}{T}\int_{-T/2}^{T/2}f(t)\sin(m\Omega_0t)\,dt,
$$

其中 $m\ge 1$，而

$$
c_0=\frac{2}{T}\int_{-T/2}^{T/2}f(t)\,dt.
$$

实系数形式强调“余弦分量”和“正弦分量”；复系数形式强调“振幅和相位”。两者完全等价。对 $m>0$，有

<a id='math:eq:real_complex_coeff_relation'></a><!--\label{math:eq:real_complex_coeff_relation}-->
$$
a_m=\frac{1}{2}(c_m-i s_m),
\qquad
a_{-m}=\frac{1}{2}(c_m+i s_m),
\qquad
a_0=\frac{c_0}{2}.
$$

反过来，

$$
c_m=a_m+a_{-m},
\qquad
s_m=i(a_m-a_{-m}).
$$

在射电干涉测量中，复形式通常更自然，因为可见度本身就是复数，幅度和相位不宜拆开处理。


### 2.3.4 矩形脉冲的傅里叶级数：`sinc` 系数与 Gibbs 现象<a id='math:sec:fourier_series_rectangular_pulse'></a>

一个很有代表性的例子，是周期性矩形脉冲。设周期为 $T$，每个周期中只有宽度为 $w$ 的区间取值为 1，其余为 0；在一个中心周期内可写为

$$
f(t)=
\begin{cases}
1, & |t|\le w/2,\\
0, & w/2<|t|\le T/2.
\end{cases}
$$

它的复系数可以直接算出：

<a id='math:eq:rectangular_pulse_coefficients'></a><!--\label{math:eq:rectangular_pulse_coefficients}-->
$$
\begin{aligned}
a_m
&=\frac{1}{T}\int_{-w/2}^{w/2}e^{-i2\pi mt/T}\,dt\\
&=\frac{w}{T}\,\operatorname{sinc}\left(\frac{mw}{T}\right).
\end{aligned}
$$

这条公式把第 2.2 节和傅里叶级数直接连了起来：时域中的矩形窗，导致谐波系数沿着 `sinc` 包络衰减。矩形边界越尖锐，需要的高阶谐波越多；如果只保留有限个谐波，边界附近就会出现振铃。

![周期矩形脉冲的傅里叶级数](figures/fourier_series_square_wave.png)

**图 2.3.2** 周期矩形脉冲的傅里叶级数。左图显示有限阶部分和逐渐逼近矩形脉冲；右图显示系数幅度落在 `sinc` 包络上。跳变附近的过冲不会简单消失，只会随着阶数增加而变窄，这就是 Gibbs 现象。


这个例子也解释了为什么有限孔径、有限频带或有限观测时间总会留下旁瓣结构。只要原始域中存在硬边界，频域系数就不会快速消失；反过来，若希望频域旁瓣更低，通常需要用更平滑的窗函数替代矩形窗。这一思想在频谱分析、成像加权和 tapering 中都会出现。


### 2.3.5 周期 delta comb 的级数表示<a id='math:sec:fourier_series_shah'></a>

Shah 函数也可以用傅里叶级数表示。对周期为 $T$ 的 comb

$$
\operatorname{III}_T(t)=\sum_{n=-\infty}^{+\infty}\delta(t-nT),
$$

每个周期中只有一个 delta 脉冲。它的傅里叶系数为

$$
a_m=\frac{1}{T}\int_{-T/2}^{T/2}\delta(t)e^{-i2\pi mt/T}\,dt=\frac{1}{T}.
$$

因此，在分布意义下有

<a id='math:eq:comb_fourier_series'></a><!--\label{math:eq:comb_fourier_series}-->
$$
\operatorname{III}_T(t)=\frac{1}{T}\sum_{m=-\infty}^{+\infty}e^{i2\pi mt/T}.
$$

这个式子非常值得记住：时间域中的规则脉冲列需要所有谐波以相同幅度叠加才能形成。它也是下一节连续傅里叶变换中“comb 变换后仍是 comb”的周期版本。


### 2.3.6 通向连续傅里叶变换<a id='math:sec:from_series_to_transform'></a>

傅里叶级数的频率只出现在离散位置

$$
\nu_m=\frac{m}{T},
\qquad
\Delta\nu=\frac{1}{T}.
$$

当周期 $T$ 变大时，频率间隔 $\Delta\nu$ 变小；当 $T\to\infty$ 时，离散谐波逐渐变成连续频率变量。直观上，连续傅里叶变换就是把“对离散谐波求和”变成“对连续频率积分”。下一节会把这个极限写成严格的变换公式，并把它与射电干涉测量中的空间频率和可见度联系起来。
