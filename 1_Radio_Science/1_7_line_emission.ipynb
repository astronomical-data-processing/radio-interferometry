## 1.7 谱线辐射<a id='science:sec:line_emission'></a>

连续谱告诉我们源“总体有多亮、谱有多陡、亮温有多高”，而谱线则进一步回答更细致的问题：气体由什么成分组成、以多大速度运动、处于怎样的激发与密度状态。对射电天文学而言，谱线辐射的重要性几乎怎么强调都不为过，因为许多关键科学结论，例如星系旋转曲线、分子云运动学、恒星形成前体的密度结构以及气体的柱密度估计，都建立在对谱线的测量上。

与连续谱相比，谱线数据也对观测和数据处理提出了更高要求。我们不仅要测总通量，还要在频率轴上保留足够精细的结构，把频率转换为速度，并在三维数据立方体中分析空间与运动学信息。因此，本节需要学生建立的不是单一公式记忆，而是一整套“微观跃迁 -> 线轮廓 -> 速度信息 -> 数据立方体 -> 科学解释”的思路。


### 1.7.1 从微观跃迁到辐射传输

在 [$\S$ 1.3](1_3_radiative_transport.ipynb) 中，我们已经讨论过发射系数 $\varepsilon_\nu$ 和吸收系数 $\kappa_\nu$ 在辐射传输中的作用。对谱线辐射而言，这两个量可以直接和原子或分子的能级跃迁联系起来。最常用的语言就是爱因斯坦系数：

* $A_{21}$：从高能级到低能级的自发辐射概率；
* $B_{21}$：高能级到低能级的受激发射概率；
* $B_{12}$：低能级到高能级的受激吸收概率。

这些系数之间并不是独立的，而是由热平衡条件约束。它们之间的关系为：


<a id='science:eq:07_001'></a><!--\label{science:eq:07_001}-->$$
\begin{align}
    g_1 B_{12} \,&=\, g_2 B_{21}\\
    A_{21}\,&=\,\frac{8\pi h\nu_0^3}{c^3}\,B_{21}
    \\
\end{align}
$$

真实谱线并不是无限窄的单一频率尖峰，因此我们还需要引入谱线轮廓函数 $\varphi(\nu)$。线轮廓的宽度可能来自热展宽、湍流展宽、自然展宽、碰撞展宽以及由大尺度速度梯度带来的多普勒展宽。把轮廓函数写进辐射传输表达式后，吸收系数和发射系数可写为：


<a id='science:eq:07_002'></a><!--\label{science:eq:07_002}-->$$
\begin{align}
    \kappa_\nu \,&=\, \frac{h\nu_0}{c}N_1\,B_{12}\left(1-\frac{g_1\,N_2}{g_2\,N_1}\right)\varphi(\nu)\\
       \varepsilon_\nu \,&=\, \frac{h\nu_0}{4\pi}N_2\,A_{21}\varphi(\nu)
   \\
\end{align}
$$

在局部热力学平衡近似下，能级布居可以用激发温度或自旋温度来描述。若两能级布居近似满足玻尔兹曼分布，则有：


<a id='science:eq:07_003'></a><!--\label{science:eq:07_003}-->$$
\begin{align}
    \frac{N_2}{N_1}\,&=\,\frac{g_2}{g_1}e^{-\frac{h\nu}{kT}} \\
\end{align}
$$

这组公式最重要的教学意义在于：**谱线强度不仅取决于“有没有这种元素”，还取决于布居、温度、密度、背景连续谱以及光学深度。** 这就是为什么谱线分析往往比连续谱更富信息量，也更需要谨慎解释。同一条线在不同环境中既可能表现为发射，也可能表现为吸收；而线宽和线形也常常携带着动力学信息，而不仅仅是热信息。


### 1.7.2 谱线为什么特别适合测速度

谱线最有用的特征之一，是它们相对于连续谱非常狭窄，因此多普勒频移可以被高精度测量。当天体气体沿视线方向存在运动时，中心频率会偏离静止频率，从而把速度信息编码到频谱中。射电天文学里常把频率轴换成速度轴，并在数据立方体中逐道分析气体分布。

若采用常见的“射电速度”定义，则有：


<a id='science:eq:07_004'></a><!--\label{science:eq:07_004}-->$$
\begin{align}
    \frac{V_{\rm radio}(\nu)}{c}\,&=\,\frac{\nu_0-\nu}{\nu_0} \\
\end{align}
$$

这一定义提醒我们，谱线观测与连续谱观测在数据结构上有本质差别。连续谱成像主要关心总带宽内的平均强度，而谱线成像必须保留足够细的频率分辨率，否则线会被展宽、稀释甚至完全洗掉。实际处理中，观测者常需要关注：

* 频谱分辨率是否足以分离热展宽与湍流展宽；
* bandpass 标定是否足够稳定，否则频率方向上的系统起伏会伪装成谱线；
* 是否需要做 continuum subtraction，把宽带连续谱从数据中分离出去；
* 成像后如何从 data cube 中构造 channel map、moment 0、moment 1 和 moment 2 图。

换句话说，谱线观测不是在二维图像上多加一个颜色条，而是把成像问题提升成了三维问题。


### 1.7.3 H I 21 cm：射电天文学的基石<a id='science:sec:atomic_hydrogen'></a>

中性氢的 21 cm 超精细跃迁，是射电天文学中最具标志性的谱线之一。它来自氢基态中电子自旋与质子自旋相对取向的变化，静止频率约为 $1.42040575\,\rm GHz$。这条线之所以具有基础地位，并不是因为它最容易激发，而恰恰是因为氢元素极其丰富、而且中性原子氢广泛存在于星系盘、外延气体和星系间相互作用环境中。

21 cm 线的自发辐射概率极低，因此单个原子的跃迁非常罕见，但大量中性氢累积起来后就能形成可测信号。也正因为如此，它在很多场景下都近似光学薄，从而可以把观测量直接转换为柱密度。


在假定 LTE 且自旋温度 $T_{\rm spin}$ 可定义时，中性氢柱密度与光学深度的关系为：


<a id='science:eq:07_005'></a><!--\label{science:eq:07_005}-->$$
\begin{align}
    \frac{N_{\rm H\,\scriptsize I}}{{\rm atoms} \,{\rm cm}^{-2}}\,=\,1.823\times10^{18}\cdot\frac{T_{\rm spin}}{\rm K}\int \tau_\nu(V_{\rm radio})\,\frac{dV_{\rm radio}}{\rm km\,s^{-1}} \\
\end{align}
$$

在更常用的光学薄极限下，上式进一步简化为：


<a id='science:eq:07_006'></a><!--\label{science:eq:07_006}-->$$
\begin{align}
    \frac{N_{\rm H\,\scriptsize I}}{{\rm atoms} \,{\rm cm}^{-2}}\,=\,1.823\times10^{18}\int \frac{T_{\rm b}}{\rm K}\,\frac{dV_{\rm radio}}{\rm km\,s^{-1}} \\
\end{align}
$$

这两个表达式在射电天文学里非常实用，因为它们把观测得到的亮温积分直接转换成气体柱密度。于是，21 cm 数据不仅能告诉我们“哪里有气体”，还能给出“有多少气体”。结合速度信息后，我们还可以绘制星系旋转曲线、研究潮汐尾和气体剥离、分析吸积与外流，乃至通过旋转曲线平坦化现象间接推断暗物质的存在。

学生在这里必须建立一个关键观念：H I 研究并不是只有一张强度图。真正常见的工作流是一个数据立方体，沿频率或速度方向切片后得到一系列 channel map，再进一步构造积分强度图和速度场图。也正因如此，谱线干涉观测对频率设置、速度分辨率、短间距覆盖和通道噪声都比连续谱更敏感。


### 1.7.4 除了 H I，还有哪些重要射电谱线<a id='science:sec:other_line_emission'></a>

射电波段可观测的谱线远不止 21 cm。若把科学用途角度摆在首位，至少应熟悉以下几类：

* **分子转动线**：如 CO 及其同位素线，是追踪冷分子气体最常见的入口；HCN、HCO$^+$、CS 等高临界密度分子则更适合示踪致密气体；
* **射电复合线**：来自高主量子数电子跃迁，可用于研究 H II 区和电离气体的温度、密度与速度场；
* **分子反演或受激辐射线**：如 OH、H$_2$O、CH$_3$OH maser，常见于恒星形成区、晚型恒星包层和活动星系核附近，可提供极高亮温和极精细速度信息；
* **原子精细结构或超精细结构线**：在更高频率上可用于研究冷却过程和不同相位星际介质。

这几类谱线虽然微观机制不同，但在观测上共享一个共同点：都需要把频谱结构保留下来，并把线强、线宽、中心速度和线形一起解释，而不能只看总通量。


### 1.7.5 为什么谱线数据处理与连续谱不同

谱线数据处理是后续实践章节必须专门训练的一条主线。与连续谱相比，它至少有四个额外难点：

* 需要更高频谱分辨率，因此数据量更大、噪声性质也更复杂；
* 常常需要先做连续谱扣除，否则强连续谱会掩盖弱谱线；
* bandpass 误差、RFI 和频率相关校准残差会直接污染谱线结构；
* 对扩展源而言，缺短间距会让某些速度通道中的大尺度通量恢复不全。

因此，学生不应把谱线观测简单理解为“连续谱成像再多一维”。它需要独立的参数选择逻辑，也需要更清楚地理解数据立方体、矩图、速度场和柱密度估计之间的关系。这也是为什么现代射电干涉课程通常都会把谱线处理单独成章。


### 1.7.6 本节应记住什么

本节最关键的结论可以概括为：

* 谱线把微观能级跃迁与宏观气体物理联系起来，是研究成分、温度、密度和速度场的核心工具；
* 线轮廓不只是“宽一点或窄一点”，它本身就包含了热运动、湍流和系统性速度结构的信息；
* H I 21 cm 是最基础的射电谱线之一，能够直接连接气体分布、速度场和柱密度；
* 分子线、复合线和 maser 线分别对应冷分子气体、电离气体和高亮温特殊环境；
* 干涉阵中的谱线处理必须重视频谱分辨率、bandpass 稳定性、continuum subtraction 和数据立方体分析。

掌握这些概念后，再进入后续的天文射电源和干涉测量章节时，学生会更容易理解为什么不同科学目标会对频率覆盖、角分辨率、最大可恢复尺度和处理流程提出完全不同的要求。
