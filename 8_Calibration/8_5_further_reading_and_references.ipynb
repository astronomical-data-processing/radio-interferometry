## 8.5 延伸阅读与参考文献 <a id='cal:sec:refs'></a>

第 8 章把校准组织成一条从测量方程到实际解链的主线。进一步阅读时，材料大致可分为四类：基础干涉测量与 RIME、自校准和闭合量、快速求解器、方向依赖校准与宽场成像。不同材料的重点并不相同，有些适合建立理论框架，有些更适合理解真实管线中的工程选择。


### 8.5.1 校准与测量方程基础

**Thompson, Moran & Swenson, _Interferometry and Synthesis in Radio Astronomy_** 是射电干涉领域最经典的系统教材之一。校准章节应与可见度、综合孔径成像、噪声和极化章节一起理解，因为校准并不是孤立步骤，而是测量方程的一部分。

**Smirnov (2011a), “Revisiting the radio interferometer measurement equation. I. A full-sky Jones formalism”** 建立了现代 RIME 的完整 Jones 形式，是把第 7 章仪器系统和第 8 章校准连接起来的关键文献。

**Smirnov (2011b), “Revisiting the radio interferometer measurement equation. II. Calibration and direction-dependent effects”** 将 RIME 明确推进到校准和方向依赖效应，是理解 3GC 语言和误差分解方式的核心材料。


### 8.5.2 自校准、闭合量与经典成图思路

**Pearson & Readhead (1984), “Image Formation by Self-Calibration in Radio Astronomy”** 是自校准经典文献。该文展示了自校准如何与成像模型共同迭代，也揭示了模型依赖性为什么是自校准的核心风险。

**Readhead & Wilkinson (1978), “The mapping of compact radio sources from VLBI data”** 是 hybrid mapping 思想的重要来源。VLBI 观测中绝对相位信息常常不可得，闭合量和自校准之间的关系在这一背景下非常清楚。

**Cornwell 与 Fomalont 的 synthesis imaging 相关综述** 有助于把闭合相位、闭合振幅、去卷积和图像质量控制联系起来。它们适合作为从理论方程走向实际成图判断的桥梁。


### 8.5.3 高效求解器与 3GC

**Salvini & Wijnholds (2014), “Fast gain calibration in radio astronomy using alternating direction implicit methods”** 介绍了 StEFCal，是理解大阵列快速方向无关增益求解的重要入门材料。它清楚展示了天线基结构如何被算法利用。

**Mitra et al. (2015), “Incorporation of antenna primary beam patterns in radio-interferometric data reduction to produce wide-field, high-dynamic-range images”** 适合与本章 8.4 的主波束例子配合阅读，可帮助理解宽场高动态范围成像中主波束模型为何必须进入处理链。

**Tasse 及相关方向依赖校准工作**，包括 DDFacet、KillMS、faceting 和差分增益等路线，是追踪现代 3GC 实现的重要入口。这些工作展示了物理模型、局部方向求解和成像器耦合之间的实际折中。


### 8.5.4 与第 9 章实践部分的衔接

进入 [第 9 章实践部分](../9_Practical/9_1_visualisation-inspection.ipynb) 后，重点不应只是记住软件命令，而是把命令和物理对象对应起来。`listobs` 与 `plotms` 对应数据质量判断，`gaincal` 与 `bandpass` 对应方向无关增益和频率响应，`applycal` 对应解转移，自校准则对应模型更新与增益求解的闭环。

第 8 章的练习页 [校准习题](8_problem_set.ipynb) 可作为进入实践章前的过渡。它把参考天线、最小二乘收敛、模型不完整和交替增益求解集中到同一组合成数据上，有助于把公式判断转化为数据诊断判断。
