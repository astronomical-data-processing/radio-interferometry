## 2.7 傅里叶定理：移位、缩放与卷积的统一语言<a id='math:sec:fourier_theorems'></a>

傅里叶变换的定义说明了“怎样从一个域走到另一个域”；傅里叶定理说明了“一个域里的操作在另一个域里变成什么”。后者才是傅里叶分析真正好用的原因。移位、缩放、调制、卷积、相关和微分看似是不同操作，经过傅里叶变换后会变成相位因子、坐标缩放、谱搬移、逐点乘法或简单的频率因子。

本节沿用第 2.4 节的约定

$$
F(s)=\mathcal{F}\{f\}(s)=\int f(x)e^{-2\pi ixs}dx,
\qquad
f(x)=\mathcal{F}^{-1}\{F\}(x)=\int F(s)e^{2\pi ixs}ds.
$$

为了突出结构，下面默认积分范围为整个实轴，并假定函数足够良好，使变量替换、交换积分次序和分部积分合法。遇到 delta 函数和 Shah 函数时，相关公式在分布意义下成立。


### 2.7.1 线性：叠加原理的傅里叶形式<a id='math:sec:linearity_of_the_fourier_transform'></a>

傅里叶变换是线性算子：

<a id='math:eq:ft_linearity'></a><!--\label{math:eq:ft_linearity}-->
$$
\mathcal{F}\{a f+b g\}=aF+bG,
\qquad a,b\in\mathbb{C}.
$$

证明直接来自积分的线性：

$$
\int (af+bg)e^{-2\pi ixs}dx
=a\int f e^{-2\pi ixs}dx+b\int g e^{-2\pi ixs}dx.
$$

它的物理意义比证明更重要。若天空可以写成多个点源或多个扩展分量的叠加，则可见度也可以写成各分量可见度的叠加；若测量方程在线性近似下成立，就可以用矩阵把许多基函数的响应组织起来。这是后面成像反演能够成立的基本前提。


### 2.7.2 缩放：窄结构对应宽频谱<a id='math:sec:similarity_theorem'></a>

若 $f_a(x)=f(ax)$，则

<a id='math:eq:similarity_theorem'></a><!--\label{math:eq:similarity_theorem}-->
$$
\mathcal{F}\{f(ax)\}(s)=\frac{1}{|a|}F\left(\frac{s}{a}\right).
$$

变量替换 $u=ax$ 给出证明：

$$
\begin{aligned}
\mathcal{F}\{f(ax)\}(s)
&=\int f(ax)e^{-2\pi ixs}dx\\
&=\frac{1}{|a|}\int f(u)e^{-2\pi iu(s/a)}du
=\frac{1}{|a|}F(s/a).
\end{aligned}
$$

若 $|a|>1$，原始域函数被压窄，傅里叶域函数被展宽；若 $|a|<1$，原始域函数被拉宽，傅里叶域函数被压窄。这个定理是“分辨率需要长基线”“窄脉冲需要宽带宽”这类说法的数学根源。

在多维情形中，若 $\mathbf{x}$ 被非奇异矩阵 $\mathbf{A}$ 线性变换，结果为

<a id='math:eq:multidim_similarity_theorem'></a><!--\label{math:eq:multidim_similarity_theorem}-->
$$
\mathcal{F}\{f(\mathbf{A}\mathbf{x})\}(\mathbf{s})
=\frac{1}{|\det\mathbf{A}|}
F(\mathbf{A}^{-T}\mathbf{s}).
$$

这里的 $|\det\mathbf{A}|$ 是面积或体积缩放因子。二维图像的旋转、拉伸和椭圆波束都可以用这个形式理解。


### 2.7.3 移位与调制：位置写进相位，调制搬移频谱<a id='math:sec:shift_modulation_theorem'></a>

若原始域函数平移 $x_0$，

<a id='math:eq:shift_theorem'></a><!--\label{math:eq:shift_theorem}-->
$$
\mathcal{F}\{f(x-x_0)\}(s)=e^{-2\pi isx_0}F(s).
$$

平移不改变傅里叶幅度，只引入线性相位。证明只需令 $u=x-x_0$：

$$
\int f(x-x_0)e^{-2\pi ixs}dx
=e^{-2\pi isx_0}\int f(u)e^{-2\pi ius}du.
$$

反过来，若在原始域乘以复指数，频谱会被搬移：

<a id='math:eq:modulation_complex_theorem'></a><!--\label{math:eq:modulation_complex_theorem}-->
$$
\mathcal{F}\{e^{2\pi is_0x}f(x)\}(s)=F(s-s_0).
$$

若乘以余弦，由欧拉公式得到两个边带：

<a id='math:eq:modulation_cosine_theorem'></a><!--\label{math:eq:modulation_cosine_theorem}-->
$$
\mathcal{F}\{f(x)\cos(2\pi s_0x)\}(s)
=\frac{1}{2}\left[F(s-s_0)+F(s+s_0)\right].
$$

![移位、缩放与调制](figures/fourier_theorems_shift_scale_modulation.png)

**图 2.7.1** 移位、缩放和调制的傅里叶域表现。平移主要改变相位，压窄结构会展宽频谱，调制会把频谱搬移到新的中心频率附近。


对干涉测量而言，移位定理尤其重要。天空中位于 $(l_0,m_0)$ 的点源会给可见度带来相位因子

$$
\exp[-2\pi i(ul_0+vm_0)].
$$

因此，源离相位中心越远，相位随基线变化越快。许多相位校准、相位中心移动和图像重采样操作，本质上都在使用这条定理。


### 2.7.4 卷积定理与乘积定理<a id='math:sec:convolution_theorem'></a>

卷积定理指出

<a id='math:eq:convolution_theorem'></a><!--\label{math:eq:convolution_theorem}-->
$$
\mathcal{F}\{f*g\}=FG.
$$

证明从定义出发：

$$
\begin{aligned}
\mathcal{F}\{f*g\}(s)
&=\int\left[\int f(t)g(x-t)dt\right]e^{-2\pi ixs}dx\\
&=\int f(t)\left[\int g(x-t)e^{-2\pi ixs}dx\right]dt.
\end{aligned}
$$

在内层积分中令 $u=x-t$，得到

$$
\int g(u)e^{-2\pi i(u+t)s}du=e^{-2\pi its}G(s).
$$

代回外层积分，便有

$$
\mathcal{F}\{f*g\}(s)=G(s)\int f(t)e^{-2\pi its}dt=F(s)G(s).
$$

互补的乘积定理为

<a id='math:eq:product_theorem'></a><!--\label{math:eq:product_theorem}-->
$$
\mathcal{F}\{fg\}=F*G.
$$

在图像域中做卷积，等价于在傅里叶域中相乘；在图像域中相乘，等价于在傅里叶域中卷积。这两句话是理解波束、采样、加权和窗口函数的核心。

![卷积定理示意](figures/fourier_theorems_convolution.png)

**图 2.7.2** 卷积定理示意。直接在原始域做卷积，与先在傅里叶域相乘再反变换，得到同一结果。


在干涉成像中，若 `uv` 平面采样函数为 $S(u,v)$，理想连续可见度为 $V(u,v)$，实际只得到 $S(u,v)V(u,v)$。根据乘积定理，反变换后的脏图像就是理想天空与采样函数反变换的卷积。后者正是脏波束。这就是

$$
I_{\rm D}=I*B_{\rm D}
$$

这一基本关系的傅里叶根源。


### 2.7.5 相关定理与维纳-辛钦定理<a id='math:sec:correlation_theorems'></a>

采用第 2.6 节的互相关定义

$$
(f\star g)(\tau)=\int f^*(t)g(t+\tau)dt,
$$

其傅里叶变换为

<a id='math:eq:cross_correlation_theorem'></a><!--\label{math:eq:cross_correlation_theorem}-->
$$
\mathcal{F}\{f\star g\}=F^*G.
$$

这是因为 $f\star g=f_-^**g$，而 $\mathcal{F}\{f_-^*\}=F^*$。当 $g=f$ 时，得到自相关定理，也就是维纳-辛钦关系：

<a id='math:eq:auto_correlation_theorem'></a><!--\label{math:eq:auto_correlation_theorem}-->
$$
\mathcal{F}\{f\star f\}=|F|^2.
$$

这条定理把自相关和功率谱连接起来。相关器、功率谱估计、噪声自相关和谱线分析都依赖这一思想：时间域或空间域的相关结构，可以等价地用频域的功率分布描述。


### 2.7.6 Parseval 与 Rayleigh：能量在换域中守恒<a id='math:sec:parseval_rayleigh'></a>

Parseval 定理说明，傅里叶变换保持内积：

<a id='math:eq:parseval_theorem'></a><!--\label{math:eq:parseval_theorem}-->
$$
\int f(x)g^*(x)dx=\int F(s)G^*(s)ds.
$$

令 $g=f$，得到 Rayleigh 定理：

<a id='math:eq:rayleigh_theorem'></a><!--\label{math:eq:rayleigh_theorem}-->
$$
\int |f(x)|^2dx=\int |F(s)|^2ds.
$$

这并不是说频域和原始域中的“形状”相同，而是说总能量或平方范数相同。频谱分析中的功率分布、滤波前后的噪声方差估计，以及最小二乘中的正交投影，都可以从这个角度理解。


### 2.7.7 导数定理：微分变成频率乘法<a id='math:sec:derivative_theorem'></a>

若 $f$ 在无穷远处足够快地趋于 0，则分部积分给出

<a id='math:eq:derivative_theorem'></a><!--\label{math:eq:derivative_theorem}-->
$$
\mathcal{F}\left\{\frac{df}{dx}\right\}(s)=2\pi i s F(s).
$$

证明为

$$
\begin{aligned}
\int f'(x)e^{-2\pi ixs}dx
&=\left[f(x)e^{-2\pi ixs}\right]_{-\infty}^{+\infty}
+2\pi is\int f(x)e^{-2\pi ixs}dx\\
&=2\pi isF(s).
\end{aligned}
$$

高阶导数对应更高次的频率因子：

$$
\mathcal{F}\{f^{(n)}\}(s)=(2\pi is)^nF(s).
$$

这说明微分运算会强调高频结构。边缘、尖峰和噪声中的高频分量在导数后会被放大，因此数值微分通常比数值积分更容易受噪声影响。

与之互补的关系是

<a id='math:eq:multiply_by_x_theorem'></a><!--\label{math:eq:multiply_by_x_theorem}-->
$$
\mathcal{F}\{x f(x)\}(s)=-\frac{1}{2\pi i}\frac{dF}{ds},
$$

以及

<a id='math:eq:inverse_derivative_theorem'></a><!--\label{math:eq:inverse_derivative_theorem}-->
$$
\mathcal{F}^{-1}\left\{\frac{dF}{ds}\right\}(x)=-2\pi i x f(x).
$$

这些公式在波前斜率、矩量计算和微分方程的频域解法中经常出现。


### 2.7.8 本节要点<a id='math:sec:fourier_theorems_summary'></a>

傅里叶定理提供了一套换域词典。线性保证叠加可以逐项处理；缩放表达尺度互反；平移变成相位斜坡；调制搬移频谱；卷积变成乘法，乘法变成卷积；相关变成一个谱乘以另一个谱的复共轭；Parseval 和 Rayleigh 保证能量解释在两个域中一致；导数则变成频率乘法。后续进入 DFT、采样、成像和校准时，这些定理会不断作为隐含步骤出现。
