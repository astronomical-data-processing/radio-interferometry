## 2.x 延伸阅读与参考文献

第 2 章的任务，不是替代完整的数学教材，而是为后续的可见度方程、成像、去卷积与校准建立一套足够稳固的工作语言。因此，本节也不追求书目“越多越好”，而是按用途整理几组最值得继续深入的材料：哪些适合巩固傅里叶与信号处理直觉，哪些适合补强线性代数与反问题基础，哪些则能把这些数学工具直接放回射电干涉测量的真实语境中。


### 2.x.1 把数学放回射电干涉语境

* [Essential Radio Astronomy](https://www.cv.nrao.edu/~sransom/web/xxx.html)：NRAO 的开放教材。它的数学要求不高，但非常擅长把傅里叶、采样、干涉条纹和观测量直接放回射电天文学语境，适合与本章配合使用。
* [Interferometry and Synthesis in Radio Astronomy](https://link.springer.com/book/10.1007/978-3-319-44431-4)：如果希望把本章的数学真正连接到可见度方程、阵列几何、成像与校准，这是最重要的系统专著之一。
* [Tools of Radio Astronomy](https://link.springer.com/book/10.1007/978-3-642-39950-3)：兼顾辐射量、观测系统和信号处理，是把“数学工具”与“射电观测语言”结合起来的优秀桥梁。


### 2.x.2 强化傅里叶与信号处理直觉

* Bracewell 的 *The Fourier Transform and Its Applications* 依然是训练傅里叶直觉的经典选择，特别适合帮助读者把函数形状、变换性质和物理解释联系起来。
* Oppenheim 与 Schafer 的 *Signals and Systems* 或 *Discrete-Time Signal Processing* 对离散化、频谱、采样和滤波的讲解非常系统，适合与第 2.8 和第 2.9 节对应阅读。

如果你发现自己“会代公式，但不太能凭直觉判断变换结果会是什么样”，这一组材料通常最有帮助。


### 2.x.3 线性代数、最小二乘与反问题

* Trefethen 与 Bau 的 *Numerical Linear Algebra* 适合补强矩阵分解、条件数和数值稳定性这些后续成像算法非常在意的问题。
* Aster、Borchers 与 Thurber 的 *Parameter Estimation and Inverse Problems* 很适合与第 2.11 节配合阅读，帮助读者把最小二乘从“求一个最佳拟合参数”提升为“如何理解不适定问题与约束”的更大框架。

对于准备进一步深入校准、自校准和高级成像的读者，这一组阅读会非常重要。


### 2.x.4 建议的阅读顺序

对大多数读者，一个比较顺畅的顺序是：

1. 先配合本章主线，把 `2.1`、`2.4`、`2.8`、`2.9`、`2.10`、`2.11` 当作第一轮重点；
2. 同时以 *Essential Radio Astronomy* 或 *Tools of Radio Astronomy* 作为语境参照，避免把数学学成脱离观测的抽象符号；
3. 当你进入第 4 到第 8 章时，再把 *Interferometry and Synthesis in Radio Astronomy* 当作更严格的理论参照书；
4. 如果后续希望深入成像算法、正则化与数值稳定性，再系统补读数值线性代数和反问题教材。

这样的顺序有一个好处：读者不会在一开始就被完整的理论细节压住，而是能沿着“先建立工作语言，再进入测量方程，最后深入数值方法”的路径推进。这也正是本书组织第 2 章的基本思路。
