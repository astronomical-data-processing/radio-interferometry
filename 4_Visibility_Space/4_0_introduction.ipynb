# 第 4 章：可见度空间<a id='visibility:sec:intro'></a>

干涉阵由一系列单元构成，每一对单元做互相关得到一条基线的测量结果。干涉测量是一项强大的技术，通过它可以间接地测得空中天体的准确位置以及空间特征信息。利用综合孔径，我们还能从这些测量结果中获得图像。


本章我们将定义干涉仪的测量量，以及这些量跟我们所观测的源之间的关系。本章主要回答以下问题：

  * 射电干涉究竟测量了什么？
  * 一条干涉基线能提供什么观测量？
  * 如何利用干涉测量结果来推断天空结构？
  
<img src='figures/basic_block.svg' width=60%>

回答这些问题之前，我们将先在 [$\S$ 4.1 &#10142;](4_1_periodic_functions_and_fourier_analysis.ipynb) 讨论图像平面与傅里叶平面之间的关系，这是理解干涉测量的基础。然后，在 [$\S$ 4.2 &#10142;](4_2_the_baseline_and_its_representation_in_space.ipynb) 中定义基线及其在不同参考系中的表示形式。

在 [$\S$ 4.3 &#10142;](4_3_the_2-element_interferometer.ipynb) 中，我们介绍二元干涉仪。它是最简单的干涉仪模型，可以看作对可见度函数 [$\S$ 4.4 &#10142;](4_4_the_visibility_function.ipynb) 进行采样的工具，而可见度函数又与天空亮度分布直接相关。根据干涉仪在地球上的布局，这种采样会沿着相应的 $uv$ 轨迹 [$\S$ 4.5.1 &#10142;](4_5_1_uv_coverage_uv_tracks.ipynb) 进行。在 [$\S$ 4.5.2 &#10142;](4_5_2_uv_coverage_improving_your_coverage.ipynb) 中，我们进一步讨论如何提高干涉阵列的 $uv$ 覆盖，从而更充分地采样可见度函数。

在最后一节 [$\S$ 4.6 &#10142;](4_6_the_fourier_approximation_van_cittert-zernike_theorem.ipynb) 中，我们介绍 van Cittert-Zernike 定理。正是在若干近似条件下，这一定理把天空亮度分布与干涉测量结果之间建立起了傅里叶关系。
