# 第 3 章：干涉测量中的位置天文学<a id='pos:sec:intro'></a>

第 2 章已经准备了球面三角、向量与傅里叶语言；本章要把这些工具接到真实观测几何上。对射电干涉阵而言，“天源在天上哪里”并不是一个孤立的位置问题，而是一条连续的变换链：星表给出赤经和赤纬，观测时刻给出地方恒星时，地方恒星时决定时角，时角与台站纬度共同决定望远镜在本地天空中的方位角与高度角；随后，再围绕相位中心引入方向余弦坐标，为下一章的可见度与成像方程做准备。

这种链条之所以重要，是因为干涉测量把“方向”同时放进了三种语境。第一，它是源目录中的惯性方向；第二，它是某一台望远镜在某一时刻必须执行的本地指向；第三，它是相对于相位中心的局部角偏移，并最终与基线投影形成傅里叶共轭关系。只要其中任意一步的坐标约定、时间约定或符号约定混乱，后面的相位、延迟和成像坐标都会跟着出错。

本章并不试图把位置天文学扩展成完整的天体测量学课程。岁差、章动、光行差、地球定向参数、极移和折射改正等内容，在高精度 VLBI、脉冲星定时或精密定轨中非常重要，但它们通常由专业软件与标准服务统一处理。本章先建立干涉测量中最常反复出现的几何主线：目录位置、观测时刻、台站位置、相位中心和局部像面坐标之间究竟如何对应。


#### 本章的阅读路径

本章可以看成从“天球上的固定方向”走向“阵列实际测量的几何变量”。赤道坐标说明源目录和相位中心如何被定义；地方恒星时与时角把目录方向接到地球自转；地平坐标把这个方向变成望远镜本地指向和可观测性约束；方向余弦则把全天球方向压到相位中心附近的局部像面。下一章的 `u,v,w` 坐标正是在这条路径的末端出现。


#### 本章结构

第 3.1 节首先讨论赤道坐标以及目录位置的参考架和历元问题，说明为什么天源在数据库中通常以 `RA/Dec` 形式出现。第 3.2 节转向时间几何：一旦给定观测时刻，地方恒星时与时角就决定该天源相对于本地子午圈的位置，也预告了地球自转合成为什么能够改变阵列投影。第 3.3 节再把视角从“天球上的固定目录坐标”转移到“台站本地的指向坐标”，并用方位角与高度角讨论可观测性、仰角限制以及观测窗口。第 3.4 节围绕相位中心引入方向余弦坐标 `(l,m,n)`，把天球几何变成局部像面几何，为第 4 章的可见度空间和成像方程作出直接铺垫。第 3.5 节进一步讨论时间标准、参考系和精密天体测量边界，说明坐标链条如何进入相位、延迟和位置误差。
