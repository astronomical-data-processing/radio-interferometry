## 5.x 延伸阅读与参考文献 <a id='imaging:sec:refs'></a>


### 综合成像总论

1. [<cite data-cite='taylor1999synthesis'>Synthesis Imaging in Radio Astronomy II（Lecture 7）</cite> &#10548;](http://adsabs.harvard.edu/abs/1999sira.conf.....T)：本章中关于成像、加权与采样函数的很多基础内容都可追溯到这份讲义。
2. [<cite data-cite='thompson2008interferometry'>Interferometry and Synthesis in Radio Astronomy（Chapter 10）</cite> &#10548;](http://onlinelibrary.wiley.com/book/10.1002/9783527617845)：可与 *Synthesis Imaging in Radio Astronomy II* 第 7 讲配合阅读，是一章很好的补充材料。

### 权重函数

1. [<cite data-cite='briggs1995high'>High Fidelity Deconvolution of Moderately Resolved Sources</cite> &#10548;](http://www.aoc.nrao.edu/dissertations/dbriggs/)：提出鲁棒权重的原始工作，也对各类权重方案作了较系统的综述。

### 宽视场效应

1. [<cite data-cite='taylor1999synthesis'>Synthesis Imaging in Radio Astronomy II（Lecture 19）</cite> &#10548;](http://adsabs.harvard.edu/abs/1999sira.conf.....T)：更深入地讨论宽视场成像问题。
2. [<cite data-cite='cornwell2008noncoplanar'>The noncoplanar baselines effect in radio interferometry: The W-projection algorithm</cite> &#10548;](http://ieeexplore.ieee.org/xpl/login.jsp?tp=&arnumber=4703511&url=http%3A%2F%2Fieeexplore.ieee.org%2Fxpls%2Fabs_all.jsp%3Farnumber%3D4703511)：详细介绍 W-projection 算法。
3. [<cite data-cite='cornwell1992radio'>Radio-interferometric imaging of very large fields: The problem of non-coplanar arrays</cite> &#10548;](http://adsabs.harvard.edu/full/1992A%26A...261..353C)：概述宽视场问题以及多面体解法，并讨论其他相关技术。

### 网格化与去网格化

1. [<cite data-cite='rhodes1970spheroidal'>On the Spheroidal Functions</cite> &#10548;](http://cdm16009.contentdm.oclc.org/cdm/compoundobject/collection/p13011coll6/id/61576/rec/50)
2. [<cite data-cite='tan1986aperture'>Aperture-synthesis mapping and parameter estimation</cite> &#10548;](http://ethos.bl.uk/OrderDetails.do?uin=uk.bl.ethos.384529)：关于最优网格化与去网格化函数的进一步讨论。
