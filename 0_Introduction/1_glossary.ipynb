## 术语表 <a id='preface:sec:glossary'></a>

### A

**azimuth（方位角）:** 在天球地平坐标系中，从正北沿东向量取的角距离。

**altitude（高度角）:** 天体与地平面之间的夹角。

**altitude-azimuth mount（高度-方位式支架）:** 一种让天线分别绕高度轴（竖直）和方位轴（水平）转动，从而跟踪天空目标的天线架设方式。

### B

### C

**celestial equator（天球赤道）:** 地球赤道所在平面向天球延伸后得到的大圆。

**celestial horizon（天球地平圈）:** 观测者的水平面在天球上的投影，是地平坐标系的基本参考平面。

**celestial sphere（天球）:** 一个包围地球的想象单位球面，宇宙中所有天体都可视为投影其上。

### D

**declination（赤纬）:** 天体沿其时圈相对于天球赤道的角距离；北天球为正，南天球为负。

**density weighting function（密度加权函数）:** 见 *weighting, density*。

**direction cosine coordinates（方向余弦坐标）:** 一种以方向余弦 $l$、$m$、$n$ 为基础的天文坐标系，常用于射电干涉测量中描述目标源附近的局部天空图像。在该坐标系下，可见度空间与图像空间之间的傅里叶关系会表现得尤为直观。

**direction-dependent effects（方向依赖效应）:** 随观测方向变化而变化的传播或仪器效应。

**direction-independent effects（方向无关效应）:** 在所有方向上都相同的传播或仪器效应。

**dirty image（脏图像）:** 见 *image, dirty*。

### E

**ecliptic（黄道）:** 太阳在天球上视运动所描出的轨迹。

**elevation（仰角）:** 见 *altitude*。

**equatorial coordinate system（赤道坐标系）:** 天文学中广泛使用的坐标系，其基本平面由地球赤道向天球延伸得到。

**equatorial mount（赤道式支架）:** 一种使天线绕极轴转动来跟踪天体的架设方式，该极轴指向天极。

### F

**field centre（视场中心）:** 见 *phase centre*。

**fringe pattern（条纹图样）:** 一种正弦型函数，用来描述空间域中的源在空间频率域中的响应位置。

**first point of Aries（白羊宫第一点）:** 太阳从天球赤道南侧穿越到北侧时对应的交点，也就是春分点。

**full width at half maximum, FWHM（半高全宽）:** 指函数在峰值一半处两侧位置之间的宽度，常用于描述分布或波束的尺度。

### G

**great circular arc（大圆弧）:** 大圆上的一段弧。

**great circle（大圆）:** 通过球心的平面与球面相交所得的圆。

### H

**horizontal coordinate system（地平坐标系）:** 一种基于观测者本地天空的坐标系统，用于描述天体在本地天空中的位置。

**hour circle（时圈）:** 经过天体和北天极、并与天球赤道垂直的大圆。

**hour angle（时角）:** 天体时圈与本地子午圈之间沿天球赤道向西量取的角距离，通常以小时计。

### I

**image, dirty（脏图像）:** 对干涉观测中采样后的可见度做傅里叶变换所得到的图像，近似等于理想天空图像与阵列 PSF 的卷积结果。它通常是后续反卷积算法的输入图像。

### J

### K

### L

**latitude（纬度）:** 地球表面某地点相对于赤道的南北角距离。

**local meridian（本地子午圈）:** 连接天顶与天球北极所形成的时圈。

**local sidereal time, LST（本地恒星时）:** 春分点的时角。

**longitude（经度）:** 地球表面某地点相对于格林尼治子午线的东西角距离。

**low noise amplifier, LNA（低噪声放大器）:** 模拟前端链路中的第一级放大器，用于放大微弱天体信号，同时尽量只引入较少的系统噪声。

### M

### N

**nadir（天底）:** 天球上与天顶相对的位置。

**natural weighting（自然加权）:** 见 *weighting, natural*。

**north celestial pole, NCP（北天极）:** 地球北极向天球延伸后与天球的交点。

### O

### P

**parallactic angle（视差角）:** 由“源到天顶的大圆”和“源到北天极的大圆”之间形成的球面夹角。

**phase centre（相位中心）:** 阵列相位校准所指向的天空参考位置，也是方向余弦坐标系中的参考中心。

**point spread function, PSF（点扩散函数）:** 测量系统在图像域中对点源产生的响应，本质上可视为可见度采样函数的傅里叶变换。在较早文献中，它也常被称为 *synthesized beam*（综合波束）。

**Polaris（北极星）:** 一颗位于北天极附近的恒星。

**precession（岁差）:** 旋转天体自转轴方向随时间发生的缓慢变化。

**primary beam（主波束）:** 天线增益随方向变化的响应。

### Q

### R

**radiation pattern（辐射方向图）:** 见 **primary beam**。

**radio frequency interference, RFI（射频干扰）:** 人为产生的无线电信号，对目标天体信号造成污染。

**right ascension（赤经）:** 天体时圈与春分点之间沿天球赤道向东量取的角距离。

**robust weighting（稳健加权）:** 见 *weighting, robust*。

### S

**sampling function（采样函数）:** 在信号处理中，把连续信号转换成离散信号时所使用的离散函数。

**sidereal day（恒星日）:** 任意一颗恒星回到天空同一位置所需的时间。

**solar day（太阳日）:** 太阳回到天空同一位置所需的时间。

**spatial domain（空间域）:** 一种信号表示域，其中样本之间的相对距离直接反映信号在物理空间中的相对位置，例如星场图像。

**spatial frequency domain（空间频率域）:** 一种信号表示域，其中样本的振幅和相位用来描述复正弦波的强度与偏移；样本间的相对位置反映它们在频率和方向上的相似程度。

**spherical triangle（球面三角形）:** 由三段大圆弧两两相交形成的三角形。

**south celestial pole, SCP（南天极）:** 地球南极向天球延伸后与天球的交点。

**synthesized beam（综合波束）:** 见 *point spread function (PSF)*。

**system temperature（系统温度）:** 描述望远镜与电子系统噪声水平的量，与系统灵敏度密切相关。

### T

**transit（中天）:** 当天体穿过本地子午圈时所处的状态。

**taper weighting function（加窗加权函数）:** 见 *weighting, taper*。

### U

**uniform weighting（均匀加权）:** 见 *weighting, uniform*。

### V

**vernal equinox（春分点）:** 见 *first point of Aries*。

**visibilities（可见度）:** 干涉基线对空间频率域的离散测量结果。

### W

**weighting, density（密度加权）:** 在合并重叠的可见度样本时，用于设定样本显著性的函数，是合成图像时的重要步骤。

**weighting, natural（自然加权）:** 一种优先提高灵敏度、降低 PSF 旁瓣、但会牺牲部分分辨率的加权方式，也常称为逆方差加权。

**weighting, robust（稳健加权）:** 一种可通过参数调节、在分辨率与灵敏度之间折中的加权方案。

**weighting, taper（加窗加权）:** 一种充当空间滤波器的加权函数，用于在成像时选择特定空间尺度。

**weighting, uniform（均匀加权）:** 一种优先提高分辨率、但通常会带来更高 PSF 旁瓣的加权方式，也常称为单位加权。

### X

### Y

### Z

**zenith（天顶）:** 天球上位于观测者正上方的点。

### 数学符号 <a id='preface:sec:symbols'></a>

* $\lambda$ : 波长（m）
* $\nu$ : 频率（Hz）

* $\mathbf{B}$ : 亮度相干矩阵
* $\boldsymbol{\mathcal{D}}$, $\boldsymbol{\mathcal{M}}$ : 非极化可见度矩阵
* $\mathbf{e}$ : 电场向量
* $\boldsymbol{\mathcal{G}}$ : 非极化增益矩阵
* $\boldsymbol{\mathscr{G}}$ : $\mathbf{g}\mathbf{g}^H$
* $I$ : 天空亮度
* $I,Q,U,V$ : Stokes 参数，不要与天空亮度或可见度混淆
* $\mathbf{J}$ : Jones 矩阵，常见例子包括：
    * $\mathbf{D}$ : 馈源泄漏
    * $\mathbf{E}$ : 主波束
    * $\mathbf{F}$ : 法拉第旋转
    * $\mathbf{G}$ : 增益
    * $\mathbf{K}$ : 几何延迟
    * $\mathbf{P}$ : 视差角
* Jy : Jansky，流量密度单位，$1\,Jy = 10^{-26}\,W\,m^{-2}\,Hz^{-1}$
* $\boldsymbol{\mathcal{M}}$ : 见 $\boldsymbol{\mathcal{D}}$
* $P$ : 功率（W, ergs）
* $\mathbf{S}$ : Stokes 向量，包含 $I,Q,U,V$
* $T$ : 温度（K）
* $\mathbf{v}$ : 电压向量
* $\mathscr{V}$ : 可见度函数，表示底层连续函数而非离散测量值
* $V$ : 可见度测量集合
* $\mathbf{V}$ : 极化可见度

* $\gamma$ - 春分点

* $H$ - 时角
* $\mathcal{A}$ - 方位角
* $\mathcal{E}$ - 仰角
