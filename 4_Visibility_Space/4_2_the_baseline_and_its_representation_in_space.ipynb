# 4.2 基线及其空间表示

***
上一节简要回顾了本章要用到的数学工具。本节则转向干涉测量的几何与物理基础，在讨论可见度与天空亮度之间的关系之前，先把干涉仪中最核心的几何量定义清楚。

首先需要引入“基线”这一概念。基线是干涉阵中任意两面天线之间的分离矢量，是干涉测量中的基本对象；从傅里叶空间的角度看，一条基线对应着一次对某个空间频率点的采样。

一个干涉阵由许多基线构成，因为阵列中的每一对天线都可以形成一条基线。因此，基线的集合完全由阵列布局决定。本节将在此基础上进一步引入“投影基线”，并推导它在不同参考系下的表达式。

下面先统一本节所用的符号和坐标变换路径。我们的目标，是把地面观测者的本地坐标一步步变换到与天空方向相关的参考系中：

1. 从本地 $(x,y,z)$ 坐标变换到 $(E,N,U)$ 坐标。
2. 从本地 $(E,N,U)$ 坐标变换到地平坐标 $(\mathcal{A}, \mathcal{E})$。
3. 从地平坐标 $(\mathcal{A}, \mathcal{E})$ 变换到赤道参考系 $(H,\delta)$，等价地也可写成 $(X,Y,Z)$。
4. 再从赤道参考系 $(X,Y,Z)$ 变换到 $(u,v,w)$ 空间。

<a id="vis:sec:421"></a> <!---\label{vis:sec:421}--->
### 4.2.1 物理基线

***
物理基线是两个天线单元在三维空间中的几何分离矢量。与之对应，投影基线则是这条三维基线沿观测方向投影到二维平面后的结果。在国际地球参考架中，物理基线可以视为常量；但对于固定在天空中的天源而言，投影基线会随着地球自转不断变化。

#### 4.2.1.1 本地基线向量
<a id="vis:sec:4211"></a> <!---\label{vis:sec:4211}--->

***
三维空间中的位置可以用向量来描述。定义原点 $O$、一组正交基 $\mathcal{B}$，以及一个固定在地面的本地笛卡尔参考系 $\mathcal{R}$，其基向量为 $\mathbf{\hat{e}_x}, \mathbf{\hat{e}_y}, \mathbf{\hat{e}_z}$（见 [图 4.2.1 &#10549;](#vis:fig:421)）。

设在这个参考系中有两面天线，它们的位置分别为

$$\textbf{a}_1=\vec{OA}_1=x_1 \mathbf{\hat{e}_x} + y_1 \mathbf{\hat{e}_y} + z_1 \mathbf{\hat{e}_z}$$

$$\textbf{a}_2=\vec{OA}_2=x_2\mathbf{\hat{e}_x} + y_2 \mathbf{\hat{e}_y} + z_2 \mathbf{\hat{e}_z}$$

也就是说，两面天线的位置由 $(x_1,y_1,z_1)$ 与 $(x_2,y_2,z_2)$ 完全确定。

若两天线之间的距离足够短，以至于可以忽略两者之间地球曲率的影响，那么局部地面就可近似看成平面。本地笛卡尔坐标下的物理基线于是定义为两位置向量之差：

$$\mathbf{b} = \mathbf{a}_2 - \mathbf{a}_1 = \vec{A_1A_2} $$

要注意，物理基线是一个矢量，它只依赖于两面天线的相对位置，而与参考系原点的具体选取无关。

<img src='figures/vectorxyz.svg' width=30%>

**图 4.2.1**：本地三维笛卡尔坐标系中的向量 $a_1$ 与 $a_2$

<a id="vis:sec:4212"></a> <!--\label{vis:sec:4212}--></a>
#### 4.2.1.2 $(E,N,U)$ 参考架：基线向量的全局定义

***
在前面的本地表达式中，$\mathbf{\hat{e}_x}, \mathbf{\hat{e}_y}, \mathbf{\hat{e}_z}$ 的方向仍然是任意指定的。为了让基线定义与地球表面的绝对方向联系起来，我们引入地球参考架，也就是由地理方位定义的新笛卡尔基：

* 选取地球表面某一点作为原点 $O$。这个点可以任意选，但通常会取阵列中心或某一面天线的位置。
* 定义一个全局右手正交基 $\mathcal{B}(\mathbf{\hat{e}_E}, \mathbf{\hat{e}_N}, \mathbf{\hat{e}_U})$，分别指向地理东、地理北以及当地天顶方向。

这样，基线向量虽然仍是在本地参考架中描述，但其坐标轴已经与地理上的东、北和天顶对齐（见 [图 4.2.2 &#10549;](#vis:fig:422)）。

#### 4.2.1.2 地平坐标系 $(\mathcal{A}, \mathcal{E})$ 中的基线表示

正如 [$\S$ 3.4 &#10142;](../3_Positional_Astronomy/3_3_horizontal_coordinates.ipynb) 所述，地面观测者通常用方位角 $\mathcal{A}$ 和高度角 $\mathcal{E}$ 来描述天空方向。类似地，基线方向也可以在这个地平参考系中加以描述。

设天线 1 为地平参考系的原点，North、East 与 Up 三个方向按照 [$\S$ 4.2.1.2 &#10549;](#vis:sec:4212) 来定义。

在这个参考系中，方位角是地平面内从北向东顺时针测得的角，高度角则是从地平线向当地天顶测得的角。

<a id="vis:fig:422"></a> <!--\label{vis:fig:422}--></a>
<img src='figures/enu_alt_az.svg' width=40%>

**图 4.2.2**：地平坐标系 $(\mathcal{A}, \mathcal{E})$ 与笛卡尔坐标系 $(E,N,U)$ 的关系

以基 $\mathcal{B}'$ 表示时，基线向量可写为：

\begin{equation}
\mathbf{b}_{\text{ENU}}
=
\lvert \mathbf{b} \rvert
\begin{bmatrix}
\sin \mathcal{A} \cos \mathcal{E}\\
\cos \mathcal{A} \cos \mathcal{E}\\
\sin \mathcal{E}
\end{bmatrix}
\end{equation}

<div class=advice>
<b>注意：</b>如果天线 1 和天线 2 在 $\mathbf{\hat{e}_U}$ 方向上的分量相同，也就是位于同一水平面上，那么这条基线的高度角就是 $\mathcal{E}=0^\circ$。
</div>

<div class=warn>
<b>提示：</b><br>
<ul>
<li>地理学与天文学对方位角原点和增大方向的约定并不完全相同，这里采用的是天文学中的约定，见 [$\S$ 3 &#10142;](../3_Positional_Astronomy/3_0_introduction.ipynb)。</li>
<li>地平坐标系与二维球坐标系非常相似，但在角度约定上略有差别。</li>
<li>这里暂未显式写入纬度 $L_a$，但它在后面会变得十分重要。</li>
</ul>
</div>

#### 4.2.1.3 赤道坐标架 $(X,Y,Z)$ 中的基线表示

现在我们已经用方位角 $\mathcal{A}$ 和高度角 $\mathcal{E}$ 描述了 $ENU$ 参考系中的基线。按照 [$\S$ 3.1 &#10142;](../3_Positional_Astronomy/3_1_equatorial_coordinates.ipynb) 的定义，对于位于纬度 $L_a$ 的观测者，沿着基线方向延拓到天空后，可以用赤道坐标 $(H,\delta)$ 来描述该方向，其中 $H$ 为时角，$\delta$ 为赤纬。

为了便于后续推导，我们进一步引入一个中间参考架 $XYZ$，其基向量为 $(\mathbf{\hat{e}_X},\mathbf{\hat{e}_Y},\mathbf{\hat{e}_Z})$。该参考架与地球相连，并相对于赤道坐标作如下定义：

* $X$ 轴指向 $(H=0^\textrm{h}, \delta = 0^{\circ})$，也就是春分点穿过当地子午线时的方向。
* $Y$ 轴指向 $(H=-6^\textrm{h}, \delta = 0^{\circ})$，即向东的方向。
* $Z$ 轴指向北天极（NCP）。

[图 4.2.3 &#10549;](#vis:fig:413) 表明，阵列所在的局部平面与天赤道平面之间可以通过坐标变换联系起来。

<a id="vis:fig:413"></a> <!--\label{vis:fig:413}--></a>
<img src='figures/conversion.svg' width=30%>

**图 4.2.3**：$(X,Y,Z)$ 参考架。$T$ 为当地天顶，$(X,Y)$ 为与北天极垂直的天赤道平面，$(N,S,E,W)$ 所在平面为天线局部平面

<div class=warn>
<b>提示：</b>
<ul>
<li>虽然这里用赤道坐标来定义 $XYZ$ 坐标系，但它本质上仍是一个与地球相连的参考架，因为对地面局部观测者而言，$X$、$Y$、$Z$ 三轴不会自行旋转。</li>
<li>不过，由于赤纬是相对于天赤道定义的，因此在实际应用中仍必须知道观测者的纬度 $L_a$。</li>
</ul>
</div>

于是，可把 $(\mathcal{A}, \mathcal{E}, L_a)$ 转换到 $(H, \delta)$，写成

\begin{equation}
\begin{bmatrix}
X\\
Y\\
Z
\end{bmatrix}
=
\begin{bmatrix}
\lvert \mathbf{b} \rvert \cos \delta \cos H\\
-\lvert \mathbf{b} \rvert \cos \delta \sin H\\
\lvert \mathbf{b} \rvert \sin \delta
\end{bmatrix}
= \lvert \mathbf{b} \rvert
\begin{bmatrix}
\cos L_a \sin \mathcal{E} - \sin L_a \cos \mathcal{E} \cos \mathcal{A}\nonumber\\ 
\cos \mathcal{E} \sin \mathcal{A} \nonumber\\
\sin L_a \sin \mathcal{E} + \cos L_a \cos \mathcal{E} \cos \mathcal{A}\\
\end{bmatrix}
\end{equation}
<a id="vis:eq:3_1"></a> <!---<a id="vis:eq:3_1"></a>--->

**公式 3.1**：把基线矢量从 $(\mathcal{E},\mathcal{A})$ 变换到 $XYZ$。其中 $\mathbf{b}$ 为基线长度，$H$ 为时角，$\delta$ 为赤纬，$L_a$ 为阵列所在纬度。

<div class=advice>
<b>说明：</b>[公式 3.1 &#10142;](#vis:eq:3_1) 从球面三角形 <b>ZTS</b> 出发的完整推导，可见 [$\S$ A.1 &#10142;](../0_Introduction/2_Appendix.ipynb)。
</div>

### 4.2.2 $(u,v,w)$ 空间

现在我们已经在地球参考系 $XYZ$ 中定义了基线，可以进一步把它转换到与天空方向直接相关的 $uvw$ 参考系中。设 $(H_0, \delta_0)$ 是天源 $\mathbf{s_0}$ 在天球上的位置，则定义新坐标轴如下：

* $u$ 轴位于天赤道平面内，并指向时角 $H_0-6^\text{h}$ 的方向。
* $v$ 轴位于时角 $H_0$ 所在的大圆平面内，并指向赤纬 $\frac{\pi}{2}-\delta_0$ 的方向。
* $w$ 轴指向源 $\mathbf{s_0}$ 本身。

因此，$(u,v)$ 张成的平面与源方向 $\mathbf{s_0}$ 垂直。

把基线从 $(X,Y,Z)$ 坐标变换到 $(u,v,w)$ 坐标，需要对其赤道坐标连续做两次旋转：

1. $\mathcal{R}_\text{Z}^{\frac{\pi}{2}-H_0}$：绕 $Z=Z'$ 轴旋转角度 $\frac{\pi}{2}-H_0$，得到新的 $X'$、$Y'$ 轴（见 [图 4.2.4 &#10549;](#vis:fig:414) 中绿色部分）。
2. $\mathcal{R}_\text{X'}^{\frac{\pi}{2}-\delta_0}$：再绕 $X'=u$ 轴旋转角度 $\frac{\pi}{2}-\delta_0$，得到 $u$、$v$ 轴（见图中的红色部分）。

<a id="vis:fig:414"></a> <!--\label{vis:fig:414}--></a>
<img src='figures/coordinates2b.svg' width=50%>

**图 4.2.4**：$(X,Y,Z)$ 坐标系与 $(u,v,w)$ 坐标系之间的关系

由 $XYZ$ 到 $uvw$ 的变换可写成

\begin{equation}
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\mathcal{R}_\text{X'}^{\frac{\pi}{2}-\delta_0}
\mathcal{R}_\text{Z}^{\frac{\pi}{2}-H_0}
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
\end{equation}

where
\begin{equation}
\mathcal{R}_\text{X'}^{\frac{\pi}{2}-\delta_0}
=
\begin{pmatrix}
1& 0& 0\\
0 & \cos (\frac{\pi}{2}- \delta_0) & \sin (\frac{\pi}{2}- \delta_0)  \\
0 &-\sin (\frac{\pi}{2}- \delta_0) & \cos (\frac{\pi}{2}- \delta_0) \\
\end{pmatrix} 
=
\begin{pmatrix}
1& 0& 0\\
0 & \sin \delta_0 & \cos \delta_0  \\
0 &-\cos \delta_0 & \sin \delta_0 \\
\end{pmatrix} 
\end{equation}

and \begin{equation}
\mathcal{R}_\text{Z}^{\frac{\pi}{2}-H_0}
=
\begin{pmatrix}
\cos (\frac{\pi}{2}- H_0) & \sin (\frac{\pi}{2}- H_0) & 0 \\
-\sin (\frac{\pi}{2}- H_0) & \cos (\frac{\pi}{2}- H_0) & 0 \\
0 & 0 & 1
\end{pmatrix} 
=
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0 \\
-\cos H_0 & \sin H_0 & 0 \\
0 & 0 & 1
\end{pmatrix} 
\end{equation}

<div class=warn>
<b>注意：</b>与线性代数中的通常规则一致，向量变换的作用顺序应从右往左理解。
</div>

继续化简后，得到如下转换矩阵：

\begin{equation}
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\ 
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0\\
\end{pmatrix} 
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
\end{equation}

<div class=warn>
<b>注意：</b>通常我们把 $(u,v,w)$ 写成“以波长为单位”的量，也就是把基线矢量除以观测波长 $\lambda$：
</div>

\begin{equation}
\mathbf{b_\lambda}
= 
\frac{\mathbf{b}}{\lambda}
= 
\begin{pmatrix}
u\\
v\\
w
\end{pmatrix}
=
\frac{1}{\lambda}
\begin{pmatrix}
\sin H_0 & \cos H_0 & 0\\ 
-\sin \delta_0 \cos H_0 & \sin\delta_0\sin H_0 & \cos\delta_0\\
\cos \delta_0 \cos H_0 & -\cos\delta_0\sin H_0 & \sin\delta_0\\
\end{pmatrix} 
\begin{pmatrix}
X\\
Y\\
Z
\end{pmatrix}
\end{equation}

在后续章节中，我们通常会省略下标 $\lambda$，默认 $u$、$v$、$w$ 的单位就是波长。
