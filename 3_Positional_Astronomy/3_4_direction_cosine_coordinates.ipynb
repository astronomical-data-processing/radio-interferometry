## 3.4 方向余弦、相位中心与局部成像坐标<a id='pos:sec:dircos'></a>

赤道坐标适合描述全天球上的目录位置，但在干涉成像中，我们通常不会围绕春分点来建立局部天图，而是围绕*某个选定的中心方向*来展开视场。这个中心方向就是*相位中心*（phase center）或视场中心。方向余弦坐标 $(l,m,n)$ 的作用，正是把天球上的任意方向改写成“相对于相位中心的局部坐标”，从而为可见度方程提供最自然的像面变量。


### 方向余弦的基本定义

设天源方向向量为 $\mathbf{s}$，相对于一组三维直角基底的夹角分别为 $\alpha, \beta, \gamma$，则方向余弦定义为：

$$
 l = \cos \alpha, \qquad m = \cos \beta, \qquad n = \cos \gamma.
$$

若 $\mathbf{s}$ 是单位向量，则自然有
$$
 l^2 + m^2 + n^2 = 1.
$$
因此 $(l,m,n)$ 既可以被看成方向余弦，也可以被看成单位球面上的笛卡尔坐标。


<a id='pos:fig:cosines'></a><img src='figures/cosine.png' width='34%'>

*图 3.4.1*：方向余弦的定义。若方向向量是单位长度，则 $(l,m,n)$ 同时也是其在三维直角坐标轴上的分量。


### 围绕相位中心建立局部坐标系

在干涉测量里，我们并不直接把 $(l,m,n)$ 建立在某个固定全球坐标轴上，而是让 $n$ 轴指向相位中心 $\mathbf{s}_0$。这样一来，视场中心自然对应 $(l,m,n) = (0,0,1)$；其余天源则以“相对于中心方向的偏移”来表示。几何上，这相当于把整个天球坐标系重新平移到一个局部切平面附近。


<a id='pos:fig:convert_lmn_ra_dec'></a><img src='figures/conversion2.png' width='42%'>

*图 3.4.2*：以相位中心 $\mathbf{s}_0$ 为参考方向建立方向余弦坐标。此时 $n$ 轴指向相位中心，$l$ 与 $m$ 则描述视场内相对于中心的局部偏移。


### 与赤道坐标之间的精确关系

若相位中心的赤道坐标为 $(\alpha_0, \delta_0)$，任意天源的赤道坐标为 $(\alpha, \delta)$，记 $\Delta\alpha = \alpha - \alpha_0$。可以在相位中心处建立三条局部基向量：$l$ 轴指向赤经增加方向，$m$ 轴指向赤纬增加方向，$n$ 轴指向相位中心本身。把源方向单位向量分别投影到这三条基向量上，就得到

$$
 l = \cos\delta\,\sin(\Delta\alpha),
$$

$$
 m = \sin\delta\,\cos\delta_0 - \cos\delta\,\sin\delta_0\,\cos(\Delta\alpha),
$$

$$
 n = \sin\delta\,\sin\delta_0 + \cos\delta\,\cos\delta_0\,\cos(\Delta\alpha).
$$

这里的 $n$ 实际上就是源方向与相位中心方向的点乘，因此也等于两者夹角的余弦。对可见天球上的源而言，通常取正号分支

$$
 n = +\sqrt{1-l^2-m^2}.
$$

这说明 $(l,m)$ 已经足以描述局部像面位置，而 $n$ 则记录该方向偏离切平面的三维几何信息。


<div class='advice'>
<b>小视场近似：</b> 当源离相位中心很近、角偏移足够小时，若 $\Delta\alpha$ 与 $\Delta\delta$ 都用弧度表示，可近似写成
$$
 l \approx \Delta\alpha\cos\delta_0, \qquad m \approx \Delta\delta, \qquad n \approx 1 - \tfrac{1}{2}(l^2+m^2).
$$
因此在小视场下，$l$ 和 $m$ 几乎就是图像平面上的两个角坐标；但在宽视场下，$n$ 不再近似等于 1，非共面效应与 $w$ 项就不能忽略。
</div>


<a id='pos:fig:understand_lm'></a><img src='figures/und_lm.png' width='44%'>

*图 3.4.3*：把天球方向向量投影到 $lm$ 平面后，$\sqrt{l^2+m^2}$ 近似表示与相位中心的角距离。这一近似只在小视场条件下成立。


### 为什么 $(l,m,n)$ 对干涉成像如此重要

方向余弦坐标把相位中心直接写进了天空坐标定义。相位中心对应 $(0,0,1)$，视场内其他方向则以相对于中心的偏移表示；这使得相关器、相位旋转和成像网格都能围绕同一个参考方向工作。在小视场近似下，$l$ 和 $m$ 几乎就是图像平面上的横纵角坐标，因此天图可以像普通二维图像一样处理。

更重要的是，$(l,m,n)$ 与下一章的 $(u,v,w)$ 直接配对。基线投影写成 $(u,v,w)$ 后，可见度积分中的相位项就是 $ul+vm+w(n-1)$。换句话说，方向余弦不是额外发明的一套坐标，而是让天空方向与基线投影在同一几何语言中相遇。


### 示例：把相位中心附近的几个源转换到 $(l,m,n)$

下面构造一个小视场示例。我们选定相位中心 $(\alpha_0, \delta_0) = (5^\mathrm h 30^\mathrm m, 60^\circ)$，并在其附近放置三个较弱天源。通过这个例子，可以直接看到 $(l,m)$ 如何描述局部角偏移，以及 $n$ 如何记录偏离切平面的程度。


```python
phase_center = ('phase center', (5, 30,  0.0), (+1, 60,  0,  0.0), np.nan)
nearby_sources = [
    ('source A', (5, 32,  0.4), (+1, 60, 17, 57.0), 1.0),
    ('source B', (5, 36, 12.8), (+1, 61, 12,  6.9), 0.5),
    ('source C', (5, 40, 45.5), (+1, 61, 56, 34.0), 0.2),
]

sources = [phase_center] + nearby_sources


def hms_to_rad(h, m, s):
    return np.deg2rad(15.0 * (h + m / 60.0 + s / 3600.0))


def dms_to_rad(sign, d, m, s):
    deg = sign * (abs(d) + m / 60.0 + s / 3600.0)
    return np.deg2rad(deg)


names = []
flux_jy = []
ra_rad = []
dec_rad = []
for name, ra_hms, dec_dms, flux in sources:
    names.append(name)
    flux_jy.append(flux)
    ra_rad.append(hms_to_rad(*ra_hms))
    dec_rad.append(dms_to_rad(*dec_dms))

flux_jy = np.array(flux_jy)
ra_rad = np.array(ra_rad)
dec_rad = np.array(dec_rad)

alpha0 = ra_rad[0]
delta0 = dec_rad[0]
```

```python
delta_alpha = ra_rad - alpha0
l = np.cos(dec_rad) * np.sin(delta_alpha)
m = np.sin(dec_rad) * np.cos(delta0) - np.cos(dec_rad) * np.sin(delta0) * np.cos(delta_alpha)
n = np.sin(dec_rad) * np.sin(delta0) + np.cos(dec_rad) * np.cos(delta0) * np.cos(delta_alpha)

theta_exact_deg = np.rad2deg(np.arccos(np.clip(n, -1.0, 1.0)))
theta_small_deg = np.rad2deg(np.sqrt(l**2 + m**2))

for name, l_val, m_val, n_val, theta_e, theta_s in zip(names, l, m, n, theta_exact_deg, theta_small_deg):
    print(f'{name:12s}  l = {np.rad2deg(l_val):+7.3f} deg  m = {np.rad2deg(m_val):+7.3f} deg  n = {n_val:0.6f}  exact theta = {theta_e:6.3f} deg  small-angle = {theta_s:6.3f} deg')
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(np.rad2deg(l[0]), np.rad2deg(m[0]), 'bx', ms=10, mew=2, label='phase center')
ax.scatter(np.rad2deg(l[1:]), np.rad2deg(m[1:]), s=120, color='tab:red', label='nearby sources')

for name, x, y, flux in zip(names[1:], np.rad2deg(l[1:]), np.rad2deg(m[1:]), flux_jy[1:]):
    ax.text(x + 0.08, y + 0.08, f'{name} ({flux:.1f} Jy)', fontsize=10)

ax.set_xlabel('$l$ [deg]')
ax.set_ylabel('$m$ [deg]')
ax.set_title('Sources near the phase center in the $(l,m)$ plane')
ax.grid(alpha=0.3)
ax.set_aspect('equal', adjustable='box')
ax.legend(loc='upper left')
plt.show()
```

从输出中可以看到，相位中心严格位于 $(0,0,1)$，而其余天源都表现为相对于中心方向的小偏移。对这些几度以内的偏移，小角近似 $\sqrt{l^2+m^2} \approx \theta$ 已经很好；但如果视场继续增大，$n$ 与 1 的偏离就会变得明显，这正是宽视场成像里 $w$ 项必须被认真处理的原因。


#### 代码演示：小视场近似在多大视场下开始失效

在小视场条件下，我们常把 $\sqrt{l^2+m^2}$ 直接看成角距离 $\theta$，并把 $n$ 近似写成 1。下面用一个最简单的径向偏移例子，看看这两个近似会随着视场半径增大而怎样偏离精确球面结果。


```python
theta_deg = np.linspace(0, 35, 400)
theta_rad = np.deg2rad(theta_deg)
projected_radius_deg = np.rad2deg(np.sin(theta_rad))
n_exact = np.cos(theta_rad)
flat_error_arcmin = 60.0 * (theta_deg - projected_radius_deg)

fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))
axes[0].plot(theta_deg, theta_deg, lw=2, label=r'flat-sky approximation: $\theta$')
axes[0].plot(theta_deg, projected_radius_deg, lw=2, ls='--', label=r'exact projection: $\sin\theta$')
axes[0].set_xlabel(r'True angular distance $\theta$ [deg]')
axes[0].set_ylabel(r'Radial coordinate in the $(l,m)$ plane [deg]')
axes[0].set_title('Angular distance vs. projected radius')
axes[0].grid(alpha=0.3)
axes[0].legend()

axes[1].plot(theta_deg, flat_error_arcmin, lw=2, color='tab:red', label=r'$\theta-\sin\theta$')
axes[1].plot(theta_deg, 1 - n_exact, lw=2, color='tab:purple', label=r'$1-n$')
axes[1].set_xlabel(r'True angular distance $\theta$ [deg]')
axes[1].set_ylabel('Error scale')
axes[1].set_title('Growth of the small-field approximation error')
axes[1].grid(alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```

这张图回答了一个实践问题：为什么几度视场内常常可以放心使用平面近似，而到了十几度、几十度以后就必须认真对待球面几何与 $w$ 项。误差并不是突然出现的，而是随着视场半径平滑增长；只是在窄视场里它足够小，通常可以忽略。


### 通向第 4 章的桥梁

在下一章里，可见度会自然写成相位中心坐标下的积分形式：

$$
V(u,v,w) = \iint \frac{I(l,m)}{n}
\exp\left[-2\pi i\left(ul + vm + w(n-1)\right)\right] \, dl\,dm.
$$

这里暂时不必急于推导每一项，只需要抓住核心关系：$(u,v,w)$ 是基线相对于相位中心的投影坐标，$(l,m,n)$ 是天球亮度相对于相位中心的方向坐标。一旦这对几何变量建立起来，天球、阵列和可见度之间的联系就会非常清楚。也正因如此，方向余弦坐标是整个射电干涉成像语言中最关键的一步过渡。


#### 本节要点

* 方向余弦坐标围绕相位中心建立，是局部天图和干涉成像中最自然的天空坐标；
* $(l,m)$ 在小视场下几乎就是图像平面角坐标，而 $n$ 则记录了偏离切平面的几何信息；
* $n \approx 1$ 只在小视场中成立；宽视场下，$w(n-1)$ 项不能随意忽略；
* 到这里为止，本章已经完成了从目录坐标到局部成像坐标的全部铺垫，可以进入第 4 章的可见度几何。
