## 9.x 延伸阅读与后续实践方向

第 9 章把数据检查、基础校准、成像、自校准、图像测量、平均展宽、谱线、宽带宽场、偏振、短间距、方向依赖处理和端到端连续谱案例串成一条实践链。进一步学习时，应把软件命令和测量方程对应起来：每一步究竟在改变数据、模型、权重、flag 还是图像解释。

推荐继续阅读 CASA Guides、WSClean 文档、DDFacet/KillMS 相关资料、SoFiA 谱线 source finding 文档、RM-Tools 与 RM synthesis 资料，以及对应望远镜的数据处理 cookbook。阅读这些材料时，重点不是复制命令，而是识别默认假设、适用频段、校准策略和质量诊断方式。


### 后续实践路线

较自然的训练顺序是：先对一个小型连续谱数据集完成 inspection、flag、1GC 和基础成像；随后做一轮保守 phase-only 自校准并评估图像质量；再用 9.14 的端到端案例把流程闭合到科学通量测量；之后进入宽带主波束校正、谱线连续谱扣除和 moment 图；最后处理偏振、RM synthesis、短间距和方向依赖宽场成像。每一轮实践都应保留处理日志、关键诊断图和失败案例，因为真实数据处理能力往往来自对失败模式的识别。
