## 1.5 热辐射

在射电天文学中，连续谱辐射大体可以分为热辐射和非热辐射两大类。这里的“热”并不只是日常语言中的高温，而是指辐射场与物质之间存在明确的温度联系：辐射强度要么由热平衡直接决定，要么至少可以用局部热平衡下的粒子速度分布来描述。后面的同步辐射、脉冲星辐射和其他相干过程，则属于不同意义上的非热辐射。

热辐射之所以必须放在第一章，是因为它给射电观测提供了几种最基本的语言。普朗克谱告诉我们热平衡辐射的基准形状；瑞利-金斯近似解释了为什么射电里常把亮度写成温度；亮温把“每单位立体角的辐射强度”转化成更接近物理直觉的量；自由-自由辐射则把电离气体和射电连续谱联系起来。本节的主线就是从这些概念出发，建立“温度、光深、谱形和物理环境”之间的关系。


### 1.5.1 黑体辐射<a id='science:sec:blackbody_emission'></a>

黑体辐射是热辐射最理想化、也最有用的基准。当辐射与物质充分交换能量并达到热平衡时，辐射场的比强度只由温度决定，而与材料成分、表面形状和发射机制细节无关。在天体物理中，真正严格满足黑体条件的系统并不多；但黑体谱仍然是判断热辐射、定义亮温和理解光学厚极限的出发点。

黑体比强度由普朗克公式给出：

\begin{equation}
B_{\nu}(T)=\frac{2h\nu^{3}}{c^{2}}\frac{1}{\exp(h\nu/kT)-1},
\end{equation}

其中 $h$ 是普朗克常数，$k$ 是玻尔兹曼常数，$c$ 是光速。这个公式同时包含了两个尺度：频率尺度 $\nu$ 和温度尺度 $T$。二者真正比较的对象不是 $\nu$ 本身，而是无量纲量

$$
x={h\nu\over kT}.
$$

当 $x\gg 1$ 时，光子能量远高于热运动能量，指数项主导，普朗克谱进入维恩尾端，

$$
B_{\nu}(T)\simeq {2h\nu^3\over c^2}\exp\left(-{h\nu\over kT}\right).
$$

当 $x\ll 1$ 时，可以展开 $\exp(x)-1\simeq x$，于是得到瑞利-金斯近似：

$$
B_{\nu}(T)\simeq {2kT\nu^2\over c^2}.
$$

这个低频极限对射电天文学极其重要。它说明在许多厘米波和毫米波问题中，热平衡辐射的比强度与温度近似成正比，而频率依赖主要表现为 $\nu^2$。后面把亮度写成亮温、把谱线强度写成 $T_b(v)$、以及把 Jy/beam 转换成 K，实质上都依赖这个近似。


普朗克谱峰值的位置由维恩位移定律给出。若用频率形式表示，峰值频率近似为

$$
\nu_{\max}\simeq 58.8\,T\ {\rm GHz},
$$

其中温度单位为 K。宇宙微波背景的温度约为 $2.725$ K，因此其按 $B_\nu$ 表示的谱峰落在约 $160$ GHz 附近；温度更高的热源，谱峰会向更高频移动。图 1.5.1 同时给出完整普朗克谱和瑞利-金斯近似，可以看出低频端二者重合，而接近谱峰以后瑞利-金斯近似会明显失效。

#### 1.5.1.1 亮温为什么重要

瑞利-金斯近似自然引出亮温。对于任意辐射源，只要已知它在某个方向上的比强度 $I_\nu$，都可以定义

$$
T_b \equiv {c^2\over 2k\nu^2} I_\nu.
$$

这个定义并不要求辐射源真的是黑体。它的含义是：如果这个比强度来自瑞利-金斯极限下的黑体辐射，那么对应的温度是多少？因此，亮温不是简单的“真实气体温度”，而是一种把强度转换成温度单位的表示方式。对于光学厚热源，$T_b$ 可以接近电子温度、尘埃温度或激发温度；对于非热源，$T_b$ 则常常只是一个诊断量，用来显示辐射机制是否可能由普通热过程解释。

亮温与干涉成像关系尤其密切。实际图像常以 Jy/beam 给出，而物理解释常需要 K。若合成波束近似为二维高斯，主轴半高全宽分别为 $\theta_{\rm maj}$ 和 $\theta_{\rm min}$（单位为角秒），则常用换算为

$$
T_b({\rm K})\simeq 1.222\times 10^6\,{S_\nu({\rm Jy\ beam}^{-1})\over
\nu_{\rm GHz}^2\,\theta_{\rm maj}('')\,\theta_{\rm min}('')}.
$$

这个式子提醒我们，亮温不只取决于通量密度，也取决于角分辨率。同样的 Jy/beam，在更小的波束中代表更高的面亮度；而干涉阵改变合成波束时，也会改变图像中 Jy/beam 与 K 之间的转换。这一点会在后续成像和科学解释中反复出现。


#### 1.5.1.2 从黑体到真实天体

黑体辐射是理想模型，真实天体通常只能在有限频段、有限光学厚条件下接近它。例如行星、宇宙微波背景、光学厚 H II 区、部分恒星大气以及尘埃连续谱，都可能在某些频率范围内用黑体或修正黑体语言描述。关键不在于把所有热源都称为黑体，而在于判断辐射是否接近热平衡、光深是否足够大、瑞利-金斯近似是否成立，以及亮温能否代表某个真实物理温度。

这一判断也解释了为什么射电天文学如此重视光深。在光学厚极限下，辐射场会趋向 $B_\nu(T)$，观测到的是源函数或局部温度；在光学薄极限下，观测到的则是沿视线累积的发射率，亮温通常低于真实粒子温度。热辐射的许多观测差异，最终都可以回到这两个极限之间的转换。


### 1.5.2 热轫致辐射<a id='science:sec:bremsstrahlung_emission'></a>

热轫致辐射是射电天文学中另一种极其重要的热连续谱机制，也常称为自由-自由辐射（free-free emission）。它来自自由电子在离子库仑场中偏转和加速时释放的辐射；之所以称为“自由-自由”，是因为电子在相互作用前后仍处于自由态，并没有被俘获到原子能级中。

这类辐射广泛出现在 H II 区、行星状星云、恒星风、超新星遗迹中的热等离子体、星系核区和星暴星系中。与黑体辐射相比，自由-自由辐射更直接反映电离气体的密度、温度和几何分布。它的观测价值也正在于此：只要能够区分自由-自由辐射和同步辐射，就能把电离气体和相对论电子这两类完全不同的物理成分分开。


要理解自由-自由辐射，可以先从一个带电粒子被加速会辐射这一事实开始。在经典电动力学中，带电粒子的辐射功率与加速度平方成正比。用高斯单位制写，拉莫尔公式为

$$
P={2\over 3}{q^2 a^2\over c^3}.
$$

当电子掠过离子时，库仑力会改变电子速度方向，从而产生瞬时加速度。由于电子质量远小于离子，在同样电场作用下电子的加速度大得多，因此天体自由-自由辐射主要由电子在离子电场中的运动贡献。单次相互作用的轨迹会形成连续频谱；在天体等离子体中，观测到的是大量电子、离子和速度分布平均后的总体辐射。

**图 1.5.2**：自由电子在离子库仑场中偏转并产生轫致辐射的示意图。


若电子速度服从麦克斯韦分布，对速度和冲量参数积分后，自由-自由辐射的发射率可以写成近似形式

$$
j_\nu^{\rm ff}\propto n_e n_i Z^2 T_e^{-1/2}
\exp\left(-{h\nu\over kT_e}\right)\,\bar g_{\rm ff},
$$

其中 $n_e$ 和 $n_i$ 分别是电子和离子数密度，$Z$ 是离子电荷数，$T_e$ 是电子温度，$\bar g_{\rm ff}$ 是 Gaunt 因子，用来修正经典近似与量子力学结果之间的差别。在射电频率下通常有 $h\nu\ll kT_e$，指数项接近 1，因此发射率对频率的依赖很弱，而对密度平方的依赖非常强。这就是为什么自由-自由辐射常被用来追踪电离气体的发射量。

相应的吸收系数在射电近似下近似满足

$$
\kappa_\nu^{\rm ff}\propto n_e n_i Z^2 T_e^{-3/2}\nu^{-2}\bar g_{\rm ff}.
$$

更常用的观测量是沿视线积分得到的发射量

$$
EM=\int n_e^2\,dl,
$$

以及由它决定的自由-自由光深。对氢等离子体，一个常用近似为

$$
\tau_\nu^{\rm ff}\simeq 8.24\times 10^{-2}
\left({T_e\over {\rm K}}\right)^{-1.35}
\left({\nu\over {\rm GHz}}\right)^{-2.1}
\left({EM\over {\rm pc\ cm}^{-6}}\right),
$$

其中指数 $-2.1$ 已经把 Gaunt 因子的缓慢频率变化合并进去。这个表达式的物理含义很清楚：低频更容易被自由-自由吸收，高密度或长路径的电离气体更容易变得光学厚，而温度升高会降低吸收。


把自由-自由辐射放进辐射传输方程，会得到一个非常有用的结果。若电离气体近似等温，并处于局部热平衡，源函数就是 $B_\nu(T_e)$，于是

$$
I_\nu=B_\nu(T_e)\left(1-e^{-\tau_\nu}\right).
$$

在瑞利-金斯极限下，这个式子等价于

$$
T_b=T_e\left(1-e^{-\tau_\nu}\right).
$$

这条关系直接连接了观测亮温和电离气体物理状态。若 $\tau_\nu\gg 1$，亮温趋向电子温度，谱形接近光学厚黑体的 $S_\nu\propto\nu^2$；若 $\tau_\nu\ll 1$，则 $T_b\simeq T_e\tau_\nu$，而通量密度近似满足 $S_\nu\propto\nu^{-0.1}$，表现为非常平坦的热连续谱。正是这一厚薄转换，使紧致 H II 区、恒星风和星系中的电离气体可以通过宽频连续谱观测被区分出来。

#### 1.5.2.1 热轫致辐射在观测上怎么识别

自由-自由辐射最典型的环境是电离气体。因此，在年轻大质量恒星周围的 H II 区、行星状星云、恒星风以及强恒星形成星系中，射电连续谱常包含显著的自由-自由成分。它与同步辐射相比有两个重要差别：第一，光学薄自由-自由谱很平坦，而典型同步辐射谱通常随频率升高而明显下降；第二，自由-自由辐射的亮温通常与电离气体温度和光深有关，不会无条件达到极高的非热亮温。

实际观测中，热与非热成分往往混合在一起。低频可能受到自由-自由吸收影响，高频可能受到尘埃连续谱影响，中间厘米波段则常同时包含自由-自由和平滑的同步辐射背景。因此，区分辐射机制不能只看一张图像，而要结合谱指数、亮温、偏振、空间位置和多波段环境。自由-自由辐射通常无显著线偏振，并与 H$\alpha$、红外尘埃加热和复合线等电离气体示踪量相关；同步辐射则更直接追踪相对论电子和磁场。


H II 区是自由-自由辐射最经典的天体环境之一。年轻高温恒星发出的紫外光电离周围氢气，形成温度约 $10^4$ K 的电离气体云。由于射电自由-自由辐射不易受到尘埃消光影响，它能够揭示光学波段被遮蔽的电离区，并与分子云、尘埃连续谱和谱线数据一起构成恒星形成研究的重要观测基础。

像猎户座星云这样的区域，既是热连续谱源，也是谱线和分子云研究的重要目标。一片电离气体云不会只在一条谱线上留下痕迹，它往往同时具有热连续谱、复合线、分子云边界和几何结构信息。这也是为什么射电观测常常需要连续谱与谱线结合分析：连续谱给出电离气体和能量输入，谱线给出速度、化学和密度结构，二者合在一起才更接近完整的物理图像。


#### 1.5.2.2 本节小结

本节建立了热辐射在射电天文学中的基本语言。黑体辐射是热平衡辐射场的基准模型，瑞利-金斯近似使比强度可以被写成亮温，亮温又成为连接图像单位和物理解释的桥梁。自由-自由辐射则说明，电离气体即使不是完美黑体，也会通过热电子与离子的相互作用产生可观测的射电连续谱。

从观测角度看，热连续谱的核心判断是光深、温度和谱形。光学厚时，亮温趋向物理温度；光学薄时，亮温反映沿视线累积的发射量。下一节将转向非热连续谱中最核心的一类：同步辐射。它与本节形成鲜明对照，也是许多典型射电源在图像上最亮的原因。
