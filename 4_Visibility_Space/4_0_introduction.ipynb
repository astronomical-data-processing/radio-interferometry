# 第 4 章：可见度空间<a id='visibility:sec:intro'></a>


第 3 章把天空方向、台站指向和相位中心坐标串成了一条几何链。本章进入射电干涉测量的核心：干涉仪到底测量什么。单天线直接接收某个方向范围内的功率；干涉阵则把不同天线接收到的电压信号相乘并做时间平均。这个互相关结果不是图像中的一个像素，而是天空亮度分布的一个空间频率样本，称为*可见度*（visibility）。

这个观点是整个孔径合成的转折点。干涉阵不是先拍下一张模糊照片再做锐化，而是在 `uv` 平面上逐点采样天空亮度的傅里叶信息。每一条基线、每一个时刻、每一个频率频道都会给出一个不同的采样点；许多采样点共同构成 `uv` 覆盖。若覆盖足够密，并且小视场、远场、窄带等近似成立，就可以把这些可见度反变换为图像。若覆盖稀疏，反变换得到的就是带有脏波束卷积效应的脏图像。

![可见度测量链条](figures/chapter4_visibility_pipeline.png)

**图 4.0.1** 从天空亮度到可见度采样的测量链条。相关器把两台天线的电压信号转换为某条投影基线上的复可见度；许多基线、时间和频率样本共同形成可见度空间中的采样。

本章主要回答三个问题。第一，一条基线的互相关为什么对应天空亮度的一个空间频率分量；第二，基线在地球自转中如何投影成 `u,v,w` 坐标，并在 `uv` 平面上画出轨迹；第三，在什么近似下 van Cittert-Zernike 定理允许我们把可见度函数看成天空亮度的二维傅里叶变换。这三个问题连在一起，就构成了从物理接收机到成像方程的主线。

第 4.1 节先短暂回到周期函数与傅里叶分析，目的是把“空间频率”和“图像结构”之间的直觉重新唤起。第 4.2 节定义物理基线以及它在不同参考系中的表示，并把它引向 `u,v,w` 坐标。第 4.3 节从最简单的二元干涉仪出发，说明相关器如何产生复测量量。第 4.4 节把这个测量量提升为连续的可见度函数。第 4.5 节讨论地球自转、天线布局和频率综合如何改变 `uv` 覆盖。第 4.6 节给出傅里叶近似与 van Cittert-Zernike 定理，明确哪些条件让“可见度是天空的傅里叶变换”这一说法成立。第 4.7 节继续讨论闭合量、缺短间距和 mosaicking，把可见度空间的采样约束连接到真实质量判断。
