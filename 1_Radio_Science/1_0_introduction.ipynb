# 第1章：利用干涉阵开展射电科学<a id='science:sec:intro'></a>

射电干涉测量常给初学者留下两种截然相反的印象。第一种印象是它极其强大，可以用分布在数公里、数百公里甚至地球尺度上的阵列去研究喷流、星系气体、分子云和黑洞附近的精细结构；第二种印象则是它过于复杂，仿佛必须先掌握一整套只属于“圈内人”的黑箱技巧，才有可能把数据真正处理成科学结果。本书希望打破的，正是后一种印象。

从本质上说，射电干涉测量并不是脱离物理直觉的技巧堆砌，而是一条可以逐层理解的科学链条：从天体辐射机制出发，到电场与可见度的测量，再到成像、去卷积、观测系统、定标和最终的物理解释。只要这些环节之间的关系讲清楚，所谓“干涉测量的复杂性”就会从神秘感转化为可分析、可训练、可复现的方法体系。


本章承担的任务，是在正式进入数学、几何与可见度空间之前，先把整条链条的科学动机建立起来。它不会直接给出完整的干涉测量理论推导，而是先回答几个更根本的问题：射电波段为什么值得研究？射电天体常见的辐射机制有哪些？为什么我们需要从单碟走向干涉阵？现代干涉阵为什么会长成今天这样？只有先把这些问题想明白，后面关于傅里叶变换、采样、成像和定标的技术章节才不会显得脱离科学背景。


#### 本章学习目标

* 理解射电科学在现代天体物理中的位置，以及射电观测相对于其他波段的独特信息；
* 建立对电磁辐射、比强度、亮温、辐射传输和常见射电辐射机制的基本语言；
* 熟悉典型天文射电源及其对分辨率、频率覆盖和成像能力提出的不同要求；
* 理解为什么单碟望远镜无法独自满足高分辨率射电科学的需求，以及干涉阵如何作为解决方案出现；
* 从现代阵列实例出发，初步建立“科学问题驱动仪器设计”的视角。


#### 本章结构

第 1.1 到 1.8 节先从射电科学本身出发，介绍本章后续所有观测讨论所依赖的物理背景，包括辐射测量量、辐射传输、热与非热连续谱、谱线辐射以及典型天文射电源。第 1.9 到 1.11 节则把重心转向仪器与方法，说明干涉测量的物理直觉、单碟与干涉阵的关系，以及现代射电阵列的主要类型。这样安排的目的是让读者在进入后续正式理论之前，先把“为什么要这样测”“测出来的量代表什么”“不同阵列为什么会有不同设计”这三件事放在同一幅图景里。

读完本章之后，后续全书会按方法主线展开：第 2 章介绍干涉测量所需的数学工具，第 3 章讨论位置天文学与坐标系统，第 4 章进入可见度空间，第 5 至第 6 章讲成像与去卷积，第 7 章讨论观测系统，第 8 章讨论定标，第 9 章再回到实践工作流。换句话说，第 1 章负责建立科学与仪器背景，其余章节负责把这套背景转化为完整的干涉测量方法论。
