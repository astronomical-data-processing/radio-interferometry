## 2.12 补充专题：立体角与天球面积元素<a id='math:sec:sldangle'></a>

立体角是把天文学辐射量写清楚的基础。比强度 $I_\nu$ 的单位中包含 “每单位立体角”，波束面积是一个立体角，天球上对亮度积分得到流量密度时也必须使用 $d\Omega$。因此，立体角不是几何附录里的孤立概念，而是连接图像亮度、波束和总流量的共同尺度。

平面角定义为弧长除以半径：$\theta=l/r$。立体角是这一概念在球面上的推广。半径为 $r$ 的球面上，一块面积 $A$ 对球心张开的立体角定义为

<a id='math:eq:solid_angle_definition'></a><!--\label{math:eq:solid_angle_definition}-->
$$
\Omega=\frac{A}{r^2}.
$$

立体角描述的是方向范围，而不是距离相关的真实面积。同一片天空方向投影到更远的球面上，面积按 $r^2$ 增长，但 $A/r^2$ 不变。

![立体角定义](figures/solid.png)

**图 2.12.1** 立体角的定义。整个球面的立体角为 $4\pi\,\mathrm{sr}$。


常用角度单位之间的关系为

$$
1\,\mathrm{rad}=\frac{180}{\pi}^{\circ}\simeq 57.296^{\circ}
\simeq 206265'',
$$

$$
1^{\circ}=60'=3600'',
\qquad
1\,\mathrm{h}=15^{\circ}.
$$

球面度（steradian, sr）是立体角单位。由于整个球面面积为 $4\pi r^2$，整个天空的立体角为

$$
\Omega_{\rm sphere}=4\pi\,\mathrm{sr}.
$$

数值上，

$$
1\,\mathrm{sr}=\left(\frac{180}{\pi}\right)^2\mathrm{deg}^2
\simeq 3282.8\,\mathrm{deg}^2.
$$


### 2.12.1 球坐标与微分立体角<a id='math:sec:differential_solid_angle'></a>

采用球坐标 $(\rho,\theta,\phi)$，其中 $\theta$ 是从极轴量起的极角，$\phi$ 是方位角。笛卡尔坐标为

<a id='math:eq:spherical_cartesian'></a><!--\label{math:eq:spherical_cartesian}-->
$$
x=\rho\sin\theta\cos\phi,
\qquad
y=\rho\sin\theta\sin\phi,
\qquad
z=\rho\cos\theta.
$$

![球坐标](figures/spherical.png)

**图 2.12.2** 球坐标 $(\rho,\theta,\phi)$。


固定半径 $\rho$ 时，球面上的微小面积元可以由两条正交弧长相乘得到。沿经线方向，弧长为

$$
\rho\,d\theta;
$$

沿纬圈方向，半径变为 $\rho\sin\theta$，弧长为

$$
\rho\sin\theta\,d\phi.
$$

因此球面面积元为

<a id='math:eq:spherical_area_element'></a><!--\label{math:eq:spherical_area_element}-->
$$
dS=\rho^2\sin\theta\,d\theta\,d\phi,
$$

对应的微分立体角为

<a id='math:eq:differential_solid_angle'></a><!--\label{math:eq:differential_solid_angle}-->
$$
d\Omega=\frac{dS}{\rho^2}=\sin\theta\,d\theta\,d\phi.
$$

![微分立体角](figures/spherical_dO.png)

**图 2.12.3** 球面面积元与微分立体角。$\sin\theta$ 因子来自纬圈半径随极角变化。


### 2.12.2 方向余弦与小视场近似<a id='math:sec:solid_angle_lmn'></a>

干涉成像常使用相对相位中心的方向余弦 $(l,m,n)$，其中

$$
n=\sqrt{1-l^2-m^2}.
$$

在这组坐标中，天球立体角元可写为

<a id='math:eq:solid_angle_lmn'></a><!--\label{math:eq:solid_angle_lmn}-->
$$
d\Omega=\frac{dl\,dm}{n}.
$$

当视场足够小，$l,m\ll 1$，于是 $n\approx1$，得到窄视场常用近似

$$
d\Omega\approx dl\,dm.
$$

这个近似解释了为什么小视场图像中的像素面积常可直接当作立体角面积；但宽视场、全天空成图和低频大视场观测必须保留完整球面几何。


### 2.12.3 波束面积、流量与亮度<a id='math:sec:beam_area_flux'></a>

若归一化波束为 $B(\Omega)$，波束立体角定义为

<a id='math:eq:beam_solid_angle'></a><!--\label{math:eq:beam_solid_angle}-->
$$
\Omega_B=\int B(\Omega)\,d\Omega.
$$

若天体比强度为 $I_\nu(\Omega)$，总流量密度为

<a id='math:eq:flux_from_intensity'></a><!--\label{math:eq:flux_from_intensity}-->
$$
S_\nu=\int I_\nu(\Omega)\,d\Omega.
$$

这两个积分说明，图像亮度和总流量之间差了一个立体角积分。对椭圆高斯波束，窄视场下常用近似为

<a id='math:eq:gaussian_beam_solid_angle'></a><!--\label{math:eq:gaussian_beam_solid_angle}-->
$$
\Omega_B\simeq 1.1331\,\theta_{\rm maj}\theta_{\rm min},
$$

其中两个 FWHM 必须使用弧度。图像单位从 Jy/beam 转换到 Jy/sr 或亮温时，这个因子会直接出现。

![小视场近似与波束面积](figures/solid_angle_small_field_beam_area.png)

**图 2.12.4** 小视场面积近似与高斯波束面积近似。几度以内的窄视场近似通常很好；尺度达到几十度时，球面修正不再可以忽略。
