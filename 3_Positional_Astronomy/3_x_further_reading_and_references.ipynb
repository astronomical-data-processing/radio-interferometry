## 3.x 延伸阅读与参考文献


本章的阅读材料可以分成三类：位置天文学与参考系统的背景材料，直接服务于射电干涉几何的核心教材，以及把课堂公式接到现代数据处理流程的软件与标准框架。


### 位置天文学与参考系统

* [Practical Astronomy with your Calculator or Spreadsheet](https://www.cambridge.org/core/books/practical-astronomy-with-your-calculator-or-spreadsheet/coordinate-systems/01F111D3505038668441D3715414C40F)  
  适合系统复习坐标系统、时间系统以及若干基础观测几何概念。它比本章更接近传统位置天文学教材，可作为坐标与时间变量的补充背景。
* [JPL Horizons Manual](https://ssd.jpl.nasa.gov/horizons/manual.html)  
  这份官方手册系统展示参考架、观测者位置、视位置与星历服务之间的关系，也体现了现代天文软件如何严肃处理参考系统问题。


### 射电干涉核心教材

* [Essential Radio Astronomy（NRAO）](https://www.cv.nrao.edu/~sransom/web/Ch3.html)  
  一套公开的射电天文学课程讲义，特别适合把赤道坐标、时角、本地天空和干涉仪几何放进同一条叙事链里理解。
* [Interferometry and Synthesis in Radio Astronomy](https://www.loc.gov/item/2019756474)  
  这是射电干涉领域最经典的教材之一。与本章最直接相关的是其关于几何关系、两单元干涉仪和可见度响应的章节，它会把本章的 $(H,\delta)$ 与 `(l,m,n)` 进一步接到 `u,v,w` 和可见度积分上。
* [Tools of Radio Astronomy](https://link.springer.com/book/10.1007/978-3-642-39950-3)  
  适合把单天线观测、射电辐射基础和干涉测量几何放在更宽广背景下通读。


### 现代软件与实践框架

* [Astropy Coordinates Documentation](https://docs.astropy.org/en/stable/coordinates/index.html)  
  这份文档非常适合把本章公式落实到 Python 工作流中。`SkyCoord`、`ICRS`、`AltAz`、`EarthLocation` 等对象正是现代坐标变换实践中最常见的工具。
* 各阵列的调度与观测准备软件文档  
  CASA、MeerKAT/EVN/VLBI 相关的观测准备工具通常都会在本章的基础上进一步加入阵列工程约束、扫描策略和指向模型。


### 课程与暑校材料

* [NRAO Synthesis Imaging Workshop Lectures](https://science.nrao.edu/science/meetings/2018/16th-synthesis-imaging-workshop/16th-synthesis-imaging-workshop-lectures)  
  这类课程讲义是从本章几何内容走向“地球自转如何填充 `uv` 覆盖”的良好桥梁。
* 各类 VLBI/射电干涉暑校与观测学校  
  这类材料通常最擅长把源目录、调度、相位中心、校准策略和成像流程串起来。阅读时尤其值得留意 hour angle coverage、elevation limits、phase-referencing 和 wide-field imaging 等主题。


#### 阅读建议

比较顺畅的顺序是：先用 `Practical Astronomy` 与 `Astropy` 文档稳固坐标定义和软件对象；再把 `RA/Dec`、`HA/LST` 和 `Alt/Az` 的转换放入 `Interferometry and Synthesis in Radio Astronomy` 或 NRAO 公开课程的干涉几何语境；最后反复回看本章第 3.2 到 3.4 节，因为时角、可观测性和相位中心会在后续章节不断出现。
