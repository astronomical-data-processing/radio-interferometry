## 9.34 处理报告与复查量规：从产品到可审查结论

射电干涉测量的实践训练不能止于“生成了一张图”或“得到了一张源表”。图像、源表、通量表和上限只有在证据链完整时才成为科学结论。处理报告的作用，就是把科学问题、数据身份、处理选择、质量证据、误差预算和限制声明组织成一条可审查的因果链。复查量规的作用，则是判断这条链在哪些环节充分，在哪些环节需要补证据、重跑或降低结论等级。

一份报告可能图很漂亮，但若没有输入数据身份，结果无法复查；参数描述可能很流畅，但若没有配置文件或运行编号，差异无法定位；噪声可能写得很低，但若没有局部噪声、相关噪声和系统误差，科学量的置信度会被高估。报告写作因此不是附属环节，而是数据处理的一部分。能否写出可审查报告，常常比单次软件运行是否成功更能反映实践能力。

本节把处理报告视为前面几节的出口。第 9.30 节定义运行事件，第 9.31 节定义归档复盘证据链，第 9.32 节定义文件级发布包，第 9.33 节定义轻量样本包。第 9.34 节进一步说明这些对象怎样进入报告，怎样被复查，怎样在证据不足时触发修订或降级。


![处理报告证据主线](figures/practical_report_evidence_spine.png)

图 9.34.1 展示了处理报告的证据主线。科学问题定义目标量，数据身份限定频率和尺度，处理选择说明权重和阈值，质量证据检查残差和噪声，科学测量给出通量或上限，限制声明说明适用边界。每一环都应有可追溯证据。


### 9.34.1 报告不是步骤记录，而是因果解释

处理报告常见的薄弱形式，是按时间顺序记录命令：先检查数据，再校准，再成像，再测量。这种顺序有助于回忆工作过程，却不一定能说明结论为什么可信。科学报告更需要因果顺序：目标量是什么；为什么这个数据集能够或不能测量它；为什么这些处理参数适合目标；质量证据是否支持产品；误差预算是否覆盖主要风险；最终结论适用于哪些范围。

因此，报告第一段不应从软件开始，而应从科学目标和可观测量开始。若目标是低面亮度连续谱上限，报告需要立即说明目标区域、角尺度、频率、亮温灵敏度和最大可恢复尺度。若目标是连续谱源表，报告需要说明检测阈值、局部噪声、主波束裁剪、可靠性和完整性。若目标是谱线 moment 图，报告需要说明速度约定、连续谱扣除、mask、通道噪声和 beam smearing 风险。

步骤记录仍然重要，但它应作为证据附录或运行日志进入发布包，而不是取代解释。正文负责回答“为什么可以相信这个结果”，清单和日志负责回答“结果如何被生成”。二者合在一起，报告才可审查。


### 9.34.2 处理报告的基本结构

一份完整的处理报告可以组织为七个部分。科学问题部分定义目标量和判据；数据身份部分记录项目编号、频率、阵列配置、主波束区域、校准源和产品版本；处理选择部分记录权重、taper、mask、阈值、源搜索参数和测量区域；质量证据部分展示校准解、flag 统计、残差、局部噪声和产品差异；科学测量部分给出通量、源表、上限、谱线积分或偏振量；误差预算部分区分随机误差、相关噪声、系统误差和数据限制；限制声明部分说明结论适用范围和不能回答的问题。

这个结构并不要求每个报告很长。小型练习可以每部分一段，研究级报告可以每部分展开为图表和附录。关键是每个部分都能指向文件级证据，例如清单、参数文件、产品 manifest、运行报告或质量证据板。

| 报告部分 | 需要回答的问题 | 证据对象 |
|---|---|---|
| 科学问题 | 目标量和判断标准是什么 | 目标说明、区域文件、预期尺度 |
| 数据身份 | 输入是否足以支持目标量 | 数据清单、质量报告、主波束信息 |
| 处理选择 | 参数为何匹配科学目标 | 配置文件、运行编号、产品差异 |
| 质量证据 | 产品是否被残差或系统误差限制 | 残差图、局部噪声、校准诊断 |
| 科学测量 | 测得了什么，单位和区域是什么 | 通量表、源表、上限计算 |
| 误差预算 | 不确定度由哪些项组成 | 随机项、系统项、数据限制 |
| 限制声明 | 结论不能越过哪些边界 | 复盘报告、发布包、审计记录 |


### 9.34.3 复查量规：把模糊印象定位到证据链

复查量规的目标不是机械打分，而是定位证据链断点。一个报告可以在科学问题和数据身份上充分，却在误差预算上不完整；也可以在参数记录上完整，却在质量证据上不足。量规应把这些情况分开，否则“总体不错”或“还不够清楚”都很难转化为具体改进。


![处理报告复查量规](figures/practical_report_rubric_matrix.png)

图 9.34.2 给出一个量规矩阵示例。科学问题、数据身份、处理参数、质量证据、误差预算和限制声明分别判为充分、部分充分或缺失。矩阵的价值在于指出断点，例如处理参数只写成文字、误差预算只给热噪声、限制声明过宽等。


一个实用量规可以包含六个判断维度。科学问题是否具体到可观测量；数据身份是否足以复查输入；处理参数是否以配置或清单形式保存；质量证据是否覆盖局部噪声、残差和系统结构；误差预算是否包括随机项、相关噪声、通量标尺、主波束和数据限制；限制声明是否把“适合、部分适合、只能给限制、不适合”写清楚。

量规还应允许“部分充分”。例如，一个图像测量报告若没有完整校准表，但保留了归档质量报告和产品身份，可以支持图像级测量，却不能支持校准策略复查。一个源表报告若给出了局部噪声和检测阈值，但没有注入源恢复率，可以形成候选目录，却不宜发布完整性校正后的统计结论。部分充分不是失败，而是结论等级的依据。


### 9.34.4 误差预算：不仅是热噪声

射电图像中的一个常见问题，是把热噪声当作全部误差。对一个点源峰值测量，局部均方根可能接近主要随机误差；但对积分通量、低面亮度上限、源表统计、谱线 moment 或偏振量，误差预算往往更复杂。相关噪声会使像素数不能简单当作独立样本；通量标尺误差会影响所有源的共同归一化；主波束误差随位置改变；模型选择、mask、阈值和背景区域会带来方法误差；缺短间距、频率覆盖不足或校准证据不完整则属于数据限制。

积分通量的不确定度可以写成一个简化形式：

$$
\sigma_S^2 \simeq N_{\rm beam}\sigma_{\rm rms}^2 + (f_{\rm flux}S)^2 + \sigma_{\rm pb}^2 + \sigma_{\rm model}^2,
$$

其中 $N_{\rm beam}$ 是测量区域中的独立波束数，$\sigma_{\rm rms}$ 是局部噪声，$f_{\rm flux}$ 是通量标尺相对误差，$\sigma_{\rm pb}$ 表示主波束相关误差，$\sigma_{\rm model}$ 表示背景、mask 或模型选择带来的不确定度。这个式子不是唯一标准，而是提醒报告应说明哪些误差项被纳入，哪些只能作为限制声明。


![误差预算证据板](figures/practical_report_uncertainty_board.png)

图 9.34.3 把误差预算拆成随机噪声、相关噪声、通量标尺、主波束、模型选择和数据限制。只给热噪声通常不足以支持科学结论；系统项和数据限制同样需要进入报告。


上限报告尤其需要谨慎。一个 $3\sigma$ 上限必须说明 $\sigma$ 的来源、区域大小、独立波束数、单位、主波束响应、最大可恢复尺度和假设源形态。若目标可能比最大可恢复尺度更大，则上限只适用于小于可恢复尺度的结构；若强源残差主导局部噪声，则上限不应使用全图均方根；若主波束响应低，通量上限必须纳入主波束校正后的噪声放大。报告的价值在于把这些条件写清，而不是给出一个孤立数字。


### 9.34.5 复查循环：补证据、重跑或降级

复查不应只是末尾盖章。若复查发现输入身份不完整，报告应补数据清单；若发现参数不可追溯，应补配置文件或运行编号；若发现产品差异无法解释，应回到成像或源搜索参数；若发现误差预算不足，应补局部噪声、系统项或数据限制；若关键证据无法补齐，则应降低结论等级。


![处理报告复查循环](figures/practical_report_review_loop.png)

图 9.34.4 展示了自查、交叉复查、修订和发布之间的循环。复查应能触发重跑、补证据或降低结论等级。问题未解时，降级结论比维持过宽结论更稳健。


这个循环与软件工作流直接相连。报告中的每个改动都应能回到文件级发布包：参数改变对应新的配置文件和运行编号，产品改变对应新的 manifest 和校验和，结论改变对应新的限制声明。若复查意见只停留在文字层面，而没有进入参数、产品或清单，报告与数据处理仍然是脱节的。


### 9.34.6 常见失败模式与修复

处理报告的失败模式通常对应证据链断点，而不只是格式问题。只展示最终图，说明输入身份和处理路径缺失；参数写成自然语言，说明配置不可复查；噪声只给全局值，说明局部质量证据不足；差异图只说“更好”，说明产品差异没有回到参数和科学目标；上限无角尺度，说明数据限制没有进入结论；源表没有选择函数，说明目录统计不能发布；限制声明过宽，说明结论越过了证据边界。


![处理报告常见失败模式](figures/practical_report_failure_modes.png)

图 9.34.5 把常见失败模式映射到对应修复：补清单、补参数、补验证或降级结论。失败模式不是格式问题，而是科学结论无法被重新审查的信号。


修复策略应尽量具体。缺输入身份时，补数据 manifest 和产品版本；参数不可复查时，补配置文件和运行编号；噪声证据不足时，补局部噪声图和残差区域说明；差异无解释时，补共同 beam 比较、参数差异表和科学区域测量；上限过宽时，补最大可恢复尺度、主波束范围和假设源形态；源表统计不足时，补注入源恢复、边缘区域排除和人工检查样本。无法修复的部分应进入限制声明，而不是被沉默处理。


### 9.34.7 教学案例：归档连续谱复盘报告

延续第 9.31 到第 9.33 节的连续谱归档复盘案例，一份合格报告可以这样组织。科学问题部分说明目标是给出星系团外围弥散连续谱的表面亮度上限，而不是寻找所有点源。数据身份部分说明归档项目、频率、阵列配置、主波束范围、最大可恢复尺度和质量报告证据。处理选择部分说明为什么从原 pipeline 图像转向自然权重和 taper 重跑，并给出参数文件。质量证据部分比较原图、taper 图、残差图和局部噪声。测量部分给出区域通量或上限，并把单位、beam 和主波束裁剪写清。误差预算部分包括局部噪声、相关波束数、通量标尺、主波束和强源残差。限制声明部分说明该上限只适用于某一角尺度和主波束范围，不能排除更大尺度辐射。

复查量规可以指出两个常见问题。若报告只给出 taper 后的非探测，但没有最大可恢复尺度，限制声明不充分；若报告使用全图均方根计算上限，但科学区域靠近强源残差，误差预算不充分。修复后，报告可以从“未探测到弥散辐射”改为更准确的表述：在某一频率、beam、主波束范围和可恢复尺度内，未检测到超过某一表面亮度阈值的结构；更大尺度或更低面亮度成分仍需新观测或短间距补充。

这个案例的重点不是把结论写得保守，而是让结论和证据匹配。可审查报告不惧怕限制，相反，清楚的限制正是科学判断成熟的表现。


### 9.34.8 本节结论

处理报告是实践链的出口，不是附录装饰。它需要把科学问题、数据身份、处理选择、质量证据、科学测量、误差预算和限制声明组织成可审查因果链。复查量规则把模糊评价变成可定位的改进项，帮助判断报告应补证据、重跑、修订参数，还是降低结论等级。

第 9.34 节与 9.30 至 9.33 节共同形成一套完整实践框架：运行事件保证处理可追溯，归档复盘保证判断有证据，发布包保证文件可复查，样本包保证训练可分发，处理报告和复查量规保证结论可评价。这样，第 9 章不仅训练软件操作，也训练从数据到科学结论的完整判断能力。
