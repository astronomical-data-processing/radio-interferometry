## 2.12 补充专题：立体角与天球面积元素<a id='math:sec:sldangle'></a>


立体角在本书中会反复出现于比强度、亮温、波束面积以及天球积分等概念之中。它不属于傅里叶主线的中心部分，但却是把“每单位面积、每单位频率、每单位立体角”的辐射量语言真正用顺的必要补充。

如果读者在第 1 章读到比强度、波束和天球面积元素时仍觉得抽象，可以把本节视为一个回补专题。它的主要任务，是把后续章节经常默认使用的几何量解释清楚。


回顾二维情形：半径为 $r$ 的圆弧长度 $l$ 所张的平面角定义为 $\theta=l/r$。立体角只是这一概念在球面上的推广。若半径为 $r$ 的球面上有一块面积为 $A$ 的区域，那么它所张的立体角定义为

$$ \Omega = \frac{A}{r^2}. $$

这个定义最重要的含义是：立体角描述的是“方向范围”而不是“真实面积”。同一片方向区域投影到不同半径的球面上，面积会随 $r^2$ 改变，但对应的立体角保持不变。因此在天文学中，我们才能把天空上的一个方向元当作与距离无关的几何量来使用。

</p>

半径为 $r$ 的整个球面所张的立体角可按下式计算：

$$
\Omega_{\textrm{sphere}}=\frac{4\pi r^2}{r^2}=4\pi~\textrm{sr}.
$$



### 2.12.1 极小立体角

点 $P$ 的球坐标 $(\rho,\theta,\phi)$ 如 [图 2.12.2 &#10549;](#math:fig:spherical) <!--\ref{math:fig:spherical}--> 所示。

<img src='figures/spherical.png' width=40%>

**图 2.12.2**：点 $P$ 的球坐标 $(\rho,\theta,\phi)$。<a id='math:fig:spherical'></a> <!--\label{math:fig:spherical}-->

点 $P$ 的笛卡尔坐标 $(x,y,z)$ 可以由它的球坐标 $(\rho,\theta,\phi)$ 写为：

<p class=conclusion>
  <font size=4> <b>球坐标与笛卡尔坐标之间的关系</b></font>
  <br>
  <br>

</p>

$$
\begin{aligned}
x &=& \rho \sin \theta \cos \phi\\
y &=& \rho \sin \theta \sin \phi  \\
z &=& \rho \cos \theta 
\end{aligned}
$$

这里约定 $\theta$ 是从极轴量起的极角，$\phi$ 是绕极轴的方位角。后面写面积元和立体角元时，正是这个几何约定决定了为什么会出现 $\sin\theta$ 因子。


为了得到球面上的面积元，我们只需要考察方向发生一个微小变化时，球面上对应小片区域的边长如何变化。若固定 $\rho$ 和 $\theta$、只改变 $\phi$，则点 $P$ 会在 [图 2.12.3 &#10549;](#math:fig:infarea) 上沿纬圈移动，其弧长增量为

$$ \rho\sin\theta\, d\phi. $$

若固定 $\rho$ 和 $\phi$、改变 $\theta$，则点 $P$ 会沿经线移动，其弧长增量为

$$ \rho\, d\theta. $$

因此，球面上的极小面积元为

$$ dS = \rho^2 \sin \theta\, d\theta\, d\phi. $$

再除以半径平方，就得到极小立体角元

$$ d\Omega = \frac{dS}{\rho^2}=\sin \theta\, d\theta\, d\phi. $$

这个结果非常关键，因为它说明：立体角元只依赖方向坐标，不依赖观测者到球面的距离。也正因此，天球积分通常都写成对 $d\Omega$ 的积分，而不是对某个真实面积的积分。

### 2.12.2 小视场近似与波束面积

在射电干涉成像里，我们更常使用的是方向余弦坐标 $(l,m,n)$，其中 $n=\sqrt{1-l^2-m^2}$。这时天球上的立体角元可写为

$$ d\Omega = \frac{dl\,dm}{n}. $$

当视场足够小、也就是 $l,m\ll 1$ 时，有 $n\approx 1$，于是就得到常用的小角近似

$$ d\Omega \approx dl\,dm. $$

这条近似解释了为什么在窄视场成像中，图像平面上的像素面积常可以直接视作立体角面积；但当视场增大时，这种近似会逐渐失效，必须回到完整的球面几何。


立体角的另一个直接应用，是定义波束面积与总流量。若主波束或恢复波束记为 $B(\Omega)$，则其波束立体角定义为

$$ \Omega_B = \int B(\Omega)\, d\Omega. $$

若天体的比强度为 $I_\nu(\Omega)$，则总流量密度满足

$$ S_\nu = \int I_\nu(\Omega)\, d\Omega. $$

这两条关系把“图像亮度”“波束大小”和“积分流量”直接联系起来。后面无论讨论单天线主波束、干涉图像单位，还是亮温与流量之间的换算，都会不断回到这里。


#### 代码演示：小视场近似在什么范围内可靠

立体角最常见的近似是把球面上的方向面积直接当作平面面积处理，即用 $d\Omega \approx dl\,dm$ 或圆盘面积 $\pi\theta^2$ 来代替精确球面结果。下面分别对两类场景做比较：左图是圆形天区的精确立体角与平面近似误差；右图是圆对称高斯波束的精确球面面积与常用近似公式 $\Omega_B \approx 1.133\,\theta_{\rm FWHM}^2$ 的比较。


```python
patch_radius_deg = np.linspace(0.2, 40.0, 400)
patch_radius_rad = np.deg2rad(patch_radius_deg)
omega_cap_exact = 2 * np.pi * (1 - np.cos(patch_radius_rad))
omega_cap_flat = np.pi * patch_radius_rad**2
relative_error_cap = (omega_cap_flat - omega_cap_exact) / omega_cap_exact

beam_fwhm_deg = np.array([0.5, 1, 2, 5, 10, 20, 30])
beam_fwhm_rad = np.deg2rad(beam_fwhm_deg)
theta = np.linspace(0, np.deg2rad(90), 50000)
exact_beam = []
approx_beam = []
for theta_fwhm in beam_fwhm_rad:
    beam = np.exp(-4 * np.log(2) * (theta / theta_fwhm) ** 2)
    exact_beam.append(2 * np.pi * np.trapz(beam * np.sin(theta), theta))
    approx_beam.append(np.pi * theta_fwhm**2 / (4 * np.log(2)))
exact_beam = np.array(exact_beam)
approx_beam = np.array(approx_beam)

fig, axes = plt.subplots(1, 2, figsize=(13, 4.8))

axes[0].plot(patch_radius_deg, 100 * relative_error_cap, color='tab:blue', lw=2)
axes[0].axhline(1.0, color='tab:red', ls='--', label='1% error')
axes[0].set_xlabel('Circular patch radius [deg]')
axes[0].set_ylabel('Relative error of the flat-sky approximation [%]')
axes[0].set_title(r'Circular cap: error of $\pi\theta^2$')
axes[0].grid(alpha=0.3)
axes[0].legend()

axes[1].plot(beam_fwhm_deg, exact_beam * (180 / np.pi) ** 2, 'o-', lw=2, label='exact spherical integral')
axes[1].plot(beam_fwhm_deg, approx_beam * (180 / np.pi) ** 2, 's--', lw=2, label=r'$1.133\,\theta_{\rm FWHM}^2$')
axes[1].set_xlabel('Gaussian beam FWHM [deg]')
axes[1].set_ylabel(r'Beam area [deg$^2$]')
axes[1].set_title('Circular Gaussian beam area')
axes[1].grid(alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```

从图中可以看出：只要视场仍处于“几度以内”的量级，平面近似通常非常好；但一旦视场或波束尺度上升到几十度，球面几何的修正就不再可以忽略。这也是为什么窄视场连通阵成像常常能安全使用 $dl\,dm$，而宽视场、全天空成图和低频大视场观测则必须回到完整的球面表达。
