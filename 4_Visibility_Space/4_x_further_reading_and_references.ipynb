## 4.x 延伸阅读与参考文献 <a id='visib:sec:refs'></a>


第四章的核心是把“几何基线、互相关、可见度函数、$uv$ 覆盖、Fourier 近似”连成一条物理和数学上自洽的链条。下面的阅读材料按用途排列：先补足直观图像，再进入严格推导，最后连接到现代成像算法。

### 基础与直观图像

1. [<cite data-cite='taylor1999synthesis'>Taylor, Carilli & Perley, *Synthesis Imaging in Radio Astronomy II*</cite>](http://adsabs.harvard.edu/abs/1999sira.conf.....T)，特别是前两讲。该书来自 NRAO 综合成像学校，适合在本章之后复习可见度、$uv$ 覆盖和脏束之间的关系。它的优点是图像直观，许多概念直接面向实际阵列和真实数据。

2. [<cite data-cite='condon2016essential'>Condon & Ransom, *Essential Radio Astronomy*</cite>](https://www.cv.nrao.edu/~sransom/web/EssentialRadioAstronomy.html)。其中关于干涉仪、亮温、辐射转移和天线响应的章节可作为射电天文学背景补充。该书的风格更偏天文物理，能帮助把本章的数学量与真实天体辐射联系起来。

### 严格推导与标准教材

3. [<cite data-cite='thompson2017interferometry'>Thompson, Moran & Swenson, *Interferometry and Synthesis in Radio Astronomy*</cite>](https://link.springer.com/book/10.1007/978-3-319-44431-4)。这是射电干涉测量的标准教材。第 1 章、第 2 章、第 3 章和第 4 章可用于补充可见度、相干性、相关器和响应函数；第 14 章对 van Cittert-Zernike 定理及其与经典光学相干理论的关系有更系统的讨论。

4. [<cite data-cite='born1999principles'>Born & Wolf, *Principles of Optics*</cite>](https://doi.org/10.1017/CBO9781139644181)。对相干理论和 van Cittert-Zernike 定理的光学背景感兴趣时，可参考其中关于部分相干光和衍射理论的章节。射电干涉仪与光学干涉仪的仪器实现不同，但互相干函数的数学结构具有共同根源。

### 与现代成像方法的连接

5. [<cite data-cite='cornwell2008wprojection'>Cornwell, Golap & Bhatnagar, 2008, “The Noncoplanar Baselines Effect in Radio Interferometry: The W-Projection Algorithm”</cite>](https://doi.org/10.1109/JSTSP.2008.2005290)。该文从实际成像角度说明 $w$ 项为什么不能简单忽略，并给出 $w$-projection 的基本思想。阅读重点是把本章式 (4.6.1) 中的 $w(n-1)$ 项与宽场成像误差联系起来。

6. [<cite data-cite='offringa2014wsclean'>Offringa et al., 2014, “WSClean: an implementation of a fast, generic wide-field imager for radio astronomy”</cite>](https://doi.org/10.1093/mnras/stu1368)。该文展示了现代宽场成像软件如何把 $uv$ 采样、$w$ 项、权重和去卷积组织成可运行的成像流程。它适合作为第五章和第六章的过渡阅读。

7. [<cite data-cite='hamaker1996rime'>Hamaker, Bregman & Sault, 1996, “Understanding radio polarimetry. I. Mathematical foundations”</cite>](https://doi.org/10.1051/aas:1996146)。该文是 RIME 形式主义的重要源头之一。第四章采用的是标量、方向响应简化后的可见度表达式；RIME 则把极化、仪器增益、方向相关误差等因素纳入统一矩阵框架。

### 阅读顺序建议

完成第四章后，可先读 Taylor, Carilli & Perley 的前两讲来巩固整体图像；随后读 Thompson, Moran & Swenson 中关于可见度与相干性的章节，补齐严格推导；进入第五章宽场和成像误差时，再回到 Cornwell 等关于 $w$-projection 的论文。这样形成的阅读路径与教材主线一致：先理解测量量，再理解近似条件，最后理解算法为什么必须处理这些近似失效的部分。
