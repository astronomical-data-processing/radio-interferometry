## 7.x 延伸阅读与参考文献

第 7 章覆盖的内容跨度很大：既有 Jones/RIME 这种偏数学和测量方程的内容，也有接收机、相关器、主波束、极化、传播效应与 RFI 这种强工程与强观测背景并存的话题。因此，本章的延伸阅读不宜只按“文献罗列”来组织，而更适合按“你想进一步解决什么问题”来组织。

一个比较实用的阅读顺序是：

1. 先把 `7.1` 与 `7.2` 里的 Jones 语言和 RIME 再读一遍，确保符号体系稳固；
2. 再读 `7.3` 到 `7.5`，把模拟链路、数字相关器和主波束看成同一条系统响应链；
3. 最后读 `7.6` 到 `7.8`，把极化、传播与 RFI 理解成“真实观测为何会偏离理想模型”的主要来源。


### 7.x.1 RIME、Jones 语言与方向依赖效应

如果你希望真正掌握本章的理论骨架，那么以下文献应视作核心阅读，而不是“有空再看”的补充材料：

1. [Smirnov 2011a: *Revisiting the radio interferometer measurement equation. I. A full-sky Jones formalism*](http://adsabs.harvard.edu/abs/2011A%26A...527A.106S)
2. [Smirnov 2011b: *Revisiting the radio interferometer measurement equation. II. Calibration and direction-dependent effects*](http://adsabs.harvard.edu/abs/2011A%26A...527A.107S)
3. [Smirnov 2011c: *Revisiting the radio interferometer measurement equation. III. Addressing direction-dependent effects in 21 cm WSRT observations of 3C 147*](http://adsabs.harvard.edu/abs/2011A%26A...527A.108S)
4. [Smirnov 2011d: *Revisiting the radio interferometer measurement equation. IV. A generalized tensor formalism*](http://adsabs.harvard.edu/abs/2011A%26A...531A.159S)

建议阅读方式：

- 第一篇用于建立 Jones/RIME 的统一语言；
- 第二篇用于理解本章反复出现的方向依赖效应为何必须写进测量方程；
- 第三篇最适合作为“理论如何落到真实数据”的案例；
- 第四篇适合在你已经能熟练使用前面三篇语言之后再进入。


### 7.x.2 偏振、馈源与全相干标定

本章的极化部分如果只停留在 Stokes 参数定义，会很容易显得抽象。真正需要建立的是“馈源基底如何把极化信息编码进相关积，以及泄漏如何破坏这种编码”的系统认识。以下文献在这方面仍然非常经典：

1. [Hamaker, Bregman & Sault 1996: *Understanding radio polarimetry. I. Mathematical foundations*](http://adsabs.harvard.edu/abs/1996A%26AS..117..137H)
2. [Hamaker 2000: *Understanding radio polarimetry. IV. The full-coherency analogue of scalar self-calibration*](http://adsabs.harvard.edu/abs/2000A%26AS..143..515H)

阅读建议：

- 前者重点看“线性代数语言如何描述极化”；
- 后者重点看“全相干校准”与“极化保真度”之间的关系；
- 如果你已经完成本章 `7.6`，再回头读这些论文，理解会比第一次接触时清晰得多。


### 7.x.3 数字相关器、接收机与工程实现

第 7 章并不是纯理论章。`7.3` 和 `7.4` 讨论的很多问题，本质上都属于“信号链工程学”。如果你想把“公式上的相关器”真正理解为“硬件和软件里的相关器”，下面两本书非常值得长期放在手边：

1. Richard G. Lyons, *Understanding Digital Signal Processing*
2. Paul Horowitz & Winfield Hill, *The Art of Electronics*

这两本书和干涉测量并不完全同一领域，但它们能帮助你更牢固地理解：

- 采样、量化、窗函数、滤波和 FFT 到底在工程上意味着什么；
- 为什么模拟前端、噪声温度、增益链和阻抗匹配会直接影响后端数据质量；
- 为什么“数字相关器”并不是一个抽象的黑盒，而是一连串有代价、有近似、有资源约束的设计决策。


### 7.x.4 如何把本章用于后续学习

如果你接下来要进入第 8 章“校准”，那么本章里最值得反复回看的并不是所有内容，而是以下几组“接口知识”：

- `7.1` 与 `7.2`：Jones 矩阵、亮度矩阵、可见度矩阵和 RIME；
- `7.5`：主波束为什么属于方向依赖增益；
- `7.6`：馈源旋转和 D-term 为什么会改变极化测量；
- `7.7`：传播效应为什么应被理解为传播 Jones 项；
- `7.8`：flag 为什么不是后处理细节，而是会直接影响校准与成像权重。

对真正要做数据处理的学生来说，本章最重要的产出不是记住多少名词，而是形成一个稳定判断：

**观测系统不会“忠实地搬运天空信号”，它会持续地重写、混合、衰减、旋转、污染和丢弃信息。校准的任务，就是尽可能把这些系统效应显式建模并从数据里拆回去。**
