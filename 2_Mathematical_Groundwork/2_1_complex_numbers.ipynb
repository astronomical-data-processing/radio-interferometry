## 2.1 复数、相位与相量表示<a id='math:sec:complex_numbers'></a><!--\label{math:sec:complex_numbers}-->

在射电干涉测量中，天线接收到的是随时间振荡的电场，而相关器保留下来的则是这些振荡之间的幅度关系与相位关系。复数之所以重要，并不是因为它能把公式写得更“高级”，而是因为它能在同一个对象里同时编码振幅和相位。无论是平面波、延迟补偿、可见度相位，还是后面出现的 Jones 矩阵，最终都离不开复数语言。

因此，本节的目的不是孤立地回顾一遍代数规则，而是把复数解释为干涉测量中的相量表示。读完本节之后，读者应当能够把实部、虚部、模、辐角、复共轭和欧拉公式直接联系到波的叠加、相位差以及相关测量这些后续会反复出现的物理图像。


### 2.1.1. 复数域<a id='math:sec:the_field_of_complex_numbers'></a><!--\label{math:sec:the_field_of_complex_numbers}-->

复数构成一个记作 $\mathbb{C}$ 的数域。

<a id='math:eq:0001'></a><!--\label{math:eq:0001}-->

$$z = x+\imath y \mathrm,$$

其中 $x \in \mathbb{R}$ 和 $y \in \mathbb{R}$ 都是实数，加法与乘法定义如下：

<a id='math:eq:0002'></a><!--\label{math:eq:0002}-->

$$
\begin{aligned}
z_i    &= x_i+\imath y_i\qquad i=1, 2\\
z_1+z_2&=(x_1+x_2)+\imath (y_1+y_2)\\
z_1\cdot z_2&= (x_1 x_2-y_1 y_2)+\imath (x_1 y_2+y_1 x_2) .\\
\end{aligned}
$$

复数 $z \in \mathbb{C}$ 的实部 $\Re\left\{z\right\} \in \mathbb{R}$、虚部 $\Im\left\{z\right\} \in \mathbb{R}$、复共轭 $z^* \in \mathbb{C}$，以及绝对值（或模） $|z| \in \mathbb{R}$ 定义为

<a id='math:eq:0003'></a><!--\label{math:eq:0003}-->

$$
\begin{aligned}
z &=\, x + \imath y \\
\Rightarrow \Re{\left\{z\right\}} &=\, x \\
\Im\left\{z\right\} &=\, y \\
z^* &=\, x + \imath(-y) = x - \imath y \quad\left[\Rightarrow \, z_1^*\,{z_2}^* =\, (z_1\,z_2)^*\right]\\
\lvert z\rvert &=\, \sqrt{zz^\ast} =\, \sqrt{x^2+y^2} \in \mathbb{R}
\end{aligned}
$$

实数域可以看作复数域的一个子域，其定义为

<a id='math:eq:0004'></a><!--\label{math:eq:0004}-->

$$ x \in \mathbb{R}, \qquad z_x \in \mathbb{C}\\
z_x = x \Leftrightarrow z_x = x +\imath 0 $$

虚数单位 $\imath$ 可以理解为一个数，即实数 $-1$ 的平方根。

$$
\imath = \sqrt{-1}
$$

**公式 2.1.1** <a id='math:eq:0005'></a><!--\label{math:eq:0005}-->

我们把证明“在这些定义下复数确实构成一个域”留给读者。

### 2.1.2. 欧拉公式<a id='math:sec:eulers_formula'></a><!--\label{math:sec:eulers_formula}--><!--\label{math:sec:eulers_formula}-->

欧拉公式以 [<cite data-cite='euler'>Leonhard Euler&#10548;</cite>](https://en.wikipedia.org/wiki/Leonhard_Euler) (1707-1783) 命名。它指出，对于任意实数 $\phi \in \mathbb{R}$，有

$$
\exp(\imath \phi) = e^{\imath \phi}=\cos \phi+\imath\sin \phi\mathrm,
$$

**公式 2.1.2** <a id='math:eq:001'></a><!--\label{math:eq:001}-->

其中 $e$ 是自然对数的底，$\cos$ 和 $\sin$ 分别表示余弦与正弦函数。欧拉公式可以通过把式中各项展开为 [<cite data-cite='wiki:taylorseries'>泰勒级数&#10548;</cite>](https://en.wikipedia.org/wiki/Taylor_series) 来证明。

因此，函数 $\exp(\imath \phi)$（其中 $\phi \in \mathbb{R}$）描述了复平面单位圆上一点的位置：

<img width=400 src="figures/Euler's_formula.svg">

**图 2.1.1** *复平面中的单位圆。来源：[<cite data-cite='wiki:eulersformula'>Wikipedia&#10548;</cite>](https://en.wikipedia.org/w/index.php?title=Euler%27s_formula&oldid=700599758)*

注意，对于任意角 $\phi \in \mathbb{R}$，都有

<a id='math:eq:0011'></a><!--\label{math:eq:0011}-->

$$
\cos^2 \phi + \sin^2 \phi = 1
$$

这意味着任意复数 $z \in \mathbb{C}$ 都可以写成如下形式：

<a id='math:eq:0012'></a><!--\label{math:eq:0012}-->

$$
z = \lvert z\rvert\exp{(\imath\phi)} \qquad ,
$$

其中 $\phi \in \mathbb{R}$ 称为 $z$ 的幅角。还要注意 $\lvert z \rvert \in \mathbb{R}$，且 $\lvert \exp{(\imath\phi)}\rvert = 1$。

在干涉测量里，这种极坐标形式比直角坐标形式更自然。若
$$z = |z|e^{\imath\phi},$$
那么 $|z|$ 可以直接理解为振幅，$\phi$ 则是相位；复共轭 $z^* = |z|e^{-\imath\phi}$ 对应相位取反；两个复数相乘时振幅相乘、相位相加。这正是电场、可见度和复增益习惯写成复指数形式的根本原因。


#### 代码演示：相量如何同时编码振幅与相位

下面的实验把三个同频率、不同振幅和不同相位的振荡同时画在复平面和时间域中。请特别关注两点：一是复平面中的转角如何对应时间域中的相位平移；二是两个相量相乘并取共轭后，只会留下它们之间的相对相位差。这正是干涉测量里复可见度相位最核心的直觉来源。


```python
frequency_hz = 2.0
omega = 2 * np.pi * frequency_hz
t = np.linspace(0, 1.0, 800, endpoint=False)
phases = np.deg2rad([0.0, 45.0, 120.0])
amplitudes = np.array([1.0, 1.0, 0.7])
labels = ['reference', 'phase lead 45°', 'phase lead 120°']
colors = ['tab:blue', 'tab:orange', 'tab:green']

phasors = [amp * np.exp(1j * (omega * t + phi)) for amp, phi in zip(amplitudes, phases)]
t0 = 0.18
idx0 = np.argmin(np.abs(t - t0))

fig, axes = plt.subplots(1, 3, figsize=(15, 4.6))

ax = axes[0]
for amp, phi, label, color, z in zip(amplitudes, phases, labels, colors, phasors):
    circle = amp * np.exp(1j * np.linspace(0, 2 * np.pi, 400))
    ax.plot(circle.real, circle.imag, color=color, alpha=0.25)
    ax.arrow(0, 0, z[idx0].real, z[idx0].imag,
             color=color, width=0.01, length_includes_head=True)
    ax.plot(z[idx0].real, z[idx0].imag, 'o', color=color, label=label)
ax.axhline(0, color='0.7', lw=1)
ax.axvline(0, color='0.7', lw=1)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('Real part')
ax.set_ylabel('Imag part')
ax.set_title(f'Phasors in the complex plane (t = {t[idx0]:.2f} s)')
ax.legend(loc='upper left', fontsize=9)

ax = axes[1]
for z, label, color in zip(phasors, labels, colors):
    ax.plot(t, z.real, color=color, lw=2, label=label)
ax.set_xlabel('Time [s]')
ax.set_ylabel('Real part of the signal')
ax.set_title('Time-domain projection of the phasors')
ax.grid(alpha=0.3)
ax.legend(fontsize=9)

ax = axes[2]
reference = phasors[0]
for z, label, color in zip(phasors[1:], labels[1:], colors[1:]):
    visibility = reference * np.conj(z)
    ax.plot(visibility.real[0], visibility.imag[0], 'o', ms=9, color=color,
            label=f'{label} relative to reference')
    ax.arrow(0, 0, visibility.real[0], visibility.imag[0],
             color=color, width=0.01, length_includes_head=True)
    print(f'{label:>15s}: |z_1 z_2^*| = {np.abs(visibility[0]):.3f}, phase = {np.rad2deg(np.angle(visibility[0])):6.2f} deg')
ax.axhline(0, color='0.7', lw=1)
ax.axvline(0, color='0.7', lw=1)
ax.set_aspect('equal', adjustable='box')
ax.set_xlabel('Real part')
ax.set_ylabel('Imag part')
ax.set_title(r'Relative phase: $z_1 z_2^*$')
ax.legend(fontsize=9, loc='upper left')

plt.tight_layout()
plt.show()
```

最后一幅图已经非常接近干涉测量中的基本对象了。若两个电场相量分别记为 `z_1` 与 `z_2`，那么 `z_1 z_2^*` 的幅度反映它们的相关强度，而其相位只取决于二者之间的相对相位差，而不再依赖“此刻已经转到了单位圆的哪个位置”。这正是为什么相关器能够从快速振荡的电场中提取稳定的相位信息。


### 2.1.3 周期函数与复数<a id='math:sec:periodic_functions_and_complex_numbers'></a> <!--\label{math:sec:periodic_functions_and_complex_numbers ss:pacn:pfacn}-->

利用欧拉公式（[$\S$ 2.1.2](#math:sec:eulers_formula)<!--\ref{math:sec:eulers_formula}-->），可知一个余弦型实函数

<a id='math:eq:002'></a><!--\label{math:eq:002}-->

$$
f(t)=a\,\cos (\omega t+\phi)
$$

可以改写为

<a id='math:eq:1_003'></a><!--\label{math:eq:1_003}-->

$$
\begin{aligned}
f(t) & = a\,\cos (\omega t+\phi) \\
     & =  z \, e^{\imath\omega t} + z^* \, e^{-\imath\omega t}
\end{aligned}
 ,
$$

其中

<a id='math:eq:004'></a><!--\label{math:eq:004}-->

$$
\begin{aligned}
z     & = z_1 + \imath \,z_2 \\
z_1 & =  \frac{a}{2} \cos \phi\\
z_2 & =  \frac{a}{2} \sin \phi
\end{aligned}
,
$$
**公式 2.1.3**

其中 $z_1 \in \mathbb{R}$ 且 $z_2 \in \mathbb{R}$。反过来可得（假设 $a > 0$）：

<a id='math:eq:005'></a><!--\label{math:eq:005}-->

$$
\begin{aligned}
a &= 2\sqrt{z_1^2 + z_2^2} \\
  &= 2\lvert z \rvert\\
\phi &= \operatorname{atan2}(z_2,z_1) =
\begin{cases}
  \arctan\frac{z_2}{z_1}, & z_1 > 0\\
  \arctan\frac{z_2}{z_1} - \operatorname{sign}(z_2)\pi, & z_1 < 0\\
  \operatorname{sign}(z_2)\frac{\pi}{2}, & z_1 = 0 \land z_2 \neq 0
\end{cases}
\end{aligned}
$$

对应的正弦型周期函数留给读者自行推导。

另一种把实值周期函数写成“复数乘以指数函数”形式的方法，是把 $a$ 视为复数 $z_a$：

<a id='math:eq:006'></a><!--\label{math:eq:006}-->

$$
\begin{aligned}
\Re\left\{z_a\right\} & = a\\
\Im\left\{z_a\right\} & = 0
\end{aligned}
,
$$

于是（见 [公式 2.1.1 &#10549;](#math:eq:0004) <!--\ref{math:eq:0004}-->）

<a id='math:eq:007'></a><!--\label{math:eq:007}-->

$$
\begin{aligned}
f(t) & = a\,\cos (\omega t+\phi) \\
     & =  \Re \left\{{z_a \, e^{\imath(\omega t+\phi)}}\right\}\\
     & =  \Re \left\{{a \, e^{\imath(\omega t+\phi)}}\right\}\\
     & =  a \, \Re \left\{{e^{\imath(\omega t+\phi)}}\right\}
\end{aligned}
\qquad .
$$
**公式 2.1.4**

这是最常见的约定。

要在这两种写法之间转换，只需使用 [公式 2.1.3 &#10549;](#math:eq:004) <!--\ref{math:eq:004}-->：

<a id='math:eq:008'></a><!--\label{math:eq:008}-->

$$
\begin{aligned}
z &= z_1 + \imath\,z_2 \\
z_1 &= \frac{a}{2} \cos \phi = \frac{z_a}{2} \cos \phi \\
z_2 &= \frac{a}{2} \sin \phi = \frac{z_a}{2} \sin \phi
\end{aligned}
\qquad .
$$

[公式 2.1.4 &#10549;](#math:eq:007) <!--\ref{math:eq:007}--> 还可以进一步推广：把 $a \in \mathbb{R}$ 替换为任意一个模长相同但幅角不同的复数 $z^\prime$。换句话说，就是让 $a$（或 $z_a$）乘上单位圆上的一个复数 $x$：

<a id='math:eq:009'></a><!--\label{math:eq:009}-->

$$
\begin{aligned}
f(t) & = a\,\cos (\omega t+\phi) \\
     & =  \Re \left \{{z^\prime \, e^{\imath(\omega t+\phi)}}\right\}\\
   z^\prime &= z^\prime_1 + \imath z^\prime_2\\
                   &= x \cdot a\\
                   &= (x_1 + \imath\,x_2) \cdot a\\
\lvert z^\prime \rvert & = \lvert a \rvert  \Rightarrow \\
\lvert x \rvert & = 1 \Rightarrow \\
x &= e^{\imath\alpha}\\
x_1 &= \cos \alpha\\
x_2 &= \sin \alpha\\
z^\prime_1 &= a \cdot \cos \alpha\\
z^\prime_2 &= a \cdot \sin \alpha
\end{aligned}
\qquad .
$$

同样地，要确定 $\alpha$：

<a id='math:eq:010'></a><!--\label{math:eq:010}-->

$$
\begin{aligned}
a &> 0 \Rightarrow \\
\alpha &= \operatorname{atan2}(z^\prime_2,z^\prime_1) =
\begin{cases}
  \arctan\frac{z^\prime_2}{z^\prime_1}, & z^\prime_1 > 0\\
  \arctan\frac{z^\prime_2}{z^\prime_1} - \operatorname{sign}(z^\prime_2)\pi, & z^\prime_1 < 0\\
  \operatorname{sign}(z^\prime_2)\frac{\pi}{2}, & z^\prime_1 = 0 \land z^\prime_2 \neq 0
\end{cases}
\end{aligned}
\qquad .
$$

**公式 2.1.5**

这时，只要利用恒等式，[公式 2.1.4 &#10549;](#math:eq:007) <!--\ref{math:eq:007}--> 就可写成

$$\begin{aligned}
f(t) & = a\,\cos (\omega t+\phi) \\
     & =  \Re \left \{{a \, e^{\imath(\omega t+\phi)}}\right\}\\
     & =  \Re \left \{{a \cdot 1 \cdot e^{\imath(\omega t+\phi)}}\right\}\\
     & =  \Re \left \{{a (\, x \, x^*) \, e^{\imath(\omega t+\phi)}}\right\}\\
     & =  \Re \left \{{z^\prime \left ({x^* \, e^{\imath(\omega t+\phi)}}\right)} \right\}\\
     & =  \Re \left \{{z^\prime \left ({e^{-\imath\,\alpha} \, e^{\imath(\omega t+\phi)}}\right)} \right\}\\
     & =  \Re \left \{{z^\prime \, e^{\imath(\omega t+\phi-\alpha)}} \right\}\\
     & =  \Re \left \{{z^\prime \, e^{\imath(\omega t+\phi^\prime)}} \right\}
\end{aligned}
\qquad .
$$

因此，对于任意满足 $\lvert a \rvert = \lvert z^\prime \rvert$ 的复数 $z^\prime$，都可以写成：

<a id='math:eq:012'></a><!--\label{math:eq:012}-->

$$
\begin{aligned}
f(t) & = a\,\cos (\omega t+\phi) \\
     & =  \Re \left \{{z^\prime \, e^{\imath(\omega t+\phi^\prime)}}\right\}
\end{aligned}
\qquad ,
$$

其中
<a id='math:eq:013'></a><!--\label{math:eq:013}-->

$$
\begin{aligned}
\phi^\prime & = \phi - \alpha \\
\end{aligned}
\qquad ,
$$

而 $\alpha$ 按 [公式 2.1.5 &#10549;](#math:eq:010) <!--\ref{math:eq:010}--> 计算。最常见的约定是令 $\Im\left\{z^\prime\right\} = 0$，这样就回到了 [公式 2.1.4 &#10549;](#math:eq:007) <!--\ref{math:eq:007}-->。

在本节中，我们使用了常见记号 $t$ 和 $\omega$，它们通常分别表示时间与角频率，用于描述随时间振荡的函数。当然，这也完全可以推广到其他周期函数或其他自变量。例如，把 $\omega t$ 替换为 $2\pi \frac{t}{T}$、${\bf{k}}\cdot{\bf{x}}-\omega t$，或 $2\pi(\frac{l}{\lambda}x+\frac{m}{\lambda}y+\frac{n}{\lambda}z- \nu t)$，公式形式都保持不变。
