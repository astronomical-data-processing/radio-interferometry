# 第六章：成像中的去卷积<a id='deconv:sec:intro'></a>


在综合孔径成像里，人们通常把“成像”理解为两个相互关联的过程：首先是把可见度转换成天空的空间域表示；其次是利用阵列的点扩散函数（PSF）对所得图像进行去卷积，尽可能恢复出更接近“真实”天空的结果。

正如上一章已经看到的，最终图像会明显受到阵列构型以及观测参数的影响，例如带宽、观测时长、指向赤纬等。而这些因素都是已知的，且由它们决定的阵列 PSF 本身又独立于真实天空。因此，一个非常合理的去卷积目标，就是尽可能把“仪器带来的效应”与“天空本身的结构”分离开来。

在日常说法里，人们常会说“把脏图变成洁图”。这背后的意思是：利用我们对阵列 PSF 的了解，以及对天空中可能出现的源类型的先验认识，通过某种去卷积算法，通常是 CLEAN 及其变体，去构造一幅更完整、更可信的图像。

为了理解射电干涉测量中这些去卷积算法的动机与形式，本章会先讨论天空模型通常怎样表示；然后从最基本的 CLEAN 形式开始，逐步引入更复杂的实现；接着讨论何时应停止去卷积、怎样评价图像质量，以及最终图像是如何构建出来的；随后介绍用于从图像中自动提取天体源信息的源搜索方法，并进一步把正则化、过拟合和残差统计纳入去卷积可信度判断。
