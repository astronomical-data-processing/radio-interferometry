# 格式说明 <a id='preface:sec:style'></a>

#### 数学符号

本书给出了一个符号总表来说明所用符号的含义，基本约定如下：

1. 向量、标量和矩阵：
 * $a, A$：标量
 * $\mathbf{A}$, $\boldsymbol{\mathcal{A}}$：矩阵
 * $\mathbf{a}$：向量

2. $2\times2$ 极化矩阵与 $N\times N$ 非极化矩阵：
 * $\mathbf{A}$：表示一个 $2\times2$ 的极化 Jones 矩阵
 * $\boldsymbol{\mathcal{A}}$：表示一个 $N\times N$ 的非极化矩阵，用来统一描述阵列相关的非极化量

3. Jones 与 Jacobian：
 * 用 $\mathbf{J}$ 表示 Jones 矩阵，用 $\mathbb{J}$ 表示 Jacobian 矩阵

4. 傅里叶变换：
 * 使用 $\mathscr{F}\{\cdot\}$ 表示傅里叶变换

5. 用下标消除歧义：
 * 当一个符号可能表示多个不同量时，应通过下标加以区分。例如 $\lambda$ 既可表示波长，也可能表示 LM 阻尼因子，因此可写成 $\lambda_{\textrm{LM}}$ 表示阻尼因子，而 $\lambda$ 继续表示波长。必要时请在总表中加入新的下标说明。

一般可用符号列表见 [术语表 &#10142;](1_glossary.ipynb#preface:sec:glossary)。

如果在编写过程中引入了新的专门术语，请在第一次出现时明确说明，并把它加入术语表。

#### Notebook 与目录命名规范

不同章节内容存放在不同目录下，目录通常采用“章编号 + 章名称”的命名方式，并用下划线替代空格，例如 `6_deconvolution_in_imaging`。目录中主要放 notebook 文件，其它数据或资源文件则可放在相应的全局数据目录中。

Notebook 文件名通常以章号或章内顺序号开头，并包含节名称，同样用下划线连接。例如第 6 章“天空模型”这一节可命名为 `6_1_sky_models.ipynb`。

#### Notebook 长度

一章通常由多个节构成，每节下面还可以继续划分为小节。为了保持 notebook 适中、易读，通常一个 notebook 只对应一个主节；如果某节内容过长，可以进一步拆分成多个 notebook。

#### Notebook 页眉

每个 notebook 开头都应包含导航链接，至少指向总目录、术语表、本章导言、上一节和下一节。标准 Python 模块与本节专用模块的导入应放在导航之后。随后应给出与目录项一致的 notebook 标题。

#### Notebook 页脚

每个 notebook 结尾应包含指向下一节的链接；若该 notebook 已是本章最后一节，则应给出通往下一章的链接。

#### 章节导言

每章应包含一个简短的导言 notebook，用于概述本章主题并列出本章各 notebook。导言结尾还应给出本章编辑者与贡献者名单，并尽量注明负责的具体小节。

#### 章节结语

每章最后一个 notebook 通常应包含本章总结，以及参考文献和延伸阅读材料。
