## 1.7 谱线辐射<a id='science:sec:line_emission'></a>

连续谱告诉我们源总体有多亮、谱有多陡、亮温有多高；谱线则进一步回答气体由什么成分组成、以多大速度运动、处于怎样的激发和密度状态。对射电天文学而言，谱线辐射的重要性几乎怎么强调都不为过。星系旋转曲线、分子云运动学、恒星形成前体的密度结构、气体柱密度估计、外流和吸积的速度场，许多关键结论都建立在谱线测量之上。

谱线观测与连续谱观测的差别，不只是“多了一条频率轴”。连续谱常把宽带信号平均起来提高灵敏度，而谱线必须保留频率方向上的细结构。频率偏移对应视向速度，线宽对应热运动、湍流或大尺度速度梯度，线强又取决于丰度、激发、光深和背景辐射。因此，谱线分析是一条从微观能级跃迁到三维数据立方体，再到宏观气体动力学解释的完整链条。


### 1.7.1 从微观跃迁到辐射传输

在 [§ 1.3](1_3_radiative_transport.ipynb) 中，我们已经讨论过发射系数 $\varepsilon_\nu$ 和吸收系数 $\kappa_\nu$ 在辐射传输中的作用。对谱线辐射而言，这两个量可以直接和原子或分子的能级跃迁联系起来。考虑两个能级 $1$ 和 $2$，高能级向低能级的自发辐射概率由 $A_{21}$ 描述，受激发射和受激吸收分别由 $B_{21}$ 与 $B_{12}$ 描述。这些系数并不是任意常数；在热平衡条件下，它们必须满足

<a id='science:eq:07_001'></a><!--\label{science:eq:07_001}-->
$$
 g_1 B_{12}=g_2 B_{21},
\qquad
 A_{21}={8\pi h\nu_0^3\over c^3}B_{21},
$$

其中 $g_1$ 和 $g_2$ 是能级简并度，$\nu_0$ 是静止跃迁频率。

真实谱线不是无限窄的单一频率尖峰。热运动、湍流、自然展宽、碰撞展宽，以及视线方向上的系统速度梯度，都会把谱线扩展成有限宽度的轮廓。通常引入归一化线轮廓函数 $\varphi(\nu)$，满足 $\int\varphi(\nu)d\nu=1$。把能级布居和线轮廓写入辐射传输表达式，可得

<a id='science:eq:07_002'></a><!--\label{science:eq:07_002}-->
$$
\kappa_\nu={h\nu_0\over c}N_1B_{12}
\left(1-{g_1N_2\over g_2N_1}\right)\varphi(\nu),
\qquad
\varepsilon_\nu={h\nu_0\over 4\pi}N_2A_{21}\varphi(\nu).
$$

括号中的项体现了受激发射对净吸收的抵消。如果上能级布居相对于下能级过高，净吸收甚至可以变为负值，对应 maser 放大。若两能级布居可用激发温度 $T_{\rm ex}$ 描述，则

<a id='science:eq:07_003'></a><!--\label{science:eq:07_003}-->
$$
{N_2\over N_1}={g_2\over g_1}\exp\left(-{h\nu_0\over kT_{\rm ex}}\right).
$$

这些公式说明，谱线强度不仅取决于“有没有这种原子或分子”，还取决于能级布居、密度、温度、背景连续谱和光学深度。同一条线在不同环境中既可能表现为发射，也可能表现为吸收；同一个分子不同跃迁之间的强度比，也常常比单条线总强度包含更多物理信息。


### 1.7.2 线轮廓为什么携带速度信息

谱线最有用的特征之一，是静止频率通常已由实验室测量得非常精确，因此观测频率的偏移可以直接转化为速度。若采用射电速度定义，视向速度为

<a id='science:eq:07_004'></a><!--\label{science:eq:07_004}-->
$$
{V_{\rm radio}\over c}={\nu_0-\nu\over \nu_0}.
$$

在低速极限下，这与普通多普勒公式一致。对高红移或高精度工作，还需要区分射电速度、光学速度和相对论速度定义；但在本章的基础讨论中，上式已经足以说明频率轴如何变成速度轴。频率分辨率和速度分辨率之间满足近似关系

$$
\Delta v\simeq c{\Delta\nu\over \nu_0}.
$$

例如，H I 21 cm 线在 $1.420$ GHz 附近，$5\,\rm km\,s^{-1}$ 的速度分辨率对应约 $24$ kHz 的频率分辨率。若通道太宽，窄线会被稀释，速度结构会被平滑；若通道足够细，则可以区分热展宽、湍流展宽、旋转、外流或多个速度分量。

**图 1.7.1**：谱线轮廓与速度信息。左图示意热展宽、非热运动和合成线宽；右图示意 H I 21 cm 附近频率轴与射电速度轴之间的对应关系。


线宽的解释需要谨慎。热展宽与粒子质量和温度有关，同样温度下轻元素线更宽；湍流和未分辨速度梯度会带来额外非热展宽；旋转盘、外流、吸积流或相互作用星系则可能形成非高斯、多峰或不对称线形。因此，线轮廓不是一个附属细节，而是谱线观测中最直接的动力学信息之一。


### 1.7.3 H I 21 cm：射电天文学的基石<a id='science:sec:atomic_hydrogen'></a>

中性氢的 21 cm 超精细跃迁，是射电天文学中最具标志性的谱线之一。它来自氢基态中电子自旋与质子自旋相对取向的变化，静止频率为 $1.420405751\,\rm GHz$。这条线的自发跃迁概率极低，$A_{10}\simeq 2.85\times10^{-15}\,\rm s^{-1}$，单个原子的平均跃迁时间约为千万年量级。它之所以仍然能被观测到，是因为宇宙中的中性氢极其丰富，星系盘、外延气体、潮汐结构和星系间相互作用环境中都可能积累出可测的总信号。

在定义自旋温度 $T_{\rm spin}$ 后，中性氢柱密度与光学深度满足

<a id='science:eq:07_005'></a><!--\label{science:eq:07_005}-->
$$
N_{\rm H\,I}=1.823\times10^{18}
\left({T_{\rm spin}\over {\rm K}}\right)
\int \tau(v)\,{dv\over {\rm km\,s^{-1}}}
\quad {\rm cm^{-2}}.
$$

在许多外星系 H I 发射工作中，21 cm 线可近似看作光学薄，此时 $T_b\simeq T_{\rm spin}\tau$，上式化为更常用的亮温积分形式：

<a id='science:eq:07_006'></a><!--\label{science:eq:07_006}-->
$$
N_{\rm H\,I}=1.823\times10^{18}
\int {T_b(v)\over {\rm K}}\,{dv\over {\rm km\,s^{-1}}}
\quad {\rm cm^{-2}}.
$$

这个公式把观测得到的谱线亮温积分直接转换成中性氢柱密度，使 21 cm 数据不仅能回答“哪里有气体”，还能估计“有多少气体”。结合速度信息后，就可以绘制星系旋转曲线、研究潮汐尾和气体剥离、分析吸积与外流，甚至通过旋转曲线在大半径处保持平坦这一事实间接揭示暗物质问题。

需要注意的是，光学薄近似并非永远成立。在冷而致密的 H I 云中，自吸收会使柱密度被低估；若背景存在强连续谱，同一条线还可能表现为吸收线。H I 21 cm 的强大之处不在于它没有复杂性，而在于它的基本物理足够清楚，并且能在极大尺度上稳定追踪中性气体。


### 1.7.4 数据立方体、通道图和矩图

谱线干涉观测的自然数据产品不是一张二维图，而是一个三维数据立方体：两个轴是天空坐标，第三个轴是频率或速度。沿速度轴切片得到通道图，每一张通道图显示某个速度范围内的辐射空间分布；沿速度轴积分得到 moment 0 图，通常表示积分强度或柱密度；用强度加权速度得到 moment 1 图，表示速度场；用二阶矩得到 moment 2 图，常用来表示线宽或速度弥散。

**图 1.7.2**：谱线数据立方体和常见派生产品。通道图保留速度切片信息，moment 0 表示积分强度，moment 1 表示速度场，moment 2 或线宽图则用于描述速度弥散。


从数据立方体到科学解释之间存在许多实际选择。构造 moment 图前通常需要做掩膜，否则噪声会在大量空通道中累积；速度场图在多峰线或低信噪比区域可能失去明确物理含义；对旋转盘而言，位置-速度图（PV diagram）往往比简单 moment 1 图更能显示动力学结构。谱线观测因此不仅要求会成像，还要求理解频率轴、噪声、掩膜和运动学模型之间的关系。


### 1.7.5 除了 H I，还有哪些重要射电谱线<a id='science:sec:other_line_emission'></a>

射电波段可观测的谱线远不止 21 cm。冷分子气体最常用的入口是 CO 转动线，例如 CO(1-0) 位于 $115.271$ GHz，CO(2-1) 位于 $230.538$ GHz。由于 H$_2$ 本身在冷分子云中缺乏容易观测的低能电偶极跃迁，CO 常被用作分子气体质量的替代示踪量；但 CO 到 H$_2$ 的转换因子依赖金属丰度、辐射场、密度和动力学环境，因此并不是一个绝对常数。HCN、HCO$^+$、CS 等分子具有更高临界密度，常用于追踪更致密的分子气体，但也更容易受到激发条件和光深影响。

射电复合线来自高主量子数电子在离子附近复合后发生的跃迁，可用于研究 H II 区和电离气体的温度、密度和速度场。Maser 线则代表另一类特殊但极其有用的谱线，例如 OH、H$_2$O 和 CH$_3$OH maser。它们由布居反转导致受激辐射放大，亮温可以极高，线宽很窄，常用于研究恒星形成区、晚型恒星包层和活动星系核附近的极精细速度结构。水 maser 甚至可以通过 VLBI 测量几何距离和核区盘运动。

这些谱线的微观机制不同，但观测逻辑相通：必须保留频谱结构，并把线强、线宽、中心速度、线形、空间分布和多条跃迁之间的比值放在一起解释。谱线的总通量只是起点，不是终点。


### 1.7.6 为什么谱线数据处理与连续谱不同

谱线数据处理是后续实践章节必须专门展开的一条主线。与连续谱相比，谱线观测通常需要更高频谱分辨率，因此数据量更大，单通道噪声也更高。bandpass 标定必须足够稳定，否则频率方向上的增益起伏会伪装成宽浅谱线或吸收结构。若目标同时具有强连续谱和弱谱线，还需要在 $uv$ 域或图像域做连续谱扣除，避免宽带连续谱泄漏到线立方体中。

干涉阵谱线观测还会继承空间采样问题。扩展气体在某些速度通道中可能具有很大的角尺度，若缺少短基线或总功率数据，就会丢失大尺度通量，使柱密度、积分强度和速度场受到系统影响。另一方面，频率轴上的 RFI、边带结构、Hanning 平滑、通道平均和重采样都会改变有效分辨率和噪声相关性。谱线观测因此不能简单理解为“连续谱成像再多一维”；它需要独立的参数选择逻辑，也需要清楚理解数据立方体、矩图、速度场和柱密度估计之间的关系。


### 1.7.7 本节小结

谱线辐射把微观能级跃迁与宏观气体物理联系起来，是研究成分、温度、密度、速度场和柱密度的核心工具。线轮廓不仅给出“宽”或“窄”，还包含热运动、湍流、旋转、外流和多速度分量等动力学信息。H I 21 cm 是最基础的射电谱线之一，能够把中性气体分布、速度场和柱密度直接联系起来；分子线、复合线和 maser 线则分别打开冷分子气体、电离气体和高亮温特殊环境的观测窗口。

对干涉测量而言，谱线观测要求频谱分辨率、bandpass 稳定性、连续谱扣除、短间距恢复和数据立方体分析同时成立。进入后续章节后，这些要求会逐渐转化为具体的相关器设置、成像参数、权重选择和科学产品构造方法。
