## 9.36 真实轻量样本包与项目练习材料

第 9.33 节讨论了轻量样本包的设计原则，第 9.35 节讨论了综合课程项目。本节把这些原则落成一个实际目录：`sample_packages/continuum_lightweight_replay/`。该目录不是完整观测数据，也不假装替代 Measurement Set；它提供一组可直接检查的图像数组、残差图、局部噪声图、源表、区域文件、manifest、参数文件、任务说明和报告模板，用于训练低面亮度连续谱归档复盘中的核心判断。

这个样本包的价值在于让教材中的“证据链”变成文件。图像产品可以被打开或重新统计，源表可以被检查，区域文件可以进入测量，manifest 可以核对校验和，任务说明可以区分标准路径、盲测任务和失败对照，报告模板可以把结论写成可复查形式。它不追求大，而追求边界清楚。


![真实轻量样本包结构](figures/practical_real_sample_package_layout.png)

图 9.36.1 展示样本包结构。图像产品、源表与区域、清单、参数、任务和报告模板共同形成可复查练习骨架。样本包不是一组截图，而是一套能被检查、重算和评分的练习材料。


### 9.36.1 样本包目录

`.npy` 文件保存小型二维数组，分别代表重跑图像、残差图和局部噪声图。它们不是天文标准 FITS 产品，但足以训练区域测量、局部噪声、残差结构和报告边界。manifest 中保留校验和，使样本身份可以被核对。

### 9.36.2 支持和不支持的训练目标

该样本包支持四类训练。第一，低面亮度连续谱上限：通过区域文件、局部噪声图和误差预算给出受限结论。第二，源表可靠性检查：通过源表中的致密源、弥散候选和边缘检查源训练选择函数判断。第三，产品差异审计：把图像、残差和局部噪声共同放入报告。第四，处理报告与限制声明：根据 manifest、参数和质量摘要写出可审查结论。

该样本包不支持完整校准求解、谱指数测量、短间距恢复、精确通量标尺重新定标和真实 pipeline 任务重跑。这些限制必须写入报告。样本包越小，越需要清楚说明不能回答的问题；这不是缺陷，而是训练结论边界的机会。


### 9.36.3 标准路径、盲测任务与失败对照

`tasks/standard_path.md` 定义标准路径：核对产品身份，比较图像、残差和局部噪声图，使用区域文件给出上限，并写出限制声明。`tasks/blind_task.md` 定义盲测任务：改变测量区域或主波束裁剪条件，重新判断上限是否稳定。`tasks/failure_contrasts.md` 列出失败对照：只使用全图均方根给上限，忽略残差图中的负碗结构，把边缘检查源直接纳入可靠源表，或在报告中省略样本不支持的结论。


![标准路径、盲测任务与失败对照](figures/practical_real_sample_task_split.png)

图 9.36.3 展示项目练习材料的三类任务。标准路径校准期望产物，盲测任务检验迁移能力，失败对照训练质量判断。没有盲测和失败对照，项目练习很容易退化为照抄标准答案。


### 9.36.4 简单测量公式

若测量区域包含 $N_{\rm beam}$ 个独立波束，局部噪声为 $\sigma_{\rm rms}$，通量标尺相对误差为 $f_{\rm flux}$，则区域积分通量的简化误差预算可写成

$$
\sigma_S^2 \simeq N_{\rm beam}\sigma_{\rm rms}^2 + (f_{\rm flux}S)^2 + \sigma_{\rm model}^2.
$$

样本包中的 `measurement_parameters.yaml` 已记录是否纳入相关噪声和通量标尺误差。练习报告不要求给出唯一标准答案，而要求说明噪声取自哪里、区域如何选择、是否受残差结构限制、上限适用于什么尺度，以及哪些结论不能由该样本支持。


### 9.36.5 练习产物

完成该样本包练习后，应形成四个产物。第一，产品身份核对表，说明清单和校验和是否一致。第二，质量判读表，说明图像、残差、局部噪声和源表中的主要风险。第三，测量或上限报告，说明区域、噪声、误差项和结果。第四，限制声明，说明样本支持和不支持的科学结论。若作为综合项目的一部分，这些产物可以直接进入第 9.35 节所述里程碑。

该样本包还可以扩展。后续可以加入一个真实 FITS 版本、一个谱线 cube 小样本、一个偏振/RM synthesis 小样本，或者一个包含真实 pipeline weblog 摘要的归档复盘包。扩展时应保持同一结构：manifest、配置、产品、任务、报告和量规。


### 9.36.6 本节结论

真实轻量样本包把第 9 章的可复现实践链落到文件系统中。它把数据产品、清单、参数、任务、报告和量规组织成可检查材料，使处理判断不只停留在正文说明。虽然该样本包刻意保持小规模，但它已经能训练低面亮度上限、局部噪声、残差结构、源表边界、限制声明和报告复查。这是从教材文本走向课程项目和研究训练材料的关键一步。
