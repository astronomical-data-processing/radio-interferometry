## 7.9 系统温度、等效系统通量密度、观测日志与质量控制

观测系统章节前面已经讨论了接收链路、相关器、主波束、极化、传播效应和射频干扰。真实数据处理还需要一个把仪器状态和后续图像质量连接起来的桥梁：系统温度、等效系统通量密度、权重、观测日志和质量控制。它们决定一组数据在理论上应达到什么灵敏度，也决定最终图像偏离预期时应从哪里寻找原因。

现代 pipeline 的质量报告常把这些信息折叠成少数图和标志，但教材需要把背后的物理关系讲清楚。系统温度不是一个工程附录；它进入可见度权重、热噪声、灵敏度估计和误差预算。观测日志也不是行政记录；它记录天气、天线状态、校准源扫描、系统温度异常、人工标记和维护事件，是图像伪影和校准问题的重要证据来源。


![系统温度噪声预算](figures/system_tsys_noise_budget.png)

图 7.9.1 展示系统温度的组成。天空背景、接收机、大气和地面溢出共同决定噪声预算。系统温度直接进入热噪声尺度，因此也进入第 9 章的质量控制和误差预算。


### 7.9.1 系统温度的物理含义

系统温度 $T_{\rm sys}$ 用温度单位表示接收系统看到的总噪声功率。一个简化分解为

$$
T_{\rm sys}=T_{\rm rx}+T_{\rm sky}+T_{\rm atm}+T_{\rm spill}+T_{\rm loss},
$$

其中 $T_{\rm rx}$ 是接收机噪声，$T_{\rm sky}$ 包括银河背景和天体背景，$T_{\rm atm}$ 表示大气发射与吸收带来的等效噪声，$T_{\rm spill}$ 表示地面和旁瓣溢出贡献，$T_{\rm loss}$ 表示损耗项。不同频段中主导项不同：低频常受天空背景和 RFI 环境影响，高频常受大气透明度、水汽和指向误差影响。

系统温度随时间、频率、天线和仰角变化。若某一段时间 $T_{\rm sys}$ 明显升高，数据权重应降低；若某些天线或频窗系统温度异常，校准解和最终图像可能出现局部系统误差。只看最终图像噪声而不看 $T_{\rm sys}$，很难判断噪声升高来自天气、RFI、标记比例、权重错误还是成像参数。


### 7.9.2 等效系统通量密度与灵敏度

系统温度可以通过有效面积转成等效系统通量密度：

$$
S_{\rm sys}=\frac{2kT_{\rm sys}}{A_{\rm eff}},
$$

其中 $k$ 是 Boltzmann 常数，$A_{\rm eff}$ 是天线有效面积。该量把温度噪声转换为通量密度尺度。对理想化的双极化连续谱观测，热噪声量级可写成

$$
\sigma_S \simeq \frac{S_{\rm sys}}{\eta_s\sqrt{n_{\rm pol}\,\Delta\nu\,t}},
$$

其中 $n_{\rm pol}$ 是极化数，$\Delta\nu$ 是有效带宽，$t$ 是有效积分时间，$\eta_s$ 表示系统效率和权重等影响。对干涉阵，还需要把基线数、权重、flag 比例和不同天线灵敏度纳入等效估计。这个公式的意义在于建立数量级判断：给定系统状态、带宽和时间，图像噪声应大致落在哪里。


![等效系统通量密度与灵敏度](figures/system_sefd_sensitivity_chain.png)

图 7.9.2 展示从系统温度、有效面积到通量灵敏度的链路。灵敏度估计必须说明带宽、积分时间、极化数和权重。实测噪声偏离预期时，应回到这些量检查。


### 7.9.3 观测日志是质量证据

观测日志应记录时间、扫描、目标场、校准源、天气、系统温度、天线状态、标记比例、权重变化和人工异常说明。它的价值不是事后回忆，而是把图像伪影定位到观测条件。相位校准不稳定可能来自天气恶化，某些频窗噪声升高可能来自系统温度或 RFI，某些基线异常可能来自天线维护或阴影遮挡，主波束边缘伪检可能与 pointing 或权重有关。


![观测日志质量证据板](figures/system_observing_log_evidence_board.png)

图 7.9.3 展示观测日志如何形成质量证据板。时间与扫描、天气、系统温度、标记比例、权重变化和人工记录共同进入后续质量判断。没有日志，很多图像伪影只能被猜测；有日志，质量判断才能定位到观测条件。


### 7.9.4 从日志到质量控制

从日志进入质量控制需要三步。第一步是读取和整理观测条件，记录系统温度、天气、天线状态和校准源覆盖。第二步是审计权重和标记比例，检查是否有某些时间、频率、天线或基线被系统性削弱。第三步是比较预期灵敏度与实测噪声，判断偏差来自热噪声、残余 RFI、校准误差、主波束、动态范围限制还是成像选择。


![观测日志到质量控制](figures/system_log_to_qa_decision.png)

图 7.9.4 展示从观测日志到质量控制的决策过程。关键日志缺失时，结论应停止或降级；系统项主导时，报告应把限制写入误差预算，而不是只给热噪声。


### 7.9.5 教学案例：高频谱线观测的天气日志

设一组高频分子线观测在成像后噪声高于理论预期，且相位残差在后半段明显增大。若只看图像，可能会误以为成像参数不当；若查看观测日志，发现后半段水汽增加、系统温度升高、相位校准间隔不足，则质量判断会完全不同。此时应把后半段数据权重、flag、校准解稳定性和最终噪声一起审计。若重成像后仍受天气限制，报告应说明该 cube 的灵敏度受系统状态主导，某些弱线非探测只能给较宽上限。

这个案例说明，系统温度和日志不是独立于科学分析的附属材料。它们决定数据是否达到预期、是否需要降权或标记、是否可以支持某个上限，以及哪些结论需要被写入限制声明。


### 7.9.6 本节结论

系统温度、等效系统通量密度和观测日志把观测系统与后续数据质量连接起来。系统温度决定热噪声尺度，等效系统通量密度把温度噪声转成通量灵敏度，日志把天气、天线、标记、权重和人工异常变成可审查证据。第 9 章中的 pipeline QA、weblog 判读、归档复盘和误差预算，都需要这条仪器到质量控制的桥。
