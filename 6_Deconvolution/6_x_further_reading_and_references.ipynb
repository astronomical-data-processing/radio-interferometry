## 6.x 延伸阅读与参考文献 <a id='deconv:sec:refs'></a>


第 6 章的重点不只是掌握一次 `CLEAN` 命令，而是把去卷积理解为一个带有先验、停止准则、质量评估和目录化后续步骤的反问题流程。本节按“概念主线、经典文献、现代算法、软件实践、源搜索”几个层次组织延伸阅读。阅读这些材料时，最有价值的问题不是某个软件参数应当怎样记忆，而是每个参数背后改变了哪一个模型假设、哪一部分残差结构或哪一种计算折中。

### 6.x.1 本章主线回顾

第 6 章可以概括为一条从脏图像到科学目录的链条：脏图像和 PSF 给出反问题的观测形式；天空模型给出可允许的解空间；minor / major cycle 在模型域和数据一致性之间迭代；恢复图像和残图用于判断解是否可信；源搜索和参数拟合把图像进一步转化为可检验的科学量。

`6.1` 和 `6.2` 的核心是点源先验为何能把病态反问题变成可执行算法。`6.3` 的核心是区分算法思想和工程实现：Högbom、Clark、Cotton-Schwab、多尺度、多频率方法并不是互斥菜单，而是在不同数据规模、视场大小、谱结构和计算资源下形成的不同折中。`6.4` 和 `6.5` 则决定图像能否真正用于科学分析；残图、局部噪声、clean bias、伪影和源拟合误差若缺少诊断，后续目录会把成像问题悄悄转化为科学偏差。

### 6.x.2 教材与讲义

[<cite data-cite='taylor1999synthesis'>Synthesis Imaging in Radio Astronomy II</cite> &#10548;](http://adsabs.harvard.edu/abs/1999sira.conf.....T) 是理解综合成像传统方法的经典讲义，适合与本章并读，用来把教学示例接回真实阵列的数据处理流程。*Interferometry and Synthesis in Radio Astronomy*（Thompson, Moran, Swenson）是系统性最强的参考书，适合查阅公式、近似条件和工程限制。NRAO 的 [Essential Radio Astronomy &#10548;](https://science.nrao.edu/opportunities/courses/era) 对去卷积细节展开较少，但擅长把图像、噪声、角分辨率、亮温和动态范围放回完整射电观测语境。*Tools of Radio Astronomy* 更偏向仪器和观测系统背景，可用于补足接收机、噪声来源和观测误差对成像结果的影响。

### 6.x.3 CLEAN 家族与现代去卷积

[<cite data-cite='1974A&AS...15..417H'>Högbom (1974)</cite> &#10548;](http://adsabs.harvard.edu/abs/1974A%26AS...15..417H) 是经典图像域 `CLEAN` 的起点，最直接地展示了“点源字典 + 逐步减去 PSF”的思想。[<cite data-cite='1980A&A....89..377C'>Clark (1980)</cite> &#10548;](http://adsabs.harvard.edu/abs/1980A%26A....89..377C) 将 `CLEAN` 推向大图像和大数据量下可执行的工程实现。[<cite data-cite='1984AJ.....89.1076S'>Schwab (1984)</cite> &#10548;](http://adsabs.harvard.edu/abs/1984AJ.....89.1076S) 进一步明确了 major / minor cycle 与可见度一致性的关系。

展源和宽带成像需要更丰富的模型先验。[<cite data-cite='2008ISTSP...2..793C'>Cornwell (2008)</cite> &#10548;](http://arxiv.org/abs/0806.2228) 的多尺度 `CLEAN` 说明天空先验可以从纯点源推广到有限尺度分量。[<cite data-cite='2011A&A...532A..71R'>Rau & Cornwell (2011)</cite> &#10548;](http://arxiv.org/abs/1106.2745) 的多尺度多频率联合去卷积是宽带连续谱成像的重要入口。[<cite data-cite='2014MNRAS.444..606O'>Offringa et al. (2014)</cite> &#10548;](http://arxiv.org/abs/1407.1943) 以 `WSClean` 展示了现代成像器如何把宽视场、并行化和多种去卷积策略放进统一软件。[<cite data-cite='2015A&A...576A...7D'>Dabbech et al. (2015)</cite> &#10548;](http://arxiv.org/abs/1412.5387) 则提供了进入稀疏重构和更一般反问题方法的自然接口。

### 6.x.4 宽视场、模型先验与联合推断

[<cite data-cite='1992A&A...261..353C'>Cornwell & Perley (1992)</cite> &#10548;](http://adsabs.harvard.edu/abs/1992A%26A...261..353C) 是宽视场成像和非共面基线问题的经典文献，有助于理解去卷积为什么不能脱离成像几何。[<cite data-cite='2002ApJ...570..447C'>Chang & Refregier (2002)</cite> &#10548;](http://arxiv.org/abs/astro-ph/0107085) 用 shapelet 基进行源建模，展示了点源和高斯之外的函数基思路。[<cite data-cite='2015MNRAS.450.1308L'>Lochner et al. (2015)</cite> &#10548;](http://arxiv.org/abs/1501.05304) 从贝叶斯建模和不确定性传播角度提醒去卷积不只是得到一张较干净的图像，还涉及模型比较、误差刻画和后验推断。

本章目前只建立这些方向的入口，尚未系统展开方向相关去卷积、联合校准-成像、偏振宽带联合重建、mosaic / single-dish 联合去卷积等高级主题。这些内容可以视为第 6 章的自然延伸：它们仍然围绕同一个核心问题，即在不完整、有噪声、有系统误差的观测约束下，怎样用合适的先验恢复可解释的天空模型。

### 6.x.5 源搜索、局部噪声与目录构建

[PyBDSF 文档 &#10548;](https://pybdsf.readthedocs.io/en/latest/) 可直接对应 `6.5` 的许多概念，包括均值图、RMS 图、Gaussian 拟合、Gaussian 分组、shapelet / wavelet 描述以及偏振和谱指数相关量。[AegeanTools 文档 &#10548;](https://aegeantools.readthedocs.io/en/stable/) 除了 `aegean` 本体，还包含 `BANE` 等局部背景和局部噪声估计工具，特别适合作为局部阈值方法的实践参照。Hancock et al. (2012) 的 [*Compact continuum source-finding for next generation radio surveys* &#10548;](https://arxiv.org/abs/1202.4500) 讨论下一代巡天条件下的完整率、可靠率和参数恢复问题；Hancock et al. (2018) 的 [*Aegean 2.0* &#10548;](https://arxiv.org/abs/1801.05548) 进一步展示源搜索器如何适应更大规模图像和更多科学场景。[Hydra II: source finder comparison &#10548;](https://arxiv.org/abs/2304.14357) 这类基准测试则有助于比较 `PyBDSF`、`Aegean`、`Selavy` 等软件在不同源形态和噪声条件下的偏差与盲点。

源搜索最关键的问题并不是阈值取几倍 RMS，而是背景是否平稳、局部 RMS 是否可信、相邻源是否需要 deblending、拟合模型是否与合成波束一致、目录参数是否已经考虑主波束和系统误差。成像质量评估和源目录构建因此应被放在同一个闭环中理解。

### 6.x.6 软件与实践路径

[CASA `tclean` 文档 &#10548;](https://casadocs.readthedocs.io/en/latest/api/tt/casatasks.imaging.tclean.html) 把 `hogbom`、`multiscale`、`mtmfs`、`auto-multithresh`、`wproject` 等选项放在同一任务框架内，是理解参数组合与去卷积策略关系的重要资料。[WSClean deconvolution 手册 &#10548;](https://wsclean.readthedocs.io/en/latest/deconvolution.html) 系统整理了 `auto-masking`、`local RMS thresholding`、`wideband deconvolution`、`multi-scale cleaning`、`MORESANE` 和压缩感知相关模式，适合与 `6.3` 到 `6.5` 对照阅读。

实践上，比较稳妥的路径是先复现 `6.2` 到 `6.4` 的合成示例，并能解释每一次迭代为何停止、残图里还剩下什么；随后改变源形态、噪声水平、阈值和 PSF 宽度，观察算法何时开始失效；最后转向真实连续谱图像，用独立源搜索器生成目录，并把目录与残图、RMS 图和波束参数一起交叉检查。只追求把图像“清干净”容易过度去卷积，只追求源表数量又容易把伪影写入目录。成熟的数据处理习惯，是把去卷积、图像质量评估和源搜索作为同一个推断流程的三个环节。
