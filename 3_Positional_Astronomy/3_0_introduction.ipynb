# 第 3 章：干涉测量中的位置天文学<a id='pos:sec:intro'></a>

第 2 章已经准备了球面三角、向量与傅里叶语言；本章要把这些工具接到真实观测几何上。对射电干涉阵而言，“天源在天上哪里”并不是一个孤立的天文学问题，而是一条连续的变换链：星表给出赤经和赤纬，观测时刻给出地方恒星时，地方恒星时决定时角，时角与台站纬度共同决定望远镜在本地天空中的方位角与高度角；随后，我们再围绕相位中心引入方向余弦坐标，为下一章的可见度与成像方程做准备。

本章并不试图把位置天文学扩展成完整的天体测量学课程。诸如岁差、章动、光行差、地球定向参数、极移和折射改正等内容，在高精度 VLBI 或精密定轨中非常重要，但它们通常由专业软件与标准服务统一处理。本章的目标，是先把干涉测量中最常反复出现的几何主线建立清楚：目录位置、观测时刻、台站位置、相位中心和局部像面坐标之间究竟如何对应。


#### 本章学习目标

* 理解赤道坐标为何是射电源目录、校准源列表和相位中心定义的出发点；
* 掌握时角与地方恒星时的物理含义，知道它们如何决定过中天时刻和跟踪几何；
* 熟悉赤道坐标到地平坐标的转换，并能用它讨论可观测性、最低仰角和观测时间窗；
* 理解方向余弦坐标 `(l,m,n)` 与相位中心、局部天图以及小视场近似之间的关系；
* 为第 4 章中的 `u,v,w` 几何和可见度积分建立必要的坐标直觉。


#### 本章结构

本章沿着一条明确的观测流程组织。第 3.1 节首先讨论赤道坐标以及目录位置的参考架和历元问题，说明为什么天源在数据库中通常以 `RA/Dec` 形式出现。第 3.2 节转向时间几何：一旦给定观测时刻，地方恒星时与时角就决定了该天源相对于本地子午圈的位置，也预告了地球自转合成为什么能够改变阵列投影。

第 3.3 节再把视角从“天球上的固定目录坐标”转移到“台站本地的指向坐标”，即方位角与高度角，并用它讨论可观测性、仰角限制以及观测窗口。最后，第 3.4 节围绕相位中心引入方向余弦坐标 `(l,m,n)`，把天球几何变成局部像面几何，为第 4 章的可见度空间和成像方程作出直接铺垫。


#### 坐标链条总览

* **目录位置**：天源首先以赤经 $\alpha$ 和赤纬 $\delta$ 记录在参考星表或射电源表中；
* **时间几何**：给定观测时刻后，地方恒星时 `LST` 与时角 `H` 决定该源相对于本地子午圈的位置；
* **望远镜指向**：台站纬度 $\varphi$ 与 $(H,\delta)$ 一起决定本地天空中的方位角 $\mathcal{A}$ 和高度角 $\mathcal{E}$；
* **局部成像坐标**：选定相位中心后，方向余弦 `(l,m,n)` 描述视场中心附近的局部角偏移；
* **可见度几何**：下一章会把基线投影写成 `u,v,w`，并把它与 `(l,m,n)` 联系起来。
